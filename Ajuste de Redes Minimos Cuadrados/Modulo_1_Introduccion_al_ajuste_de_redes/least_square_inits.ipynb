### Método de los Mínimos Cuadrados

Hoy en día, los métodos tradicionales de ajuste de mediciones son inadecuados para levantamientos de precisión que involucran grandes mediciones redundantes (para verificar errores groseros en las mediciones), diferentes técnicas de medición (como en levantamientos de triangulación y trilateración), y precisiones de medición variables. Por ejemplo, los métodos tradicionales muestran claramente sus debilidades cuando se ajustan redes de polígonos con amarres cruzados y en el ajuste de redes GNSS (Sistema Global de Navegación por Satélite) posprocesados donde se involucran muchas observaciones. El método de ajuste que puede acomodar todas estas propiedades y evaluar mediciones redundantes al mismo tiempo es el método de los mínimos cuadrados, que generalmente se refiere como estimación de mínimos cuadrados o ajuste de mínimos cuadrados.

El método de los mínimos cuadrados proporciona un tratamiento organizado y matemáticamente lógico de errores aleatorios de observación e inconsistencias en modelos matemáticos. Utiliza un cierto enfoque de distribución de errores al aplicar correcciones a las observaciones. Las características principales del ajuste de mínimos cuadrados son ajustar las observaciones para eliminar posibles cierres defectuosos y mantener los cambios en las observaciones lo más pequeños posible (según lo que se conoce como el criterio de mínimos cuadrados).

### Historia del Método de los Mínimos Cuadrados

Los conceptos del método de los mínimos cuadrados han evolucionado a lo largo de los años. A continuación se presenta un resumen de algunos de los aspectos importantes de la historia del método:

- En 1788, **Laplace** (científico francés) utilizó el método de promedios para combinar diferentes datos recopilados bajo las mismas condiciones.
- **Boscovich** en 1757 y **Laplace** en 1799 sugirieron el uso del método de desviación absoluta mínima para combinar diferentes datos recopilados bajo diferentes condiciones.
- En 1795, 1806 y 1809, **Legendre** (matemático francés) y **Gauss** (científico alemán) propusieron el concepto de optimización de la suma de cuadrados de correcciones a las mediciones. El nombre "método de los mínimos cuadrados" fue publicado por primera vez por Legendre (1806), pero la descripción detallada del método fue proporcionada por Gauss (1809), quien afirmó que había desarrollado el método de los mínimos cuadrados de forma independiente desde 1795. Legendre y Gauss, sin embargo, se entiende que abordaron el método de los mínimos cuadrados de manera diferente:
  - **Legendre** utilizó modelos matemáticos para eliminar cierres defectuosos y proporcionó consistencia numérica de las observaciones conectadas por modelos.
  - **Gauss** asumió la normalidad de las mediciones para producir resultados con la mayor probabilidad de ser ciertos. Descubrió que los errores observacionales siguen la ley de la función de distribución de probabilidad normal, que utilizó para proporcionar propiedades numéricas y estadísticas de los parámetros desconocidos.
- En general, alrededor de 1800, **Legendre**, **Gauss** y **Laplace** pudieron ajustar modelos funcionales a los datos a través de diferentes tipos de métodos de mínimos cuadrados.
- **Adrian** (agrimensor estadounidense) (1808) formuló independientemente el principio de análisis de mínimos cuadrados, pero no se refirió a él por ese nombre. También sugirió el uso de la distribución normal para describir errores en las mediciones ya en 1808, incluso antes de Gauss.
- **Bessel** (astrónomo y geodesta alemán) en 1837 resolvió el problema del ajuste de mínimos cuadrados de modelos paramétricos determinando los elementos de la figura de la Tierra basados en las coordenadas geodésicas disponibles.
- **Helmert** (geodesta alemán) en 1907 determinó los coeficientes de transformación lineal por el método de los mínimos cuadrados paramétricos conocido como la transformación de Helmert.
- **Bjerhammar** (geodesta sueco) (1951) contribuyó a la simplificación del método de los mínimos cuadrados utilizando notaciones matriciales.

### Resumen

El método de los mínimos cuadrados es una técnica matemática utilizada para ajustar observaciones redundantes en levantamientos de precisión, eliminando errores y asegurando la consistencia de los datos. Este método ha evolucionado a lo largo de los años con contribuciones significativas de científicos como Laplace, Boscovich, Legendre, Gauss y otros. Su aplicación ha sido fundamental en geodesia y topografía, mejorando la precisión y fiabilidad de los levantamientos y modelos geodésicos.

### Criterio de los Mínimos Cuadrados

Para ilustrar algunas de las características del método de los mínimos cuadrados, considere un problema de resección en el cual las coordenadas $(x, y)$ del punto $P$ deben determinarse. En este problema, los ángulos horizontales $(\alpha, \beta, \gamma)$ y las distancias horizontales $(d_1, d_2, d_3)$ con sus desviaciones estándar asociadas se miden desde el punto $P$ a tres puntos de control $A, B$ y $C$ cuyas coordenadas son bien conocidas.

Las seis mediciones en la Figura 4.2 deben ajustarse para hacerlas coincidir en la determinación de las dos coordenadas desconocidas $(x, y)$ del punto $P$. En este caso, la suma de los ángulos en el punto $P$ sobre el horizonte debe ser igual a $360^\circ$, y las correcciones a las mediciones de distancia deben ser lo más pequeñas posible; todas estas condiciones deben cumplirse simultáneamente. Las correcciones (o residuos) a aplicar a las mediciones dependen de cuán bien las mediciones se ajusten entre sí. Un enfoque para resolver las coordenadas del punto $P$ es adivinar las coordenadas del punto $P$ utilizando las coordenadas de los puntos $A, B$ y $C$ como guía, luego calcular independientemente las mediciones, y utilizarlas para determinar los residuos. El método de los mínimos cuadrados corrige iterativamente las coordenadas iniciales adivinadas para que la suma ponderada de las correcciones a las mediciones sea un valor mínimo. Esto significa que el método de los mínimos cuadrados está diseñado para mantener los cambios (o ajustes) a las mediciones originales lo más pequeños posible. Esto lleva a lo que se conoce como el criterio de mínimos cuadrados o la condición de mínimos cuadrados. El criterio general de mínimos cuadrados o condición de mínimos cuadrados establece que la suma de los cuadrados ponderados de las correcciones a las mediciones es un mínimo. Esto se puede expresar matemáticamente como:

$$ v^T Pv = \text{mínimo} $$

donde $ v $ es el vector de residuos de observación (correcciones a aplicar a las mediciones después del ajuste o las diferencias entre las mediciones reales y los valores predichos por el modelo) y $ P $ es la matriz de pesos de las observaciones derivadas de las desviaciones estándar de las observaciones. Si la matriz de pesos es diagonal (con elementos fuera de la diagonal como ceros), el criterio de mínimos cuadrados en la Ecuación (4.20) se puede reescribir como:

$$ v_1^2p_{11} + v_2^2p_{22} + v_3^2p_{33} + \cdots + v_n^2p_{nn} = \text{mínimo} $$

donde $ P = \text{diag}(p_{11}, p_{22}, p_{33}, \ldots, p_{nn}) $ es la matriz de pesos de las mediciones y $ v = (v_1, v_2, v_3, \ldots, v_n)^T $ es el vector de residuos, que debe determinarse. Solo las cantidades conocidas para la variable observable son las mediciones y sus pesos. El criterio de mínimos cuadrados se puede utilizar para determinar las incógnitas de la siguiente manera. El criterio de mínimos cuadrados se formula como en la Ecuación (4.21) de la siguiente manera:

$$ v_1^2p_{11} + v_2^2p_{22} + v_3^2p_{33} + \cdots + v_n^2p_{nn} = \text{mínimo} $$

El mínimo de la Ecuación (4.22) se encuentra estableciendo las derivadas parciales de la ecuación (con respecto a cada residuo de observación) a cero de la siguiente manera:

$$ \frac{\partial (v_1^2p_{11})}{\partial v_1} + \frac{\partial (v_2^2p_{22})}{\partial v_2} + \frac{\partial (v_3^2p_{33})}{\partial v_3} + \cdots + \frac{\partial (v_n^2p_{nn})}{\partial v_n} = 0 $$

o

$$ 2v_1p_{11} + 2v_2p_{22} + 2v_3p_{33} + \cdots + 2v_np_{nn} = 0 $$

Recordando que un residuo $(v_i)$ es la diferencia entre la medición $(\ell_i)$ y su valor promedio $(\bar{x})$, expresado como $ v_i = (\bar{x} - \ell_i) $, la Ecuación (4.24) se puede reescribir como

$$ (\bar{x} - \ell_1)p_{11} + (\bar{x} - \ell_2)p_{22} + (\bar{x} - \ell_3)p_{33} + \cdots + (\bar{x} - \ell_n)p_{nn} = 0 $$

La Ecuación (4.25) se puede resolver para dar

$$ \bar{x} = \frac{\ell_1p_{11} + \ell_2p_{22} + \ell_3p_{33} + \cdots + \ell_np_{nn}}{p_{11} + p_{22} + p_{33} + \cdots + p_{nn}} $$

Esto significa que la solución de mínimos cuadrados para las mediciones $\ell$ con una matriz de pesos $P$. Se puede afirmar, por lo tanto, que la estimación de mínimos cuadrados es simplemente un promedio ponderado de las observaciones para el parámetro desconocido. En este caso, todas las observaciones de alguna manera contribuyen al "mejor" parámetro estimado. El valor promedio ponderado es el valor más probable en este problema. El método ordinario de mínimos cuadrados se refiere como el mejor estimador lineal insesgado (BLUE) de los parámetros en relación con el modelo de Gauss-Markov. Un modelo de Gauss-Markov es un modelo de regresión lineal en el cual los errores no correlacionados tienen una expectativa cero e igual varianzas. El término "mejor" significa que el valor medio cuadrado más bajo del parámetro estimado se logra. También se sabe que cuando se minimiza la suma de residuos cuadrados ponderados, como en la Ecuación (4.21), el parámetro estimado es BLUE si cada peso es igual a la recíproca de la varianza de la medición. Si los errores están correlacionados, el estimador resultante es también BLUE si la matriz de pesos es igual a la inversa de la matriz de covarianza de la observación, como en el caso de la Ecuación (4.20). Las relaciones entre las desviaciones estándar de las mediciones, la matriz de varianza-covarianza, la matriz de cofactors, y la matriz de pesos fueron discutidas en el Capítulo 1. De ese capítulo, se puede ver que la matriz de pesos $P$ de las observaciones se puede dar como:

$$ P = \sigma_0^2 C_{\ell}^{-1} $$

donde $ C_{\ell} = \sigma_{\ell}^2 Q $ es la matriz de varianza-covarianza del vector de observación $\ell$ (con los cuadrados de las desviaciones estándar o varianzas de las observaciones como elementos diagonales), $\sigma_0^2$ es el factor de varianza a priori de unidad de peso (que siempre se toma como la varianza poblacional general con un valor de 1), y $Q$ es la matriz de cofactors del vector de observación.

### Resumen

El criterio de mínimos cuadrados se utiliza para ajustar las mediciones de manera que la suma de las correcciones ponderadas sea mínima. Este método es fundamental en la topografía y la geodesia para ajustar observaciones redundantes y obtener coordenadas precisas. Utiliza modelos matemáticos y estadísticos para proporcionar las estimaciones más probables de los parámetros desconocidos. El concepto de BLUE (Mejor Estimador Lineal Insesgado) es central en este enfoque, asegurando que las estimaciones sean las más precisas y consistentes posibles dadas las mediciones disponibles.

### Tipos de Modelos de Ajuste de Mínimos Cuadrados

Los conceptos de observables de levantamientos, ajustes, modelos funcionales y modelos estocásticos se discutieron en el Capítulo 1. Los modelos funcionales relacionan los observables de levantamientos (como distancias, direcciones, acimuts, ángulos y diferencias de elevación) con los parámetros desconocidos como las coordenadas de los puntos de la red de levantamiento. Antes de que se pueda realizar cualquier ajuste, se debe formular un modelo funcional que relacione los observables con los parámetros, o se debe identificar y formular un modelo funcional que relacione los observables entre sí.

La necesidad de ajuste de modelos se basa en la redundancia (o sobredeterminación) de los modelos; no habrá necesidad de ajuste si los modelos no están sobredeterminados. Dependiendo de cómo se relacionan estos observables con los parámetros desconocidos o entre sí, los modelos matemáticos formados pueden referirse como paramétricos, condicionales o generales. Los modelos paramétricos tienen un observable $(\ell)$ por ecuación, que es una función de algunos o todos los parámetros desconocidos $(x)$. Dado que el número de ecuaciones es el mismo que el número de observaciones, el número de grados de libertad (redundancia) es el número de ecuaciones (u observaciones) menos el número de parámetros. Los modelos condicionales relacionan geométricamente solo las observaciones $(\ell)$ juntas sin involucrar ningún parámetro en las ecuaciones. Dado que el número de ecuaciones independientes involucradas puede ser menor que el número de observaciones, el número de grados de libertad para el ajuste condicional es el mismo que el número de ecuaciones del modelo independiente formuladas. Los modelos generales de alguna manera tienen las observaciones $(\ell)$ entrelazadas con los parámetros $(x)$ en las mismas ecuaciones; el número de ecuaciones involucradas menos el número de parámetros es el número de grados de libertad. Sin embargo, se debe mencionar que el número de ecuaciones involucradas en el ajuste del modelo general no es necesariamente el mismo que el número de observaciones, como en el caso del ajuste del modelo paramétrico; generalmente es menor.

El caso especial de modelos generales en el cual cada parámetro desconocido es una función de las observaciones se refiere como modelo especial en este libro. Este tipo de modelo se formuló para problemas de propagación de errores en el Capítulo 2. Las relaciones entre los diferentes modelos se muestran en la Figura 4.3. Se presentan ejemplos de cada modelo en los capítulos apropiados que tratan sobre métodos de ajuste de mínimos cuadrados paramétricos, condicionales y generales. La convención que se seguirá en los siguientes métodos de ajuste de mínimos cuadrados es tal que los símbolos que representan observaciones no ajustadas y los parámetros no ajustados no tendrán sombreros, como $\ell$ para observaciones (cantidades medidas) y $x$ como parámetros no ajustados (cantidades que no se miden pero cuyos valores deben estimarse); las observaciones ajustadas se representarán con sombreros como $\hat{\ell}$, y los parámetros ajustados se representarán con sombreros como $\hat{x}$. Se debe enfatizar que siempre que se formule algún tipo de los diferentes modelos (paramétrico, condicional, general), el orden de las letras del modelo debe seguir el orden de los parámetros y observaciones ajustados en el mismo lado de las ecuaciones es muy importante. La idea detrás del uso de cantidades ajustadas (aquellos representados con "sombrero" en ellos) en la formulación de la ecuación del modelo es indicar que esas cantidades se van a ajustar y las ecuaciones así formadas son completas y consistentes, es decir, no producirán cierres defectuosos.

Dado que el álgebra matricial es conveniente para representar ecuaciones de modelo en la estimación de mínimos cuadrados, se utilizará a lo largo de este libro. Como forma de revisión, las matrices introductorias se presentan en el Capítulo 1, y algunos lemas importantes de matrices se proporcionan en el Apéndice E. Se debe mencionar aquí que siempre que se estén formulando ecuaciones de modelos matemáticos, los valores numéricos de las observaciones no deben usarse de inmediato, sino que deben recordarse los símbolos apropiados que representan esas mediciones, como $\hat{\ell}_1, \hat{\ell}_2, \Delta \hat{h}_1, \Delta \hat{h}_2$. Se debe recordar también que en la etapa de formulación de ecuaciones, uno no está resolviendo las ecuaciones sino simplemente presentándolas simbólicamente. También se debe notar que, en esta etapa, lo que es importante es cómo identificar parámetros desconocidos y observaciones. Los parámetros desconocidos $(x)$ son cantidades que uno quiere determinar, y las observaciones $(\ell)$ son las mediciones o cantidades proporcionadas con sus desviaciones estándar o matriz de covarianza que se utilizarán para determinar los parámetros desconocidos.

### Tabla de Tipos de Modelos de Ajuste de Mínimos Cuadrados

| Tipo de Modelo    | Descripción | Fórmula |
|-------------------|-------------|---------|
| Modelo General    | Los parámetros ajustados ($\hat{x}$) y las observaciones ajustadas ($\hat{\ell}$) están en el mismo lado de las ecuaciones. | $ f(\hat{x}, \hat{\ell}) = 0 $ |
| Modelo Paramétrico | Cada observación ($\hat{\ell}$) se representa por una ecuación que involucra los parámetros ajustados ($\hat{x}$). | $\hat{\ell} = f(\hat{x})$ |
| Modelo Condicional | Las observaciones ($\hat{\ell}$) se representan en una ecuación de acuerdo con las condiciones que satisfacen, sin involucrar parámetros. | $ f(\hat{\ell}) = 0 $ |
| Modelo Especial    | Relaciona las observaciones ajustadas ($\hat{\ell}$) con cada parámetro ($\hat{x}$) al determinar valores para los parámetros después del ajuste condicional o en la propagación de errores. | $ \hat{x} = f(\hat{\ell}) $ |

La idea detrás del uso de cantidades ajustadas (representadas con sombreros) en la formulación de la ecuación del modelo es indicar que esas cantidades se ajustarán y las ecuaciones así formadas son completas y consistentes.

### Pasos para el Ajuste de Mínimos Cuadrados

Después de que se hayan recopilado las mediciones de levantamiento, las mediciones deben corregirse para algunas condiciones físicas como reducciones de marca a marca, correcciones de refracción a ángulos cenitales, correcciones debido a deflexiones de la vertical para convertir el acimut astronómico a geodésico, corrección de escala, etc. Después de que se hayan aplicado estas reducciones y correcciones a las mediciones, se obtienen observaciones corregidas. Las observaciones corregidas no deben confundirse con las observaciones ajustadas. Las observaciones ajustadas son observaciones corregidas cuyos errores aleatorios se han minimizado a través de procesos de ajuste que involucran observaciones redundantes; las observaciones ajustadas producen los mejores valores de las cantidades desconocidas. Las observaciones reducidas y corregidas se evalúan utilizando la herramienta de mínimos cuadrados como se ilustra en la Figura 4.4.

Básicamente, el método de mínimos cuadrados requiere los siguientes pasos como se indica en la Figura 4.4:

- Formulación de algunos modelos matemáticos adecuados (paramétrico, condicional o general) y modelo estocástico (ver Capítulos 5, 12 y 13).
- Incorporación de los modelos en el criterio de mínimos cuadrados (formando lo que se conoce como función de variación) (ver Secciones 5.5, 12.3 y 13.4).
- Encontrar las derivadas parciales del criterio de mínimos cuadrados o la función de variación (formulada a partir del criterio de mínimos cuadrados), estableciendo las derivadas parciales a cero y resolviéndolas. El "mínimo" del criterio de mínimos cuadrados o de la función de variación se encuentra estableciendo las derivadas parciales del criterio de mínimos cuadrados o de la función de variación a cero (ver Secciones 5.5-5.7, 12.5 y 13.5).
- Detectar posibles errores que puedan degradar la solución utilizando un enfoque de prueba estadística (ver Capítulo 11).
- Calcular las soluciones finales y sus propiedades estadísticas (ver Secciones 5.7-5.8, 12.4 y 13.6-13.8).
- Evaluar los resultados del ajuste para determinar si se cumplen los estándares deseados.

El punto de partida del ajuste de mínimos cuadrados es la formulación de modelos matemáticos funcionales que relacionen observaciones y los parámetros desconocidos (las cantidades \( x \)) en los casos de ajustes de modelos paramétricos y generales o relacionen observaciones entre sí como en el caso del ajuste del modelo condicional. Los modelos funcionales en geomática pueden ser no lineales. El método de mínimos cuadrados, sin embargo, generalmente se realiza con ecuaciones lineales. Esto requiere que se adopten medios para linearizar modelos no lineales. El medio comúnmente utilizado para linearizar las ecuaciones en geomática es mediante el uso de los términos de primer orden de la expansión de series de Taylor como se discutió en el Capítulo 2. Dado que solo los términos lineales (términos de orden cero y primero) de las series de Taylor se utilizan para aproximar los modelos funcionales que se deben ajustar, se requiere que los valores aproximados de las cantidades desconocidas se determinen lo suficientemente cerca de los resultados finales del ajuste, es decir, las cantidades ajustadas. Si este no es el caso, entonces la expansión de series de Taylor no representará adecuadamente el modelo funcional original y los valores corregidos de los parámetros del ajuste deben usarse nuevamente (ahora como un nuevo conjunto de valores aproximados o iniciales) en el nuevo procedimiento de ajuste de mínimos cuadrados. Este procedimiento puede repetirse tantas veces como sea necesario hasta que la solución final no difiera significativamente de la anterior. Cuanto más cerca estén los valores iniciales de la solución final, menor será el número de iteraciones necesarias. Esto quiere decir que la elección de los valores iniciales de las cantidades desconocidas para ser determinadas en el ajuste de mínimos cuadrados es importante; sin embargo, no hay una forma especial de elegirlas. Generalmente, algo de experiencia o algunos medios computacionales simples y rápidos pueden ser utilizados para elegirlas.

### Restricciones en el Ajuste

Las **ecuaciones de restricción** son las ecuaciones necesarias para completar la definición de un datum. Esto es necesario ya que el datum para las coordenadas desconocidas a determinar está incompletamente definido por las mediciones. El tipo y el número de ecuaciones de restricción necesarias dependen del número de **parámetros de datum** que están indefinidos por las mediciones de la red que se están ajustando, es decir, el número de defectos de datum. Estos defectos deben ser atendidos formulando ecuaciones de restricción apropiadas. El número de ecuaciones de restricción requeridas en un ajuste depende del tipo de posicionamiento o coordenadas a determinar y también del número de parámetros de datum atendidos por las mediciones del levantamiento. Generalmente, algunos de los parámetros de datum atendidos por las mediciones de levantamiento son los siguientes:

- **Parámetro de escala**: Este se atiende mediante mediciones de distancia horizontal, mediciones de distancia de marca a marca de levantamientos GPS, mediciones de pendiente, o mediciones de diferencia de altura.
- **Parámetros de posición**: Estos se atienden proporcionando coordenadas apropiadas $(1D, 2D,3D)$ de un punto dependiendo de si se están determinando coordenadas $1D$, $2D$, o $3D$ en el ajuste.
- **Parámetros de orientación**: Estos se atienden midiendo o fijando un acimut (o rumbo) de una línea para el posicionamiento $2D$ $(x, y)$. Para la determinación $3D$ $(x, y, z)$, los parámetros de orientación se definirán midiendo un acimut y ángulos horizontales y verticales; acimut y diferencias de altura; acimut, distancias horizontales y diferencias de altura.

Se puede fijar una coordenada incluyendo una ecuación apropiada que especifique el valor a asignar a la coordenada, como:

$$ x + v_x = \hat{x} \tag{4.28} $$
$$ y + v_y = \hat{y} \tag{4.29} $$

donde las coordenadas medidas $(x, y)$ con sus correcciones $(v_x, v_y)$ se expresan como funciones de las coordenadas ajustadas $(\hat{x}, \hat{y}$) mismas. Cuando las coordenadas se expresan matemáticamente como en las Ecuaciones (4.28) y (4.29), las coordenadas se observan directamente, formando las **ecuaciones de restricción**. En este caso, siempre que se formen las ecuaciones paramétricas, es necesario agregar un conjunto de ecuaciones de restricción que completarán la definición del datum de coordenadas. Estas ecuaciones de restricción obligan a esos parámetros a cumplir las reglas definidas por las ecuaciones. La otra forma de aplicar estas ecuaciones de restricción en el ajuste es dejarlas fuera al formar la primera matriz de diseño; en este caso, se reduce el número de parámetros a determinar en el ajuste. Este enfoque se adopta a lo largo de este libro.

#### Ajustes de Restricción Mínima

Cuando el número y tipo de ecuaciones de restricción son suficientes para definir un datum dado, el ajuste resultante de tal sistema se llamará **ajuste de restricción mínima**. Este tipo de ajuste proporcionará coordenadas ajustadas que dependen únicamente de las observaciones y la elección arbitraria de los parámetros de datum. El ajuste de mínimos cuadrados será imposible sin al menos las ecuaciones de restricción mínima impuestas. En la práctica, se dice que una red de levantamiento tiene restricciones mínimas si cumple cualquiera de las siguientes condiciones:

1. En una **red de nivelación $1D$** (donde solo se miden diferencias de elevación), se puede fijar la elevación de un punto en la red (no sujeto a cambios) mientras que todos los puntos restantes de la red son libres de cambiar mediante el proceso de ajuste.
2. En una **red de triangulación $2D$** (donde solo se miden ángulos), se pueden fijar las coordenadas de dos puntos en la red (no sujetos a cambios) mientras que todos los puntos restantes de la red son libres de cambiar (no fijados). En este caso, las coordenadas de uno de los dos puntos son para definir la ubicación de la red; el rumbo y la distancia calculada entre los dos puntos son para proporcionar la orientación (rumbo) y la escala necesarias para completar la definición del datum.
3. En una **red de trilateración $2D$** (donde solo se miden distancias), se pueden fijar las coordenadas de un punto en la red (para definir la ubicación de la red); una coordenada (x o y) de otro punto de la red se puede fijar para completar el datum si las coordenadas aproximadas de los otros puntos de la red son "bien conocidas". En este caso, las mediciones de distancia proporcionarán la escala y las coordenadas fijadas, y las coordenadas aproximadas de los otros puntos de la red completarán la definición del datum.
4. En una **red combinada de triangulación $2D$ y trilateración $2D$** (también conocida como **red de triangulateración $2D$**) en la que se mide un acimut de una línea, se pueden fijar las coordenadas de un punto en la red (para definir la ubicación de la red); el acimut y la medición de ángulo definirán la orientación de la red, mientras que las mediciones de distancia proporcionarán la escala necesaria para que la red complete la definición del datum.
5. En una **red de poligonal $2D$** en la que se miden ángulos y distancias en las estaciones del polígono, el datum se define de manera similar a los casos de red de trilateración $2D$ y una red combinada de triangulación $2D$ y trilateración $2D$.
6. En una **red de poligonal $3D$** en la que se miden ángulos horizontales, ángulos cenitales, distancias de pendiente y diferencias de elevación, el datum se define fijando las coordenadas $3D$ de uno de los puntos de la red y el acimut de una línea.

Fijar cualquier otra coordenada en cualquiera de las redes anteriores sobrecargará la red, pero agregar más mediciones como ángulos, distancias e incluso acimuts no sobrecargará la red. Ajustar una red con las restricciones mínimas especificadas anteriormente se conoce como ajuste de restricción mínima. Por ejemplo, la solución de solo GPS (en la que se conocen las coordenadas ($x$, $y$, $z$) de un punto de referencia con los vectores de línea base ($dx$, $dy$, $dz$) utilizados) es una solución de restricción mínima; las coordenadas del punto de referencia proporcionan el parámetro de ubicación, y el vector de línea base proporciona tanto las orientaciones como los parámetros de escala. También se debe recordar que si no se establece un datum para una red, será imposible ajustar la red mediante el método de mínimos cuadrados sin importar la cantidad de observaciones redundantes involucradas.

### Ajustes Sobrerrestricciones y Con Restricciones de Peso

En el ajuste de restricción mínima, las coordenadas publicadas (si no se fijan en el ajuste) aún pueden moverse en una dirección (lo cual es normalmente inaceptable). Podría ser preferible asegurarse de que la red de control existente permanezca fija restringiendo (o fijando) sus coordenadas. Esto es común en la práctica, asumiendo que la calidad de la red de control es muy superior a la calidad de las nuevas observaciones que se agregan. Cuando una red está restringida a dos o más estaciones de control, se dice que está sobrerrestringida. Una red sobrerrestringida se ve obligada a acomodar tanto las mediciones, el control previamente establecido (errores en encuestas anteriores) como la nueva encuesta. En esta red, cualquier error en los parámetros fijos deformará las observaciones ajustadas y los parámetros ajustados. En otras palabras, cuando se utilizan más ecuaciones de restricción de las necesarias, el ajuste se considera como ajuste sobrerrestringido o simplemente como ajuste sobrerrestringido. En este caso, las cantidades ajustadas resultantes están restringidas a estar en el sistema de coordenadas adecuado y también a cumplir con las restricciones adicionales. Las elipses de error derivadas para los nuevos puntos, sin embargo, parecerán optimistas en el sentido de que los errores en los puntos de control no se consideran. Cada vez que una nueva encuesta se ajusta a un sistema de coordenadas existente utilizando puntos de control existentes, se realiza un ajuste sobrerrestringido (o simplemente restringido).

Los ajustes con restricciones tienen algunas ventajas sobre los ajustes de restricción mínima, que incluyen las siguientes:

- Colocar nuevas encuestas en un sistema de coordenadas más grande proporcionando el datum (origen y orientación) para la nueva encuesta.
- Proporcionar medios para detectar errores en la nueva encuesta; esto hace que los valores ajustados observados sean más precisos y las coordenadas ajustadas más precisas que las derivadas de un ajuste de restricción mínima.
- Controlar o amortiguar la acumulación del efecto de errores observacionales (accidentales) en las coordenadas ajustadas.

Algunas de las desventajas de los ajustes con restricciones incluyen las siguientes:

- Los puntos de control fijados (con sus coordenadas asumidas como correctas y sin errores) para el ajuste pueden ser inadecuados en el sentido de que pueden llevar a cierres defectuosos que pueden ser mucho mayores de lo esperado. Esto es generalmente el caso ya que los cierres defectuosos en tales ajustes generalmente se asocian con errores en las nuevas encuestas (que pueden estar equivocadas) en lugar de en los puntos de control fijados. La consecuencia de esta asociación es que la diferencia entre los valores observados y ajustados de las observaciones puede ser mucho mayor de lo que puede explicarse por el error observacional (accidental).
- Cada ajuste restringido en el cual los puntos de control se mantienen fijos (con coordenadas con varianzas cero) es subóptimo. Esto significa que los ajustes de mínimos cuadrados utilizando una matriz de pesos que no es inversamente proporcional a la matriz de covarianza de las observaciones no tienen la propiedad de varianza mínima (son subóptimos). Nótese que las coordenadas de los puntos de control fijados son pseudoobservaciones, pero sus matrices de covarianza no se utilizan en la ponderación, ya que tienen valores cero como sus elementos. Por ejemplo, una encuesta GPS de alta precisión puede distorsionarse o degradarse cuando se ajusta a puntos de control de menor calidad. En este caso, la matriz de covarianza de las coordenadas ajustadas puede no ser menor que la de las observaciones reales como debería ser. Al fijar puntos de control, los valores ajustados de las observaciones probablemente se vuelvan peores que los valores observados. Por ejemplo, la desviación estándar de una encuesta GPS de alta precisión puede ser de aproximadamente 1 ppm en puntos de control fijos basada en la red NAD83 original que es precisa en aproximadamente 1:300,000. Esto significa que las covarianzas de los puntos de control fijados deben considerarse en el ajuste para tener en cuenta los efectos de los errores de los puntos de control fijados.

Los ajustes con restricciones a menudo se dice que distorsionan las observaciones, ya que producen observaciones ajustadas que son más inciertas o son probablemente peores que los errores reales de las observaciones. Por esta razón, cuando los agrimensores procesan, por ejemplo, sus vectores de línea base GPS, generalmente prefieren el ajuste de restricción mínima al ajuste con restricciones, ya que los ajustes de restricción mínima no distorsionan las mediciones después de los ajustes. El enfoque más correcto para ajustar mediciones de encuestas es realizar un ajuste con restricciones de peso, que considera tanto los errores en las nuevas encuestas como en los puntos de control. Este enfoque, sin embargo, tiene el problema de incluir las coordenadas de los puntos de control para el ajuste, cambiando así sus valores después del ajuste. Dado que los agrimensores no desean que las coordenadas de los puntos de control sigan cambiando en cualquier ajuste, la opción más aceptable para los agrimensores es mantener los puntos de control fijos (definiendo el datum) y ajustar las nuevas encuestas relativas a ellos.

El tipo de ajuste que se realizará en las coordenadas de la encuesta de una nueva encuesta depende de los tipos de restricciones impuestas a las coordenadas de la estación de la encuesta. Las coordenadas de la estación de la encuesta se pueden definir como fijas (o absolutamente restringidas), restringidas por peso, o libres. Las coordenadas de la estación de la encuesta fijas (o absolutamente restringidas) tendrán coordenadas que permanecerán iguales después del ajuste ya que sus coordenadas no están abiertas a ningún ajuste. Estas estaciones se consideran pseudo-observaciones con valores de varianza cero o una matriz de cofactors con valores cero como sus elementos. Cuando las estaciones de la encuesta se fijan en un ajuste para definir un datum, se realiza el ajuste restringido con una matriz de cofactors. Las estaciones restringidas por peso (o ajustadas por pesos) tendrán coordenadas que son abiertas al ajuste en la medida limitada por las desviaciones estándar asociadas de las coordenadas de las estaciones; estas coordenadas de estaciones se pueden utilizar como pseudo-observaciones con sus desviaciones estándar asociadas como factor de ponderación. Las desviaciones estándar de los puntos ponderados son para mejorar después del ajuste y pueden también mejorar otros parámetros a través de la correlación. El ajuste de estaciones ponderadas también se conoce como ajuste de mínimos cuadrados Bayesianos.

Normalmente, una estimación de mínimos cuadrados basada en coordenadas iniciales asociadas con una matriz de cofactors (ajuste ponderado) dará como resultado coordenadas ajustadas que son casi idénticas a las obtenidas por el ajuste de restricción mínima; los valores ajustados basados en este método están en algún lugar entre los valores derivados utilizando el método de restricción y aquellos basados en el método de restricción mínima. Esto se puede interpretar como que los datos estimados previamente en un ajuste no distorsionan significativamente las nuevas mediciones. En este caso, la geometría (posición relativa de los puntos) se define por las mediciones, y las restricciones de peso no afectan la geometría. Cuando las coordenadas de las estaciones de la encuesta son desconocidas y están abiertas a un ajuste ilimitado o no restringido, las estaciones se denominan estaciones libres.
