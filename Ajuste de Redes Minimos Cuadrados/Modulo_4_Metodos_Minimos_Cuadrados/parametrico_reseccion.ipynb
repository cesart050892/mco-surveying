### 6.5 Ejemplo de Resección

Un punto $ P (x_P, y_P) $ debe ser reseccionado desde tres puntos de control A, B y C, como se muestra en la Figura 6.10, con las coordenadas de los puntos de control A, B y C, como se muestra en la Tabla 6.20. Las mediciones de campo, dadas en la Tabla 6.21, consisten en tres distancias $ d_1, d_2, d_3 $; tres ángulos $ \alpha_{PA}, \alpha_{PB}, \alpha_{PC} $ respectivamente; y un ángulo APB como $ \theta $. Forma las ecuaciones del modelo paramétrico para este problema y calcula las coordenadas ajustadas del punto P, asumiendo que las coordenadas aproximadas del punto P son $ x_P^0 = 391.165 \ \text{m}, \ y_P^0 = 405.029 \ \text{m}; $ cada ángulo tiene una desviación estándar de $ \pm 4''; $ el ángulo es preciso a $ \pm 6''; $ y las distancias son precisas a $ \pm 0.002 \ \text{m} $.

Formula las ecuaciones del modelo paramétrico $ \ell = f(x) $ (ignorando los sombreros) de la siguiente manera:

$$
\alpha_{PA} = \text{atan} \left( \frac{x_A - x_P}{y_A - y_P} \right) + 360^\circ \tag{6.50}
$$

$$
\alpha_{PB} = \text{atan} \left( \frac{x_B - x_P}{y_B - y_P} \right) + 180^\circ \tag{6.51}
$$

$$
\alpha_{PC} = \text{atan} \left( \frac{x_C - x_P}{y_C - y_P} \right) + 180^\circ \tag{6.52}
$$

$$
\theta = \alpha_{PB} - \alpha_{PA} + 360^\circ = \text{atan} \left( \frac{x_B - x_P}{y_B - y_P} \right) + 180^\circ - \left[ \text{atan} \left( \frac{x_A - x_P}{y_A - y_P} \right) + 360^\circ \right] + 360^\circ
$$

o

$$
\theta = \text{atan} \left( \frac{x_B - x_P}{y_B - y_P} \right) - \text{atan} \left( \frac{316.682 - x_P}{505.015 - y_P} \right) + 180^\circ \tag{6.53}
$$

$$
d_1 = \sqrt{(x_A - x_P)^2 + (y_A - y_P)^2} \tag{6.54}
$$

$$
d_2 = \sqrt{(x_B - x_P)^2 + (y_B - y_P)^2} \tag{6.55}
$$

$$
d_3 = \sqrt{(x_C - x_P)^2 + (y_C - y_P)^2} \tag{6.56}
$$

```python
import sympy as sp

# Definir las variables
xA, yA, xB, yB, xC, yC, xP, yP = sp.symbols('xA yA xB yB xC yC xP yP')

# Formar las ecuaciones paramétricas (6.50) - (6.56)
Apa = sp.atan((xA - xP) / (yA - yP)) + 2 * sp.pi
Apb = sp.atan((xB - xP) / (yB - yP)) + sp.pi
Apc = sp.atan((xC - xP) / (yC - yP)) + sp.pi
theta = sp.atan((xB - xP) / (yB - yP)) - sp.atan((xA - xP) / (yA - yP)) + sp.pi

d1 = sp.sqrt((xA - xP)**2 + (yA - yP)**2)
d2 = sp.sqrt((xB - xP)**2 + (yB - yP)**2)
d3 = sp.sqrt((xC - xP)**2 + (yC - yP)**2)

# Determinar la Jacobiana con respecto a los parámetros desconocidos xP e yP
J = sp.Matrix([Apa, Apb, Apc, theta, d1, d2, d3]).jacobian([xP, yP])

# Valores dados y evaluar la Jacobiana como matriz A
xA_val, yA_val = 316.682, 505.015
xB_val, yB_val = 500.005, 400.040
xC_val, yC_val = 356.804, 310.032
xP_val, yP_val = 391.165, 405.029

A = J.subs({xA: xA_val, yA: yA_val, xB: xB_val, yB: yB_val, xC: xC_val, yC: yC_val, xP: xP_val, yP: yP_val})

# Determinar las observaciones calculadas L0 y las observaciones dadas L, y formar el vector de cierre w
L0 = sp.Matrix([Apa, Apb, Apc, theta, d1, d2, d3]).subs({xA: xA_val, yA: yA_val, xB: xB_val, yB: yB_val, xC: xC_val, yC: yC_val, xP: xP_val, yP: yP_val})

L = sp.Matrix([
    323 + 19/60 + 10/3600,
    92 + 37/60 + 15/3600,
    199 + 53/60 + 0/3600,
    129 + 18/60 + 15/3600,
    124.679,
    108.957,
    101.015
]) * sp.pi / 180  # Convertir ángulos a radianes

w = L0 - L

# Formar la matriz de pesos P y la solución δ
P = sp.diag(1/2.6591e9, 1/2.6591e9, 1/2.6591e9, 1/1.1818e9, 1/2.5e5, 1/2.5e5, 1/2.5e5)

N = A.T * P * A
u = A.T * P * w
d = -N.inv() * u

# Determinar los parámetros ajustados xPa e yPa
xPa = xP_val + d[0]
yPa = yP_val + d[1]

# Determinar los residuos v, varianza a posteriori S02
v = A * d + w
S02 = (v.T * P * v) / 5  # Grados de libertad = 7 - 2

# Determinar la matriz cofactor Qx y la covarianza Cx de los parámetros ajustados
Qx = N.inv()
Cx = S02[0] * Qx

# Mostrar los resultados
# sp.pprint(A)
sp.pprint(w)
# sp.pprint(N)
# sp.pprint(u)
# sp.pprint(d)
# print(f"xPa: {xPa.evalf()}")
# print(f"yPa: {yPa.evalf()}")
# sp.pprint(S02)
# sp.pprint(Qx)
# sp.pprint(Cx)
```
