### 12 Ajuste de Mínimos Cuadrados de Modelos Condicionales

#### 12.1 Introducción

Como se discutió en la Sección 4.3, las ecuaciones matemáticas involucradas en el campo de la geomática no siempre se formulan solo como ecuaciones de modelo paramétrico ($\hat{\ell} = f(\hat{x})$), donde $\hat{\ell}$ es un vector de observaciones y $\hat{x}$ es un vector de parámetros ajustados. A veces, uno puede no estar interesado inmediatamente en usar las mediciones de campo para resolver cualquier parámetro (como las coordenadas de puntos de red); sino que uno puede estar interesado solo en evaluar las mediciones por su consistencia entre sí sin tener que resolver ningún parámetro. Por ejemplo, considere un caso en el que se miden tres lados de un triángulo y sus tres ángulos interiores de forma independiente; se espera que los ángulos interiores medidos sumen 180° con los lados interiores y sus correspondientes lados medidos del triángulo satisfaciendo algunas reglas de seno y coseno. Es muy probable que los ángulos interiores medidos no sumen 180° y los ángulos y los lados correspondientes del triángulo no satisfagan las reglas de seno y coseno perfectamente. Este es un problema típico de ajuste de mínimos cuadrados que no implica resolver ningún parámetro (como las coordenadas de los puntos), sino que implica formular ecuaciones que conecten observaciones juntas matemáticamente. Las ecuaciones formuladas de esta manera se denominan ecuaciones de modelo condicional. Un modelo condicional, por lo tanto, es un modelo que contiene solo observaciones ($\hat{\ell}$) y ningún parámetro y puede expresarse matemáticamente como

$$ f(\hat{\ell}) = 0 \tag{12.1} $$

En otras palabras, las observaciones son las mismas que los parámetros en los modelos condicionales. Los ejemplos de ecuaciones de modelo condicional se dan en la Sección 12.2.

#### 12.2 Ecuaciones de Modelo Condicional

Las ecuaciones de modelo condicional se formulan de tal manera que todas las mediciones y las condiciones geométricas impuestas se representan y acomodan de manera única. Una vez que se han satisfecho estas, no se deben agregar más ecuaciones innecesarias. Una ecuación se considerará innecesaria si alguna combinación de ecuaciones ya formadas puede reproducir esa ecuación. Formar más ecuaciones de las necesarias puede resultar en que las ecuaciones se vuelvan imposibles de resolver. Dado que no hay parámetros involucrados en los modelos condicionales, el número de ecuaciones condicionales es el mismo que el número de redundancias para el problema. Esto significa que la redundancia o el número de grados de libertad de las ecuaciones de modelo condicional se determina como el número de ecuaciones de modelo independientes formadas. Formar ecuaciones únicas es usualmente el aspecto más difícil del ajuste de mínimos cuadrados de modelo condicional. Los pasos involucrados en la formulación de ecuaciones de modelo condicional son:

1) Identificar qué constituye las observaciones de la pregunta. Note que los datos dados se identifican como observaciones cuando las desviaciones estándar están asociadas con ellos o cuando los datos se refieren como medidos u observados.

2) A partir de la comprensión del tipo de modelo $(f(\hat{\ell}) = 0)$ a formular, las ecuaciones deben incluir solo las observaciones y algunos valores constantes (si es necesario); no debe haber ningún parámetro desconocido involucrado. Cada ecuación también debe arreglarse de tal manera que se establezca igual a cero; y debe asegurarse que cada ecuación sea única (basada en relaciones matemáticas sensatas que ocurran entre las mediciones) y no puede formularse combinando (sumando o restando o haciendo ambas cosas) cualquiera de las ecuaciones ya formadas.

Un sistema lineal de ecuaciones se llama sistema de ecuaciones linealmente dependiente si una o más de las ecuaciones es redundante, es decir, si al menos una de las ecuaciones es una combinación lineal de las otras. Un sistema es linealmente dependiente si una de las ecuaciones puede eliminarse sin afectar las soluciones generales. Si ninguna de las ecuaciones puede obtenerse como una combinación lineal de las otras, entonces el sistema se llama sistema de ecuaciones linealmente independiente. Los siguientes ejemplos son para ilustrar cómo formar ecuaciones de modelo condicional únicas e independientes, que excluyen ecuaciones innecesarias que harán que el problema sea irresoluble.

### 12.2.1 Ejemplos de Ecuaciones de Modelo

Se dan los siguientes ejemplos para ilustrar cómo se formulan las ecuaciones de modelo condicional.

**Ejemplo 12.1** Se realizaron mediciones de distancia entre los puntos A, B y C como se muestra en la Figura 12.1, donde $\hat{\ell}_1$, $\hat{\ell}_2$ y $\hat{\ell}_3$ son las distancias ajustadas a determinar y las distancias medidas son 190.40, 203.16 y 393.65 m, respectivamente. Forme el modelo condicional $f(\hat{\ell}) = 0$ para este problema.

**Solución:**
Las ecuaciones de modelo condicional deben tener la forma $f(\hat{\ell}) = 0$. Siguiendo los pasos dados en la Sección 12.2, se forma la siguiente ecuación:

$$ \hat{\ell}_3 - (\hat{\ell}_1 + \hat{\ell}_2) = 0 \tag{12.2} $$

donde los sombreros en las mediciones indican que las mediciones deben ajustarse; son necesarias para la consistencia de la ecuación, es decir, para que la ecuación no tenga misclosura.

**Ejemplo 12.2** En la Figura 12.2, los tres ángulos horizontales medidos alrededor del horizonte son $ \alpha = 22°12'13'' $, $ \beta = 59°56'15'' $, $ \gamma = 277°51'35'' $. Forme una ecuación adecuada $f(\hat{\ell}) = 0$ para ajustar las observaciones.

**Solución:**
Las ecuaciones de modelo condicional deben tener la forma $f(\hat{\ell}) = 0$. Siguiendo los pasos dados en el Ejemplo 12.1, la relación matemática sensata entre las observaciones es que todos los ángulos en un círculo deben sumar 360. Esto se representa matemáticamente como

$$ \hat{\alpha} + \hat{\beta} + \hat{\gamma} - 360° = 0 \tag{12.3} $$

**Ejemplo 12.3** Dado un circuito de nivelación cerrado en la Figura 12.3 con diferencias de elevación medidas para cuatro secciones como $\Delta h_1$, $\Delta h_2$, $\Delta h_3$, $\Delta h_4$, dé el modelo condicional para este problema.

**Solución:**
En un circuito de nivelación cerrado, la suma de las diferencias de elevación (siguiendo la dirección de nivelación indicada por las flechas) debe ser igual a cero. La dirección de la flecha para la diferencia de elevación $\Delta h_1$ se utilizará como la dirección de referencia alrededor del circuito (cualquier dirección puede usarse como referencia siempre que se mantenga la consistencia, cualquier diferencia de elevación en la misma dirección se representará como positiva, y cualquier dirección opuesta se representará como negativa). El modelo condicional $f(\hat{\ell}) = 0$ se da como

$$ \Delta h_1 - \Delta h_2 + \Delta h_3 + \Delta h_4 = 0 \tag{12.4} $$

**Ejemplo 12.4** Las mediciones de ángulo y distancia para una red de triangulación simple en la Figura 12.4 son las siguientes:

Ángulos: $\ell_1, \ell_2, \ell_3$

Distancias: $\ell_4, \ell_5$

Formule el modelo condicional para la red, suponiendo que la distancia $b$ es conocida y fija.

**Solución:** (considerando la condición de redundancia en el paso 2 de la Sección 12.2):

Usando las leyes del seno en las mediciones del triángulo y reorganizando para que parezca $f(\hat{\ell}) = 0$ (notando que cualquier otra ecuación será redundante),

$$ \hat{\ell}_1 + \hat{\ell}_2 + \hat{\ell}_3 - 180 = 0 \tag{12.5} $$

$$ b \sin \hat{\ell}_2 - \hat{\ell}_4 \sin \hat{\ell}_3 = 0 \tag{12.6} $$

$$ b \sin \hat{\ell}_1 - \hat{\ell}_5 \sin \hat{\ell}_3 = 0 \tag{12.7} $$

### Ejemplo 12.5

En la red de nivelación mostrada en la Figura 12.5, las elevaciones $h_A$ y $h_B$ de los puntos A y B, respectivamente, deben determinarse. Las elevaciones de referencia (en m) se muestran entre paréntesis, y las mediciones de campo se muestran en la figura como $dh_1$, $dh_2$, $dh_3$, $dh_4$ y $dh_5$ con las flechas indicando las direcciones de nivelación.

Dé el modelo matemático $(f(\hat{\ell}) = 0)$ para este problema, donde $\hat{\ell}$ es un vector de observaciones ajustadas.

**Solución:**

Tome nota de lo siguiente al formular las ecuaciones:

- Se puede ver en la red de nivelación en la Figura 12.5 que solo hay dos incógnitas (elevaciones de los puntos A y B), necesitando solo dos mediciones (diferencia de elevación a A y a B) de las cinco mediciones para determinar de manera única las dos incógnitas. Las tres mediciones restantes son redundantes, lo que significa que se deben formular tres ecuaciones condicionales.

- La suma de las diferencias de elevación entre dos puntos de referencia conocidos debe resultar en la diferencia entre las elevaciones de los dos puntos de referencia (siguiendo una flecha particular como valor positivo y la flecha opuesta como valor negativo).

- Siguiendo la ruta de nivelación BM1-A-B-BM3 ($dh_3$ tiene una flecha opuesta por lo que se le asigna negativo), se obtiene lo siguiente:

$$ \hat{dh}_1 - \hat{dh}_3 + \hat{dh}_4 - (794.881 - 785.232) = 0 \tag{12.8} $$

- Siguiendo la ruta de nivelación BM4-B-A-BM2 (todas las flechas están en la misma dirección), se obtiene lo siguiente:

$$ \hat{dh}_2 + \hat{dh}_3 + \hat{dh}_5 - (805.410 - 801.930) = 0 \tag{12.9} $$

- Aunque todas las mediciones se han tenido en cuenta en las Ecuaciones (12.8) y (12.9), aún es necesario asegurarse de que los circuitos BM2-A-BM1 y BM4-B-BM3 se cierren después del ajuste; si uno de estos circuitos se satisface junto con las condiciones establecidas en las Ecuaciones (12.8) y (12.9), el otro circuito se satisfará automáticamente. La condición para fijar el circuito BM2-A-BM1 se puede dar como sigue:

$$ \hat{dh}_1 + \hat{dh}_2 - (805.410 - 785.232) = 0 \tag{12.10} $$

o la condición para fijar el circuito BM4-B-BM3 como

$$ \hat{dh}_5 + \hat{dh}_4 - (794.881 - 801.930) = 0 \tag{12.11} $$

Solo hay tres ecuaciones únicas que acomodan todas las mediciones. El número de ecuaciones condicionales puede determinarse aproximadamente como el número de mediciones menos la cantidad de cantidades no medidas cuyos valores se están determinando implícitamente. En este problema, hay cinco mediciones ($dh_1, dh_2, dh_3, dh_4, dh_5$) y dos cantidades desconocidas ($h_A$ y $h_B$) no medidas cuyos valores se están determinando implícitamente, dando cinco menos dos ecuaciones condicionales (o tres ecuaciones condicionales). Las tres ecuaciones condicionales simplificadas de las Ecuaciones (12.8) a (12.10) se pueden dar como sigue:

$$ \hat{dh}_1 - \hat{dh}_3 + \hat{dh}_4 - 9.649 = 0 \tag{12.12} $$

$$ \hat{dh}_2 + \hat{dh}_3 + \hat{dh}_5 - 3.480 = 0 \tag{12.13} $$

$$ \hat{dh}_1 + \hat{dh}_2 - 20.178 = 0 \tag{12.14} $$

o de las Ecuaciones (12.8), (12.9) y (12.11):

$$ \hat{dh}_1 - \hat{dh}_3 + \hat{dh}_4 - 9.649 = 0 \tag{12.15} $$

$$ \hat{dh}_2 + \hat{dh}_3 + \hat{dh}_5 - 3.480 = 0 \tag{12.16} $$

$$ \hat{dh}_5 + \hat{dh}_4 + 7.049 = 0 \tag{12.17} $$

Como puede ver arriba, solo las Ecuaciones (12.14) y (12.17) son diferentes en las ecuaciones condicionales; las dos ecuaciones no deben aparecer juntas para no causar problemas al resolver las ecuaciones. Las ecuaciones condicionales deben ser únicas (independientes entre sí). Por ejemplo, se puede ver que la Ecuación (12.12) más la Ecuación (12.13) menos la Ecuación (12.14) da exactamente la Ecuación (12.17); por lo tanto, no hay necesidad de agregar la Ecuación (12.17) nuevamente a las tres ecuaciones. De manera similar, la Ecuación (12.15) más la Ecuación (12.16) menos la Ecuación (12.17) da exactamente la Ecuación (12.14). Considerando las Ecuaciones (12.8) a (12.11), la Ecuación (12.11) se dice que es redundante en el sentido de que proporciona información que ya está contenida en las otras ecuaciones.

Otra forma de formular las ecuaciones de modelo condicional para el problema en la Figura 12.5 es la siguiente:

- Siga la ruta de nivelación BM1-A-BM2 para obtener la siguiente ecuación:

$$ \hat{dh}_1 + \hat{dh}_2 - (805.410 - 785.232) = 0 \tag{12.18} $$

- Siga la ruta de nivelación BM4-B-BM3 para obtener la siguiente ecuación:

$$ \hat{dh}_5 + \hat{dh}_4 - (794.881 - 801.930) = 0 \tag{12.19} $$

- Dado que solo hay una observación más ($dh_3$) por satisfacer, cualquier ruta que acomode esta observación será aceptable. Siguiendo la ruta BM1-A-B-BM3 se obtendrá la siguiente ecuación:

$$ \hat{dh}_1 - \hat{dh}_3 + \hat{dh}_4 - (794.881 - 785.232) = 0 \tag{12.20} $$

Las Ecuaciones (12.18) a (12.20) son las ecuaciones condicionales linealmente independientes que se pueden usar en lugar de las dadas en las Ecuaciones (12.12) a (12.17). Note que cualquiera de estos tres conjuntos de ecuaciones producirá resultados idénticos en el ajuste de mínimos cuadrados de las ecuaciones.

### Ejemplo 12.6

En un tránsito cerrado mostrado en la Figura 12.6, se midieron los siguientes valores: ángulos $\theta_1$ y $\theta_2$; distancias $d_1$, $d_2$ y $d_3$; y rumbos $ \alpha_1 $ y $ \alpha_f $. Si las coordenadas del punto 1 $(x_1, y_1)$ y del punto 4 $(x_4, y_4)$ están fijas, forme las ecuaciones de modelo condicional necesarias $(f(\hat{\ell}) = 0)$ para las mediciones.

**Solución:**

Todos los ajustes de tránsito tienen un máximo de tres condiciones a satisfacer, de la siguiente manera:

1. La diferencia entre el rumbo directo $ \alpha_f $ y el rumbo directo calculado $ \alpha_f $ basado en el rumbo inicial $ \alpha_1 $ y los ángulos medidos en los puntos de tránsito debe ser igual a cero. Esto es para verificar la precisión de las mediciones angulares y puede expresarse matemáticamente como:

$$ \hat{\alpha}_f - (\hat{\alpha}_1 + \hat{\theta}_1 + \hat{\theta}_2) = 0 \tag{12.21} $$

2. La diferencia entre el cambio en las coordenadas este ($ x $) de los puntos de inicio y fin del tránsito y la suma de los cambios en las coordenadas este de las piernas individuales del tránsito debe ser igual a cero. Esto es para verificar tanto los errores en las distancias como los errores residuales en las mediciones angulares, a lo largo de la dirección este. Esto puede expresarse matemáticamente como:

$$ (x_4 - x_1) - d_1 \sin \hat{\alpha}_1 - d_2 \sin (\hat{\alpha}_1 + \hat{\theta}_1 + 180) - d_3 \sin (\hat{\alpha}_1 + \hat{\theta}_1 + \hat{\theta}_2) = 0 \tag{12.22} $$

3. La diferencia entre el cambio en las coordenadas norte ($ y $) de los puntos de inicio y fin del tránsito y la suma de los cambios en las coordenadas norte de las piernas individuales del tránsito debe ser igual a cero. Al igual que en el caso del este, esto también es para verificar tanto los errores en las distancias como los errores residuales en las mediciones angulares, a lo largo de la dirección norte. Esto también puede expresarse matemáticamente como:

$$ (y_4 - y_1) - d_1 \cos \hat{\alpha}_1 - d_2 \cos (\hat{\alpha}_1 + \hat{\theta}_1 + 180) - d_3 \cos (\hat{\alpha}_1 + \hat{\theta}_1 + \hat{\theta}_2) = 0 \tag{12.23} $$
