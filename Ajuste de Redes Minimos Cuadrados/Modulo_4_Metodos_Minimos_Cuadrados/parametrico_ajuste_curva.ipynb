### Matriz de pesos $(P)$ de las mediciones como las inversas de las varianzas de las mediciones:

```python
import sympy as sp

# Definir las variables
a, b = sp.symbols('a b')

# Formar las ecuaciones paramétricas
y = [
    a + 70.0002 * b,
    a + 219.9998 * b,
    a + 670.0132 * b,
    a + 970.0067 * b,
    a + 1469.9896 * b,
    a + 149.9996 * b,
    a + 600.0130 * b,
    a + 900.0066 * b,
    a + 1399.9894 * b,
    a + 450.0134 * b,
    a + 750.0070 * b,
    a + 1249.9898 * b,
    a + 299.9936 * b,
    a + 799.9765 * b,
    a + 499.9829 * b
]

# Evaluar la Jacobiana con respecto a a y b
J = sp.Matrix(y).jacobian([a, b])
A = J

# Datos proporcionados
L = sp.Matrix([
    70.0020, 219.9982, 670.0139, 970.0070, 1469.9906,
    149.9985, 600.0118, 900.0036, 1399.9921, 450.0116,
    750.0046, 1249.9868, 299.9925, 799.9759, 499.9839
])

# Valores aproximados
a0, b0 = 0, 1

# Calcular L0
L0 = sp.Matrix(y).subs({a: a0, b: b0})

# Vector de cierre w
w = L0 - L

# Formar la matriz de pesos P
P = sp.diag(*[1/std**2 for std in [
    0.00101, 0.00103, 0.00121, 0.00140, 0.00179,
    0.00102, 0.00110, 0.00112, 0.00173, 0.00111,
    0.00126, 0.00161, 0.00105, 0.00129, 0.00113
]])

# Calcular N y u
N = A.T * P * A
u = A.T * P * w

# Solución para δ
d = -N.inv() * u

# Parámetros ajustados
a_adj = a0 + d[0]
b_adj = b0 + d[1]

# Residuales y varianza a posteriori
v = A * d + w
S02 = (v.T * P * v)[0] / (len(L) - 2)

# Matriz cofactor Qx y matriz de covarianza Cx
Qx = N.inv()
Cx = S02 * Qx

# Mostrar resultados
A, w, N, u, d, a_adj.evalf(), b_adj.evalf(), S02, Qx, Cx
```
