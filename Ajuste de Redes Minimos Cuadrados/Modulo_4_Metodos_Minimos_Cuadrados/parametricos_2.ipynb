## 6.2.4 Ajuste de Triangulación

Una red de triangulación es una red de triángulos que combina redes de triangulación y trilateración en una sola red. En este caso, todos los ángulos en los triángulos que componen la red se miden como en un levantamiento de triangulación, y las longitudes de todos los lados de los triángulos se miden como en un levantamiento de trilateración. Para ilustrar cómo se puede usar el método de ajuste paramétrico de mínimos cuadrados para ajustar una red de triangulación, se usará un triángulo simple con el entendimiento de que el procedimiento será el mismo cuando se trate de una red más compleja. ### 6.3 Propiedades Estocásticas del Ajuste Paramétrico

El resumen de los modelos estocásticos para el ajuste paramétrico por mínimos cuadrados discutidos en el Capítulo 5 se presenta en esta sección. Los modelos se utilizan en los ejemplos que siguen.

1. Al comienzo de un ajuste paramétrico por mínimos cuadrados, las desviaciones estándar dadas de las mediciones se usan para formar la matriz de varianza-covarianza $(C_\ell)$ del vector de observación $\ell$. La matriz de varianza-covarianza formada y su matriz cofactor $(Q_\ell)$ están relacionadas por la Ecuación (6.39) como:
   
   $$
   C_\ell = \sigma_0^2 Q_\ell \tag{6.39}
   $$

   donde $\sigma_0^2 = 1$ es el factor de varianza a priori de peso unitario. En esencia, esto significa que la matriz de varianza-covarianza de las observaciones y su matriz cofactor son las mismas al comienzo de cada ajuste.

2. Durante o después del ajuste por mínimos cuadrados, se derivan las matrices cofactor de los parámetros ajustados $(Q_{\hat{x}})$, las observaciones ajustadas $(Q_{\hat{\ell}})$ y el vector residual $(Q_v)$. Estas cantidades derivadas se resumen en las Ecuaciones (6.40), (6.41) y (6.42) respectivamente:

   $$
   Q_{\hat{x}} = (A^T PA)^{-1} \tag{6.40}
   $$

   $$
   Q_{\hat{\ell}} = A (A^T PA)^{-1} A^T \tag{6.41}
   $$

   $$
   Q_v = Q_\ell - A (A^T PA)^{-1} A^T \tag{6.42}
   $$

   donde $A$ es la primera matriz de diseño y $P$ es la matriz de peso de las observaciones dadas como $P = Q_\ell^{-1}$.

3. Las siguientes matrices de varianza-covarianza de las cantidades ajustadas (Ecuaciones (6.43)–(6.45)) pueden darse a partir de las Ecuaciones (6.40), (6.41) y (6.42) respectivamente:

   $$
   C_{\hat{x}} = \sigma_0^2 (A^T PA)^{-1} \tag{6.43}
   $$

   $$
   C_{\hat{\ell}} = \sigma_0^2 A (A^T PA)^{-1} A^T \tag{6.44}
   $$

   $$
   C_v = \sigma_0^2 Q_\ell - \sigma_0^2 A (A^T PA)^{-1} A^T \tag{6.45}
   $$

   donde $C_{\hat{x}}$ es la matriz de covarianza de los parámetros ajustados, $C_{\hat{\ell}}$ es la matriz de covarianza de las observaciones ajustadas, $C_v$ es la matriz de covarianza de los residuos de las observaciones, y $\sigma_0^2$ es el factor de varianza a posteriori (APVF) de peso unitario, que se calcula después del ajuste por mínimos cuadrados como:

   $$
   \sigma_0^2 = \frac{v^T Pv}{n - u} \tag{6.46}
   $$

   con la redundancia expresada como $n - u$, $n$ el número de ecuaciones paramétricas, $u$ el número de parámetros independientes, $v$ el vector de residuos de las observaciones, y $P$ la matriz de peso de las observaciones.

4. La matriz de covarianza $(C_v)$ de los residuos de las observaciones puede reescribirse a partir de la Ecuación (6.45) como:

   $$
   C_v = \hat{C}_\ell - C_{\hat{\ell}} \tag{6.47}
   $$

   donde $\hat{C}_\ell$ es la matriz de covarianza escalada de las observaciones originales $\ell$, expresada como:

   $$
   \hat{C}_\ell = \sigma_0^2 Q_\ell \tag{6.48}
   $$

   con todos los símbolos definidos en el paso (3). La matriz de covarianza escalada $\hat{C}_\ell$ no debe confundirse con $C_\ell = \sigma_0^2 Q_\ell$ (con $\sigma_0^2 = 1$), que es la matriz de covarianza no escalada de las observaciones originales, utilizada al comienzo del ajuste por mínimos cuadrados. En realidad, $\hat{C}_\ell = \sigma_0^2 Q_\ell$ y $C_\ell = \sigma_0^2 Q_\ell$ serán idénticas cuando $\sigma_0^2 = \sigma_0^2$.
