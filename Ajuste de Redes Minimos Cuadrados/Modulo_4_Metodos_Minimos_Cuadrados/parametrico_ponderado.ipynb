### 6.7 Pasos para el ajuste con restricciones de peso

Los pasos para el método de ajuste paramétrico de mínimos cuadrados con restricciones de peso discutido en el Capítulo 5 se resumen en esta sección de la siguiente manera:

1) Proporcione el modelo paramétrico ($\ell = f(\hat{x})$) con $ P $ como la matriz de peso de las observaciones y $ C_x $ como la matriz de covarianza a priori del vector de todos los parámetros o de un sub-vector de los parámetros.

2) Forme la primera matriz de diseño ($ A $), que es la matriz que contiene las derivadas parciales de las ecuaciones paramétricas con respecto a los parámetros desconocidos. El tamaño de esta matriz será igual al número de ecuaciones (o número de observaciones) formando el tamaño de fila por el tamaño de la columna que representa el número de parámetros, incluidos aquellos parámetros asociados con la matriz de covarianza dada $ C_x $.

3) Elija valores aproximados para los parámetros ($ x^0 $); los valores de esos parámetros con la matriz de covarianza dada pueden usarse como valores aproximados para el ajuste. Por lo general, si los valores elegidos no están lo suficientemente cerca de lo que deberían ser los valores finales, los valores finales calculados pueden usarse nuevamente como valores aproximados en otra ronda de ajuste de manera iterativa hasta que los valores finales no sean demasiado diferentes de los valores aproximados (o iniciales). En este caso, el ajuste de mínimos cuadrados puede repetirse tantas veces como sea posible hasta que la solución final y la anterior converjan a un límite de tolerancia específico. Sustituya estos valores aproximados en las ecuaciones paramétricas formadas en el paso 1 para obtener los valores derivados o simulados de las observaciones ($\ell^0 = f(x^0)$).

4) Encuentre el vector de cierre ($ w = f(x^0) - \ell $) como la diferencia entre las observaciones derivadas ($\ell^0 = f(x^0)$) calculadas en el paso 3 y las observaciones reales ($\ell$).

5) Forme la matriz cofactor ($ Q $) de las observaciones (asumiendo que el factor de varianza a priori $\sigma_0^2$ es 1) utilizando las varianzas de las observaciones a lo largo de la diagonal de la matriz (en el orden en que se formaron las ecuaciones paramétricas en el paso 1). La inversa de esta matriz cofactor es la matriz de peso ($ P $) de las observaciones.

6) Tome la inversa ($ C_x^{-1} $) de la matriz de covarianza de los parámetros como el peso de los parámetros, $ P_x = C_x^{-1} $; alinee estos a lo largo de la diagonal de la matriz (en el orden en que se formaron las ecuaciones paramétricas en el paso 1); y asigne pesos cero a esos parámetros con una matriz de covarianza desconocida.

7) Determine la solución con restricción de peso de las correcciones a los parámetros aproximados de la siguiente manera:
$$ \delta = -(P_x + A^T P A)^{-1} A^T P w $$

8) Determine los parámetros ajustados: $\hat{x} = x^0 + \delta$.

9) Si es necesario, determine los residuales ($ v $) de las observaciones y las observaciones ajustadas ($\hat{\ell}$): $ v = A\delta + w $ y $ \hat{\ell} = \ell + v $.

10) Determine la matriz cofactor ($ Q_{\hat{x}} $) y la matriz de covarianza ($ C_{\hat{x}} $) de los parámetros ajustados, respectivamente, de la siguiente manera:
$$ Q_{\hat{x}} = (P_x + A^T P A)^{-1} $$
$$ C_{\hat{x}} = \sigma_0^2 Q_{\hat{x}} $$
donde el APVF de peso unitario $\sigma_0^2$ se puede dar de la siguiente manera:
$$ \sigma_0^2 = \frac{v^T P v + \delta^T P_x \delta}{n_d - u_d} $$
con $ n_d $ como el número de observaciones directas (o reales) y $ u_d $ como el número de parámetros directos (cuyos valores no tienen desviaciones estándar asociadas con ellos).

11) Determine la matriz cofactor ($ Q_{\hat{\ell}} $) y la matriz de covarianza ($ C_{\hat{\ell}} $) de las observaciones ajustadas, respectivamente, de la siguiente manera:
$$ Q_{\hat{\ell}} = A(P_x + A^T P A)^{-1} A^T $$
$$ C_{\hat{\ell}} = \sigma_0^2 Q_{\hat{\ell}} $$
donde el factor de varianza $\sigma_0^2$ se determina a partir de la Ecuación (6.60).

1. Las correcciones (±0.0188 m) a las coordenadas de los puntos ponderados FH1 y FH2 son mayores que las desviaciones estándar (±0.002 m) asociadas con los puntos. Este resultado, sin embargo, demuestra que solo los errores aleatorios que causan cierres en una red se ajustarán (o distribuirán) adecuadamente por el método de mínimos cuadrados.

Por lo tanto, el valor atípico debe eliminarse antes de que el método de mínimos cuadrados pueda ajustar adecuadamente las mediciones restantes (sin valores atípicos) para una solución correcta.
