### 12.3 Formulación del Ajuste del Modelo Condicional

Los pasos involucrados en el ajuste del modelo condicional se derivan en esta sección. Por ejemplo, considere el siguiente modelo condicional:

$$ f(\hat{\ell}) = 0 \quad P \tag{12.24} $$

donde $f$ es el vector del modelo matemático, $\hat{\ell}$ es el vector de observaciones ajustadas y $P$ es la matriz de pesos del vector de observaciones original $\ell$. Los pasos lógicos involucrados en la obtención del vector de solución para este modelo se pueden dar como sigue:

- Linearizar el modelo matemático si es no lineal.
- Formular la función de variación, que implica imponer el criterio de mínimos cuadrados en el modelo linearizado.
- Derivar las ecuaciones normales de mínimos cuadrados a partir de la función de variación.
- Resolver las ecuaciones normales de mínimos cuadrados para obtener las correcciones que se aplicarán a los valores aproximados de las cantidades desconocidas.

Para linearizar el modelo matemático en la Ecuación (12.24), asuma lo siguiente:

$$ \hat{\ell} = \ell + v \tag{12.25} $$

donde $v$ es un vector de residuos de observación que se aplicará al vector de observación original $\ell$. De las Ecuaciones (12.24) a (12.25), se puede obtener lo siguiente:

$$ f(\ell + v) = 0 \tag{12.26} $$

Mediante la linearización por expansión de la serie de Taylor, la Ecuación (12.26) en forma linearizada se puede dar como:

$$ f(\ell) + \frac{\partial f}{\partial \ell} v = 0 \tag{12.27} $$

La Ecuación (12.27) linearizada se puede expresar de forma más simple como:

$$ w + Bv = 0 \tag{12.28} $$

donde $B = \frac{\partial f}{\partial \ell}$ se refiere comúnmente como la segunda matriz de diseño y $w = f(\ell)$ es un vector de misclosures. La Ecuación (12.27) o la Ecuación (12.28) es una forma linearizada de la Ecuación (12.24).

La función de variación se puede formular de la siguiente manera. Dado que la matriz de pesos $P$ está directamente asociada con el modelo condicional en la Ecuación (12.24), el criterio de mínimos cuadrados para este problema será:

$$ v^T P v = \text{mínimo} \tag{12.29} $$

Imponer el criterio de mínimos cuadrados en el modelo linearizado en la Ecuación (12.28) basado en el enfoque de Lagrange para dar la siguiente función de variación:

$$ \varphi = v^T P v - 2k^T (Bv + w) = \text{mínimo} \tag{12.30} $$

donde $k$ es un vector de correlatos o un vector de multiplicadores de Lagrange. El sistema de ecuaciones normales de mínimos cuadrados se puede derivar encontrando las derivadas parciales de la función de variación en la Ecuación (12.30) con respecto a las incógnitas (es decir, $v$ y $k$) y estableciéndolas a cero de la siguiente manera:

$$ \frac{\partial \varphi}{\partial v} = 2v^T P - 2k^T B = 0 \tag{12.31} $$

$$ \frac{\partial \varphi}{\partial k} = -2(Bv + w) = 0 \tag{12.32} $$

El resumen de la forma más expandida de las ecuaciones normales (12.31) y (12.32) es:

$$ v^T P - k^T B = 0 \tag{12.33} $$

$$ Bv + w = 0 \tag{12.34} $$

Las Ecuaciones (12.33) y (12.34) se pueden resolver simultáneamente para las cantidades desconocidas $k$ y $v$ como sigue. De la Ecuación (12.33), se puede obtener lo siguiente:

$$ v = P^{-1} B^T k \tag{12.35} $$

Sustituyendo la Ecuación (12.35) en la Ecuación (12.34) para obtener:

$$ BP^{-1} B^T k + w = 0 \tag{12.36} $$

Si $Q = P^{-1}$ y $M = BQB^T$, se pueden derivar las siguientes funciones de solución de las Ecuaciones (12.36) y (12.35):

$$ k = -M^{-1} w \tag{12.37} $$

$$ v = QB^T k \tag{12.38} $$

De las Ecuaciones (12.25) y (12.38), se puede determinar la siguiente solución para la observación ajustada $(\hat{\ell})$:

$$ \hat{\ell} = \ell + v \tag{12.39} $$

### 12.3.1 Pasos de Ajuste del Modelo Condicional

El resumen de los pasos para el ajuste de mínimos cuadrados condicionales se da de la siguiente manera:

1. Formular las ecuaciones del modelo condicional:

$$ f(\hat{\ell}) = 0 \tag{12.40} $$

donde

$$ \hat{\ell} = \ell + v \tag{12.41} $$

2. Formular la segunda matriz de diseño ($B$) y el vector de misclosures $w$:

$$ B = \frac{\partial f}{\partial \ell} \quad \text{y} \quad w = f(\ell) \tag{12.42} $$

3. Formular las ecuaciones normales:

$$ BQB^T k + w = 0 \tag{12.43} $$

o

$$ Mk + w = 0 \tag{12.44} $$

donde

$$ M = BQB^T \tag{12.45} $$

$$ Q = P^{-1} \tag{12.46} $$

o

$$ Q = C_\ell \quad (\text{asumiendo} \; \sigma_0^2 = 1) \tag{12.47} $$

$Q$ es la matriz de cofactores, $C_\ell$ es la matriz de covarianza de las observaciones, $P$ es la matriz de pesos de las observaciones y $k$ es un vector de correlatos o vector de multiplicadores de Lagrange.

4. Resolver $k$ en las ecuaciones normales en la Ecuación (12.44):

$$ k = -M^{-1} w \tag{12.48} $$

5. Calcular el vector de residuos ($v$):

$$ v = QB^T k \tag{12.49} $$

6. Calcular las observaciones ajustadas como:

$$ \hat{\ell} = \ell + v \tag{12.50} $$

Los pasos de ajuste del modelo condicional dados anteriormente se ilustran numéricamente en los siguientes ejemplos.

### 12.4 Modelo Estocástico de Ajuste Condicional

El modelo estocástico para los ajustes condicionales se deriva como en el caso del ajuste del modelo paramétrico en la Sección 5.8. Recordemos de la Sección 1.7 y el Capítulo 5 la relación entre la matriz de varianza-covarianza ($C_\ell$) del vector de observaciones $\ell$ y la matriz de cofactores ($Q_\ell$) del vector de observaciones como:

$$ C_\ell = \sigma_0^2 Q_\ell \tag{12.55} $$

donde $\sigma_0^2$ es un factor de varianza a priori de unidad de peso (que siempre se toma como la varianza general de la población con un valor de 1.00). Dado que el vector de observaciones $\ell$ y su matriz de varianza-covarianza $C_\ell$ se usan para calcular las cantidades desconocidas, como las observaciones ajustadas ($\hat{\ell}$) y los residuos de observación ($v$), es obvio que los errores en esas observaciones se propagarán en estas cantidades también. Los conceptos de propagación de varianza-covarianza (Capítulo 2) se usarán para derivar la matriz de cofactores de las observaciones ajustadas ($Q_{\hat{\ell}}$) y la matriz de cofactores de los residuos de observación ($Q_v$).

#### 12.4.1 Derivación de la Matriz de Cofactores de Observaciones Ajustadas

El vector de observaciones ajustadas de mínimos cuadrados se da de las Ecuaciones (12.42), (12.48), (12.49) y (12.50) como sigue:

$$ \hat{\ell} = \ell - Q_\ell B^T M^{-1} f(\ell) \tag{12.56} $$

La Ecuación (12.56) es una forma de $\hat{\ell} = f(\ell)$ con la usual matriz de cofactores propagada (basada en las reglas de propagación de varianza-covarianza) dada como:

$$ Q_{\hat{\ell}} = JQ_\ell J^T \tag{12.57} $$

donde $Q_\ell$ es la matriz de cofactores de las observaciones y $J$ es la matriz Jacobiana dada como sigue:

$$ J = \frac{\partial \hat{\ell}}{\partial \ell} = I - Q_\ell B^T M^{-1} B \tag{12.58} $$

teniendo en cuenta que la derivada parcial del vector $\ell$ con respecto al vector $\ell$ da una matriz identidad $I$ y la derivada parcial de $f(\ell)$ con respecto a $\ell$ da la segunda matriz de diseño $B$ (basada en la Ecuación (12.42)). Sustituyendo la Ecuación (12.58) en la Ecuación (12.57) da lo siguiente:

$$ Q_{\hat{\ell}} = (I - Q_\ell B^T M^{-1} B) Q_\ell (I - Q_\ell B^T M^{-1} B)^T \tag{12.59} $$

Multiplicando el primer corchete por $Q_\ell$ y transponiendo los elementos en el otro corchete da la siguiente ecuación:

$$ Q_{\hat{\ell}} = (I - Q_\ell B^T M^{-1} B) Q_\ell (I - B^T M^{-1} B Q_\ell)^T \tag{12.60} $$

Expandiendo la Ecuación (12.60) da lo siguiente:

$$ Q_{\hat{\ell}} = Q_\ell - Q_\ell B^T M^{-1} B Q_\ell - Q_\ell B^T M^{-1} B Q_\ell + Q_\ell B^T M^{-1} (B^T Q_\ell B) M^{-1} B Q_\ell \tag{12.61} $$

Recordando que $M = (B^T Q_\ell B)$ y $(B^T Q_\ell B) M^{-1} = I$ reduce la Ecuación (12.61) a lo siguiente:

$$ Q_{\hat{\ell}} = Q_\ell - Q_\ell B^T M^{-1} B Q_\ell - Q_\ell B^T M^{-1} B Q_\ell + Q_\ell B^T M^{-1} B Q_\ell \tag{12.62} $$

Los últimos tres términos en la Ecuación (12.62) son idénticos, de modo que los últimos dos términos se cancelan, dando la forma reducida como sigue:

$$ Q_{\hat{\ell}} = Q_\ell - Q_\ell B^T M^{-1} B Q_\ell \tag{12.63} $$

La Ecuación (12.63) es la matriz de cofactores de las observaciones ajustadas $\hat{\ell}$.
