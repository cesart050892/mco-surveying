### Guía Completa para Entender el Ajuste por Mínimos Cuadrados  Aplicado al Problema de Angulos

#### **1. Contexto y Problema**
En topografía y otras disciplinas que dependen de mediciones precisas, es común que las observaciones no cumplan exactamente con las condiciones geométricas establecidas debido a errores en la medición. Estos errores pueden ser causados por diversos factores, como imprecisiones en los instrumentos o condiciones ambientales. El método de mínimos cuadrados es una técnica matemática utilizada para ajustar estas observaciones, buscando minimizar la suma de los cuadrados de los errores (o residuos), permitiendo obtener valores corregidos que cumplan con las condiciones geométricas de la manera más precisa posible.

#### **2. Paso a Paso del Ejemplo**

##### **Planteamiento Inicial**
Supongamos que estamos trabajando con la suma de tres ángulos que, por definición, debe ser $360^\circ$. Sin embargo, debido a errores de medición, la suma observada no es exacta:

$$
360^\circ = x + y + z = (42^\circ 12' 13'' + v_1) + (59^\circ 56' 15'' + v_2) + (257^\circ 51' 35'' + v_3)
$$

En esta ecuación:
- $x$, $y$, y $z$ representan los ángulos observados corregidos.
- $v_1$, $v_2$, y $v_3$ son los residuos o errores en las observaciones que debemos corregir para que la suma total sea exactamente $360^\circ$.

##### **Expresión del Residuo Dependiente**
Dado que sabemos que la suma total de los ángulos debe ser exactamente $360^\circ$, podemos expresar uno de los residuos ($v_3$) en función de los otros dos ($v_1$ y $v_2$):

$$
v_3 = -3'' - v_1 - v_2
$$

Este paso es crucial porque reduce el número de incógnitas en el sistema de ecuaciones, facilitando el proceso de ajuste.

##### **Función Objetivo**
El siguiente paso es formular una **función objetivo** que debemos minimizar. Esta función mide la magnitud de los errores o residuos, y se define como la suma de los cuadrados de los mismos:

$$
\Sigma v^2 = v_1^2 + v_2^2 + (-3'' - v_1 - v_2)^2
$$

Minimizar esta función significa encontrar los valores de $v_1$ y $v_2$ que minimicen el total de los errores al cuadrado, garantizando así que las correcciones aplicadas sean las más pequeñas posibles.

##### **Derivadas Parciales e Igualación a Cero**
Para encontrar los valores de $v_1$ y $v_2$ que minimizan la función objetivo, derivamos la función respecto a cada uno de los residuos y luego igualamos estas derivadas a cero:

$$
\frac{\partial \Sigma v^2}{\partial v_1} = 4v_1 + 2v_2 + 6'' = 0
$$
$$
\frac{\partial \Sigma v^2}{\partial v_2} = 2v_1 + 4v_2 + 6'' = 0
$$

Igualar a cero las derivadas parciales nos da un sistema de ecuaciones que describe las condiciones necesarias para que los residuos $v_1$ y $v_2$ minimicen la función objetivo.

##### **Resolución del Sistema de Ecuaciones**
A continuación, resolvemos el sistema de ecuaciones lineales resultante para obtener los valores específicos de $v_1$ y $v_2$:

$$
\begin{cases}
4v_1 + 2v_2 = -6'' \\
2v_1 + 4v_2 = -6''
\end{cases}
$$

La solución de este sistema nos da:

$$
v_1 = -1'' \quad \text{y} \quad v_2 = -1''
$$

Estos valores representan las correcciones necesarias que deben aplicarse a los ángulos $x$ y $y$ para que su suma, junto con $z$, sea exactamente $360^\circ$.

##### **Cálculo de $v_3$**
Una vez que tenemos los valores de $v_1$ y $v_2$, podemos calcular $v_3$ usando la expresión previamente establecida:

$$
v_3 = -3'' - (-1'') - (-1'') = -1''
$$

Este valor de $v_3$ ajusta el tercer ángulo $z$ para que la suma total de los ángulos sea $360^\circ$.

##### **Ángulos Ajustados**
Finalmente, sustituimos los valores de los residuos en las observaciones originales para obtener los ángulos ajustados:

$$
\begin{align*}
x & = 42^\circ 12' 13'' + v_1 = 42^\circ 12' 12'' \\
y & = 59^\circ 56' 15'' + v_2 = 59^\circ 56' 14'' \\
z & = 257^\circ 51' 35'' + v_3 = 257^\circ 51' 34''
\end{align*}
$$

Estos valores ajustados de $x$, $y$, y $z$ cumplen con la condición impuesta de que su suma debe ser $360^\circ$, minimizando al mismo tiempo los errores en las mediciones originales. Este proceso asegura que las correcciones aplicadas son las mínimas necesarias, lo que es el objetivo principal del ajuste por mínimos cuadrados.

### Guía Completa para Entender el Ajuste por Mínimos Cuadrados Aplicado al Problema de Distancias

#### **1. Contexto y Problema**

En este ejemplo, se nos presenta un problema clásico de ajuste por mínimos cuadrados aplicado a la medición de distancias. Tenemos tres distancias medidas entre los puntos A, B, y C. Debido a posibles errores en la medición, las distancias no cumplen perfectamente las condiciones geométricas establecidas. El objetivo es ajustar estas mediciones para minimizar los errores, asegurando que las distancias ajustadas sean lo más cercanas posible a las observaciones originales, pero cumpliendo con las condiciones impuestas.

#### **2. Paso a Paso del Ejemplo**

##### **Planteamiento Inicial**

Se nos dan las siguientes tres mediciones:
- La distancia total $x + y$ entre A y C debe ser 393.65 pies.
- La distancia $x$ entre A y B se mide como 190.40 pies.
- La distancia $y$ entre B y C se mide como 203.16 pies.

Estas mediciones están relacionadas por las ecuaciones:
$$
x + y = 393.65
$$
$$
x = 190.40
$$
$$
y = 203.16
$$

##### **Incorporación de Residuos**

Dado que sabemos que las mediciones pueden contener errores, incluimos los residuos $v_1$, $v_2$, y $v_3$ en las ecuaciones:

$$
\begin{align*}
x + y &= 393.65 + v_1 \\
x &= 190.40 + v_2 \\
y &= 203.16 + v_3
\end{align*}
$$

Aquí, $v_1$, $v_2$, y $v_3$ son los residuos que deben corregirse para que las observaciones sean consistentes con las condiciones geométricas impuestas.

##### **Función Objetivo**

El siguiente paso es formular una función objetivo que mida la suma de los cuadrados de los errores:

$$
\Sigma v^2 = (x + y - 393.65)^2 + (x - 190.40)^2 + (y - 203.16)^2
$$

Esta función mide el total de los errores al cuadrado. Minimizar esta función significa encontrar los valores de $x$ y $y$ que hagan que estos errores sean lo más pequeños posible.

##### **Derivadas Parciales e Igualación a Cero**

Para encontrar los valores de $x$ y $y$ que minimizan la función objetivo, derivamos la función con respecto a $x$ y $y$ y luego igualamos estas derivadas a cero:

$$
\frac{\partial \Sigma v^2}{\partial x} = 2(x + y - 393.65) + 2(x - 190.40) = 0
$$
$$
\frac{\partial \Sigma v^2}{\partial y} = 2(x + y - 393.65) + 2(y - 203.16) = 0
$$

Estas ecuaciones normales nos permiten resolver para los valores de $x$ y $y$ que minimizan los residuos.

##### **Resolución del Sistema de Ecuaciones**

Al resolver el sistema de ecuaciones normales:

$$
\begin{cases}
2(x + y - 393.65) + 2(x - 190.40) = 0 \\
2(x + y - 393.65) + 2(y - 203.16) = 0
\end{cases}
$$

Obtenemos:

$$
x = 190.43 \quad \text{y} \quad y = 203.19
$$

Estos son los valores ajustados de las distancias $x$ y $y$ que minimizan la suma de los errores cuadrados.

##### **Cálculo de Residuos**

Una vez que tenemos los valores ajustados de $x$ y $y$, calculamos los residuos para cada una de las observaciones:

$$
\begin{align*}
v_1 &= x + y - 393.65 = 190.43 + 203.19 - 393.65 = -0.03 \text{ pies} \\
v_2 &= x - 190.40 = 190.43 - 190.40 = 0.03 \text{ pies} \\
v_3 &= y - 203.16 = 203.19 - 203.16 = 0.03 \text{ pies}
\end{align*}
$$

Estos residuos representan las correcciones que deben aplicarse a las mediciones originales para que las condiciones geométricas se cumplan con el mínimo error cuadrático posible.

### **Conclusión**

Este ejemplo muestra cómo el ajuste por mínimos cuadrados permite corregir mediciones para que cumplan con las condiciones geométricas establecidas. Al minimizar la suma de los errores al cuadrado, obtenemos valores ajustados para $x$ y $y$ que son consistentes con la suma total requerida y que minimizan las desviaciones de las mediciones originales.
