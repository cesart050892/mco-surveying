### Ajuste de Mínimos Cuadrados No Lineales

El ajuste de mínimos cuadrados es una técnica matemática utilizada para encontrar la mejor aproximación de un conjunto de datos mediante la minimización de la suma de los cuadrados de las diferencias entre los valores observados y los valores predichos por un modelo. En el caso de ajustes no lineales, la relación entre las variables dependientes e independientes no es lineal y se representa mediante un modelo no lineal.

#### Series de Taylor y Linearización

Para resolver problemas de mínimos cuadrados no lineales, una técnica común es la linearización del modelo no lineal mediante series de Taylor. La idea es aproximar la función no lineal con una función lineal en torno a un punto de inicio, que se mejora iterativamente. Este proceso se puede resumir en los siguientes pasos:

1. **Función No Lineal:** Consideramos una función no lineal $ f(\mathbf{x}, \boldsymbol{\beta}) $, donde $\mathbf{x}$ son las variables independientes y $\boldsymbol{\beta}$ son los parámetros del modelo.

2. **Serie de Taylor:** La serie de Taylor de $ f(\mathbf{x}, \boldsymbol{\beta}) $ alrededor de un punto inicial $ \boldsymbol{\beta}_0 $ se puede escribir como:
   
   $$
   f(\mathbf{x}, \boldsymbol{\beta}) \approx f(\mathbf{x}, \boldsymbol{\beta}_0) + \nabla_{\boldsymbol{\beta}} f(\mathbf{x}, \boldsymbol{\beta}_0) (\boldsymbol{\beta} - \boldsymbol{\beta}_0)
   $$

   donde $ \nabla_{\boldsymbol{\beta}} f(\mathbf{x}, \boldsymbol{\beta}_0) $ es el gradiente de $ f $ respecto a $ \boldsymbol{\beta} $ evaluado en $ \boldsymbol{\beta}_0 $.

3. **Linearización:** Esta aproximación permite linearizar el problema de mínimos cuadrados no lineales, convirtiéndolo en un problema de mínimos cuadrados lineales para la iteración actual.

4. **Iteración:** El proceso iterativo implica resolver el problema linealizado, actualizar los parámetros $ \boldsymbol{\beta} $ y repetir hasta que la convergencia sea alcanzada.

### Formulación Matricial del Problema

1. **Función Residual:** Definimos una función residual $ \mathbf{r}(\boldsymbol{\beta}) $ tal que:

   $$
   \mathbf{r}(\boldsymbol{\beta}) = \mathbf{y} - \mathbf{f}(\mathbf{x}, \boldsymbol{\beta})
   $$

   donde $ \mathbf{y} $ son los valores observados y $ \mathbf{f}(\mathbf{x}, \boldsymbol{\beta}) $ son los valores predichos.

2. **Jacobian:** La matriz Jacobiana $ \mathbf{J} $ de $ \mathbf{f} $ respecto a $ \boldsymbol{\beta} $ se define como:

   $$
   \mathbf{J} = \frac{\partial \mathbf{f}}{\partial \boldsymbol{\beta}}
   $$

3. **Problema Linealizado:** En cada iteración, el problema de mínimos cuadrados se aproxima por:

   $$
   \min_{\boldsymbol{\beta}} \| \mathbf{r}(\boldsymbol{\beta}_0) + \mathbf{J} (\boldsymbol{\beta} - \boldsymbol{\beta}_0) \|^2
   $$

4. **Actualización de Parámetros:** La solución del problema anterior proporciona una actualización para los parámetros:

   $$
   \boldsymbol{\beta} \leftarrow \boldsymbol{\beta}_0 + \Delta \boldsymbol{\beta}
   $$

   donde $ \Delta \boldsymbol{\beta} $ es el incremento obtenido resolviendo el sistema linealizado.

Este procedimiento iterativo se repite hasta que los cambios en los parámetros $ \boldsymbol{\beta} $ sean suficientemente pequeños, indicando que se ha alcanzado la convergencia.

### Conclusión

El ajuste de mínimos cuadrados no lineales es una técnica poderosa para modelar datos cuando la relación entre las variables no es lineal. La utilización de series de Taylor y la linearización permiten transformar un problema complejo en una serie de problemas más manejables, facilitando la búsqueda de una solución óptima.

### Ecuación de la Distancia

La ecuación de la distancia entre dos puntos $ I $ y $ J $ en un plano es:

$$
(IJ) = \sqrt{(X_J - X_I)^2 + (Y_J - Y_I)^2}
$$

#### Expansión Linearizada

Para aplicar el ajuste de mínimos cuadrados, necesitamos linearizar esta ecuación utilizando la serie de Taylor. La expansión linearizada en torno a una estimación inicial $ (X_{I0}, Y_{I0}) $ y $ (X_{J0}, Y_{J0}) $ es:

$$
(IJ)_0 = \sqrt{(X_{J0} - X_{I0})^2 + (Y_{J0} - Y_{I0})^2}
$$

La expansión de Taylor de la ecuación de la distancia nos da la forma linearizada:

$$
\left( \frac{X_I - X_J}{IJ} \right)_0 dX_I + \left( \frac{Y_I - Y_J}{IJ} \right)_0 dY_I + \left( \frac{X_J - X_I}{IJ} \right)_0 dX_J + \left( \frac{Y_J - Y_I}{IJ} \right)_0 dY_J = k_{IJ} + v_{IJ}
$$

donde:

- $ dX_I $, $ dY_I $, $ dX_J $, $ dY_J $ son las correcciones a las coordenadas.
- $ k_{IJ} $ es una constante de corrección.
- $ v_{IJ} $ es el residuo de la medida.

### Derivadas Parciales

Para implementar la linearización en el ajuste de mínimos cuadrados, necesitamos las derivadas parciales de la ecuación de la distancia con respecto a las coordenadas $ X_I, Y_I, X_J, Y_J $.

Las derivadas parciales son:

$$
\frac{\partial (IJ)}{\partial X_I} = \left( \frac{X_I - X_J}{IJ} \right)
$$

$$
\frac{\partial (IJ)}{\partial Y_I} = \left( \frac{Y_I - Y_J}{IJ} \right)
$$

$$
\frac{\partial (IJ)}{\partial X_J} = \left( \frac{X_J - X_I}{IJ} \right)
$$

$$
\frac{\partial (IJ)}{\partial Y_J} = \left( \frac{Y_J - Y_I}{IJ} \right)
$$

Estas derivadas parciales se evalúan en las coordenadas iniciales $ (X_{I0}, Y_{I0}) $ y $ (X_{J0}, Y_{J0}) $.

### Implementación en el Ajuste

En el contexto del ajuste de redes topográficas, estas derivadas parciales se utilizan para construir la matriz de diseño $ \mathbf{A} $, donde cada fila representa una ecuación linearizada para una observación de distancia. La matriz $ \mathbf{A} $ y el vector de residuos $ \mathbf{v} $ se utilizan para resolver el sistema de ecuaciones linealizadas mediante el método de mínimos cuadrados.

#### Ejemplo de una Fila de la Matriz de Diseño

Para una observación de distancia $ IJ $, la fila correspondiente en la matriz de diseño $ \mathbf{A} $ tendría la forma:

$$
\mathbf{A}_{IJ} = \left[ \left( \frac{X_I - X_J}{IJ} \right)_0, \left( \frac{Y_I - Y_J}{IJ} \right)_0, \left( \frac{X_J - X_I}{IJ} \right)_0, \left( \frac{Y_J - Y_I}{IJ} \right)_0 \right]
$$

Y el vector de residuos sería:

$$
\mathbf{v}_{IJ} = k_{IJ} + v_{IJ}
$$

De esta manera, podemos construir el sistema de ecuaciones para el ajuste de mínimos cuadrados y resolver para las correcciones a las coordenadas $ \mathbf{dX} $ y $ \mathbf{dY} $.

### Conclusión

La linearización de la ecuación de la distancia y la obtención de las derivadas parciales son pasos cruciales en el proceso de ajuste de redes topográficas. Este enfoque permite transformar un problema no lineal en un sistema linealizado que se puede resolver iterativamente para obtener las mejores estimaciones de las coordenadas.

### Creación del Vector de Residuos

El vector de residuos, denotado como $ \mathbf{v} $, representa la diferencia entre las observaciones y las predicciones del modelo. En el contexto del ajuste de redes topográficas, el vector de residuos se construye utilizando las distancias observadas y las distancias calculadas a partir de las coordenadas iniciales.

#### Paso a Paso para Crear el Vector de Residuos

1. **Distancia Observada:**
   Para una observación de distancia entre dos puntos $ I $ y $ J $, denotamos la distancia observada como $ D_{IJ}^{obs} $.

2. **Distancia Calculada:**
   La distancia calculada entre los puntos $ I $ y $ J $ a partir de las coordenadas iniciales $ (X_{I0}, Y_{I0}) $ y $ (X_{J0}, Y_{J0}) $ se denota como $ D_{IJ}^{calc} $ y se calcula como:

   $$
   D_{IJ}^{calc} = \sqrt{(X_{J0} - X_{I0})^2 + (Y_{J0} - Y_{I0})^2}
   $$

3. **Residuo de la Observación:**
   El residuo de la observación $ v_{IJ} $ es la diferencia entre la distancia observada y la distancia calculada:

   $$
   v_{IJ} = D_{IJ}^{obs} - D_{IJ}^{calc}
   $$

4. **Vector de Residuos:**
   El vector de residuos $ \mathbf{v} $ se construye acumulando los residuos de todas las observaciones de distancia. Si tenemos $ n $ observaciones, el vector de residuos es:

   $$
   \mathbf{v} = \begin{bmatrix}
   v_{IJ_1} \\
   v_{IJ_2} \\
   \vdots \\
   v_{IJ_n}
   \end{bmatrix}
   $$

### Ejemplo de Cálculo del Residuo

Supongamos que tenemos una observación de distancia $ IJ $ con:

- Distancia observada $ D_{IJ}^{obs} $
- Coordenadas iniciales $ (X_{I0}, Y_{I0}) $ y $ (X_{J0}, Y_{J0}) $

Primero, calculamos la distancia inicial:

$$
D_{IJ}^{calc} = \sqrt{(X_{J0} - X_{I0})^2 + (Y_{J0} - Y_{I0})^2}
$$

Luego, obtenemos el residuo:

$$
v_{IJ} = D_{IJ}^{obs} - D_{IJ}^{calc}
$$

Finalmente, este residuo se coloca en la correspondiente posición en el vector $ \mathbf{v} $.

### Creación del Vector de Parámetros

En el ajuste de mínimos cuadrados, el vector de parámetros $ \boldsymbol{\beta} $ representa las coordenadas de los puntos que queremos ajustar. En el contexto de redes topográficas, estos parámetros son las coordenadas $ X $ e $ Y $ de los puntos de la red.

#### Paso a Paso para Crear el Vector de Parámetros

1. **Definir los Puntos de la Red:**
   Identificamos todos los puntos $ P_1, P_2, \ldots, P_n $ que forman parte de la red topográfica y cuyas coordenadas queremos ajustar.

2. **Vector de Parámetros Inicial:**
   El vector de parámetros inicial $ \boldsymbol{\beta}_0 $ contiene las coordenadas iniciales $ X $ e $ Y $ de todos los puntos. Se organiza de la siguiente manera:

   $$
   \boldsymbol{\beta}_0 = \begin{bmatrix}
   X_{P1,0} \\
   Y_{P1,0} \\
   X_{P2,0} \\
   Y_{P2,0} \\
   \vdots \\
   X_{Pn,0} \\
   Y_{Pn,0}
   \end{bmatrix}
   $$

   Aquí, $ X_{P1,0} $ y $ Y_{P1,0} $ son las coordenadas iniciales del punto $ P_1 $, $ X_{P2,0} $ y $ Y_{P2,0} $ son las coordenadas iniciales del punto $ P_2 $, y así sucesivamente.

3. **Actualización del Vector de Parámetros:**
   Durante el proceso iterativo de ajuste, el vector de parámetros se actualiza en cada iteración sumando las correcciones $ \Delta \boldsymbol{\beta} $ al vector de parámetros anterior:

   $$
   \boldsymbol{\beta}_{k+1} = \boldsymbol{\beta}_k + \Delta \boldsymbol{\beta}
   $$

   donde $ k $ indica la iteración actual.

### Ejemplo de Vector de Parámetros

Supongamos que estamos ajustando una red de 3 puntos. Las coordenadas iniciales de estos puntos son:

- Punto $ P_1 $: $ (X_{P1,0}, Y_{P1,0}) = (100, 200) $
- Punto $ P_2 $: $ (X_{P2,0}, Y_{P2,0}) = (300, 400) $
- Punto $ P_3 $: $ (X_{P3,0}, Y_{P3,0}) = (500, 600) $

El vector de parámetros inicial $ \boldsymbol{\beta}_0 $ sería:

$$
\boldsymbol{\beta}_0 = \begin{bmatrix}
100 \\
200 \\
300 \\
400 \\
500 \\
600
\end{bmatrix}
$$

### Creación del Vector Gradiente

El vector gradiente es esencial para entender cómo las variaciones en los parámetros afectan la función objetivo que estamos minimizando (en este caso, la suma de los cuadrados de los residuos). El vector gradiente se utiliza para construir la matriz de diseño $ \mathbf{A} $.

#### Paso a Paso para Crear el Vector Gradiente

1. **Definir la Función Objetivo:**
   La función objetivo en el ajuste de mínimos cuadrados es la suma de los cuadrados de los residuos. Para una observación de distancia $ D_{IJ} $, la función residual es:

   $$
   r_{IJ} = D_{IJ}^{obs} - D_{IJ}^{calc}
   $$

2. **Calcular las Derivadas Parciales:**
   Necesitamos las derivadas parciales de $ r_{IJ} $ respecto a las coordenadas de los puntos $ I $ y $ J $. Estas derivadas son los elementos del vector gradiente.

   $$
   \frac{\partial r_{IJ}}{\partial X_I} = \left( \frac{X_I - X_J}{D_{IJ}^{calc}} \right)
   $$
   $$
   \frac{\partial r_{IJ}}{\partial Y_I} = \left( \frac{Y_I - Y_J}{D_{IJ}^{calc}} \right)
   $$
   $$
   \frac{\partial r_{IJ}}{\partial X_J} = \left( \frac{X_J - X_I}{D_{IJ}^{calc}} \right)
   $$
   $$
   \frac{\partial r_{IJ}}{\partial Y_J} = \left( \frac{Y_J - Y_I}{D_{IJ}^{calc}} \right)
   $$

3. **Construcción del Vector Gradiente:**
   El vector gradiente para una observación $ D_{IJ} $ se construye colocando estas derivadas en un vector. Para $ n $ observaciones, tenemos un vector gradiente para cada observación, formando la matriz de diseño $ \mathbf{A} $.

### Ejemplo de Vector Gradiente

Para la observación de distancia $ D_{IJ} $ entre los puntos $ P_1 $ y $ P_2 $:

- Las coordenadas iniciales son $ (X_{P1,0}, Y_{P1,0}) $ y $ (X_{P2,0}, Y_{P2,0}) $.
- La distancia calculada es $ D_{12}^{calc} = \sqrt{(X_{P2,0} - X_{P1,0})^2 + (Y_{P2,0} - Y_{P1,0})^2} $.

El vector gradiente sería:

$$
\mathbf{g}_{12} = \begin{bmatrix}
\frac{\partial r_{12}}{\partial X_{P1}} \\
\frac{\partial r_{12}}{\partial Y_{P1}} \\
\frac{\partial r_{12}}{\partial X_{P2}} \\
\frac{\partial r_{12}}{\partial Y_{P2}}
\end{bmatrix} = \begin{bmatrix}
\left( \frac{X_{P1,0} - X_{P2,0}}{D_{12}^{calc}} \right) \\
\left( \frac{Y_{P1,0} - Y_{P2,0}}{D_{12}^{calc}} \right) \\
\left( \frac{X_{P2,0} - X_{P1,0}}{D_{12}^{calc}} \right) \\
\left( \frac{Y_{P2,0} - Y_{P1,0}}{D_{12}^{calc}} \right)
\end{bmatrix}
$$

### Conclusión

La creación del vector de parámetros y del vector gradiente es fundamental en el ajuste de mínimos cuadrados. El vector de parámetros contiene las coordenadas iniciales de los puntos, mientras que el vector gradiente proporciona las derivadas necesarias para construir la matriz de diseño. Estos elementos permiten transformar el problema de ajuste no lineal en un problema linealizado que se resuelve iterativamente para obtener las mejores estimaciones de las coordenadas de la red.

### Posiciones de las Parciales en el Vector Gradiente

Las posiciones de las derivadas parciales en el vector gradiente dependen de los puntos involucrados en cada observación. Estas posiciones deben alinearse con la estructura del vector de parámetros, donde cada par de coordenadas $ (X, Y) $ corresponde a un punto específico. Esto asegura que la matriz de diseño $ \mathbf{A} $ y el vector de parámetros $ \boldsymbol{\beta} $ se correspondan correctamente.

### Ejemplo con Tres Puntos (A, B, C)

Supongamos que tenemos tres puntos $ A $, $ B $ y $ C $ con las siguientes observaciones de distancia: $ AB $, $ AC $, $ BC $, y $ CA $.

#### Vector de Parámetros

El vector de parámetros inicial $ \boldsymbol{\beta}_0 $ para los puntos $ A $, $ B $, y $ C $ es:

$$
\boldsymbol{\beta}_0 = \begin{bmatrix}
X_{A,0} \\
Y_{A,0} \\
X_{B,0} \\
Y_{B,0} \\
X_{C,0} \\
Y_{C,0}
\end{bmatrix}
$$

#### Gradientes y Filas de la Matriz de Diseño

Para cada observación, calculamos las derivadas parciales y construimos una fila en la matriz de diseño $ \mathbf{A} $.

##### Observación AB

Para la observación $ AB $:

$$
D_{AB}^{calc} = \sqrt{(X_{B,0} - X_{A,0})^2 + (Y_{B,0} - Y_{A,0})^2}
$$

Las derivadas parciales son:

$$
\frac{\partial r_{AB}}{\partial X_A} = \left( \frac{X_{A,0} - X_{B,0}}{D_{AB}^{calc}} \right), \quad \frac{\partial r_{AB}}{\partial Y_A} = \left( \frac{Y_{A,0} - Y_{B,0}}{D_{AB}^{calc}} \right)
$$

$$
\frac{\partial r_{AB}}{\partial X_B} = \left( \frac{X_{B,0} - X_{A,0}}{D_{AB}^{calc}} \right), \quad \frac{\partial r_{AB}}{\partial Y_B} = \left( \frac{Y_{B,0} - Y_{A,0}}{D_{AB}^{calc}} \right)
$$

La fila correspondiente en la matriz $ \mathbf{A} $ para $ AB $ es:

$$
\mathbf{A}_{AB} = \begin{bmatrix}
\left( \frac{X_{A,0} - X_{B,0}}{D_{AB}^{calc}} \right), & \left( \frac{Y_{A,0} - Y_{B,0}}{D_{AB}^{calc}} \right), & \left( \frac{X_{B,0} - X_{A,0}}{D_{AB}^{calc}} \right), & \left( \frac{Y_{B,0} - Y_{A,0}}{D_{AB}^{calc}} \right), & 0, & 0
\end{bmatrix}
$$

##### Observación AC

Para la observación $ AC $:

$$
D_{AC}^{calc} = \sqrt{(X_{C,0} - X_{A,0})^2 + (Y_{C,0} - Y_{A,0})^2}
$$

Las derivadas parciales son:

$$
\frac{\partial r_{AC}}{\partial X_A} = \left( \frac{X_{A,0} - X_{C,0}}{D_{AC}^{calc}} \right), \quad \frac{\partial r_{AC}}{\partial Y_A} = \left( \frac{Y_{A,0} - Y_{C,0}}{D_{AC}^{calc}} \right)
$$

$$
\frac{\partial r_{AC}}{\partial X_C} = \left( \frac{X_{C,0} - X_{A,0}}{D_{AC}^{calc}} \right), \quad \frac{\partial r_{AC}}{\partial Y_C} = \left( \frac{Y_{C,0} - Y_{A,0}}{D_{AC}^{calc}} \right)
$$

La fila correspondiente en la matriz $ \mathbf{A} $ para $ AC $ es:

$$
\mathbf{A}_{AC} = \begin{bmatrix}
\left( \frac{X_{A,0} - X_{C,0}}{D_{AC}^{calc}} \right), & \left( \frac{Y_{A,0} - Y_{C,0}}{D_{AC}^{calc}} \right), & 0, & 0, & \left( \frac{X_{C,0} - X_{A,0}}{D_{AC}^{calc}} \right), & \left( \frac{Y_{C,0} - Y_{A,0}}{D_{AC}^{calc}} \right)
\end{bmatrix}
$$

##### Observación BC

Para la observación $ BC $:

$$
D_{BC}^{calc} = \sqrt{(X_{C,0} - X_{B,0})^2 + (Y_{C,0} - Y_{B,0})^2}
$$

Las derivadas parciales son:

$$
\frac{\partial r_{BC}}{\partial X_B} = \left( \frac{X_{B,0} - X_{C,0}}{D_{BC}^{calc}} \right), \quad \frac{\partial r_{BC}}{\partial Y_B} = \left( \frac{Y_{B,0} - Y_{C,0}}{D_{BC}^{calc}} \right)
$$

$$
\frac{\partial r_{BC}}{\partial X_C} = \left( \frac{X_{C,0} - X_{B,0}}{D_{BC}^{calc}} \right), \quad \frac{\partial r_{BC}}{\partial Y_C} = \left( \frac{Y_{C,0} - Y_{B,0}}{D_{BC}^{calc}} \right)
$$

La fila correspondiente en la matriz $ \mathbf{A} $ para $ BC $ es:

$$
\mathbf{A}_{BC} = \begin{bmatrix}
0, & 0, & \left( \frac{X_{B,0} - X_{C,0}}{D_{BC}^{calc}} \right), & \left( \frac{Y_{B,0} - Y_{C,0}}{D_{BC}^{calc}} \right), & \left( \frac{X_{C,0} - X_{B,0}}{D_{BC}^{calc}} \right), & \left( \frac{Y_{C,0} - Y_{B,0}}{D_{BC}^{calc}} \right)
\end{bmatrix}
$$

##### Observación CA

Para la observación $ CA $:

$$
D_{CA}^{calc} = \sqrt{(X_{A,0} - X_{C,0})^2 + (Y_{A,0} - Y_{C,0})^2}
$$

Las derivadas parciales son:

$$
\frac{\partial r_{CA}}{\partial X_C} = \left( \frac{X_{C,0} - X_{A,0}}{D_{CA}^{calc}} \right), \quad \frac{\partial r_{CA}}{\partial Y_C} = \left( \frac{Y_{C,0} - Y_{A,0}}{D_{CA}^{calc}} \right)
$$

$$
\frac{\partial r_{CA}}{\partial X_A} = \left( \frac{X_{A,0} - X_{C,0}}{D_{CA}^{calc}} \right), \quad \frac{\partial r_{CA}}{\partial Y_A} = \left( \frac{Y_{A,0} - Y_{C,0}}{D_{CA}^{calc}} \right)
$$

La fila correspondiente en la matriz $ \mathbf{A} $ para $ CA $ es:

$$
\mathbf{A}_{CA} = \begin{bmatrix}
\left( \frac{X_{A,0} - X_{C,0}}{D_{CA}^{calc}} \right), & \left( \frac{Y_{A,0} - Y_{C,0}}{D_{CA}^{calc}} \right), & 0, & 0, & \left( \frac{X_{C,0} - X_{A,0}}{D_{CA}^{calc}} \right), & \left( \frac{Y_{C,0} - Y_{A,0}}{D_{CA}^{calc}} \right)
\end{bmatrix}
$$

### Matriz de Diseño Completa

Combinando todas las filas de las observaciones $ AB $, $ AC $, $ BC $, y $ CA $, obten

emos la matriz de diseño $ \mathbf{A} $:

$$
\mathbf{A} = \begin{bmatrix}
\left( \frac{X_{A,0} - X_{B,0}}{D_{AB}^{calc}} \right) & \left( \frac{Y_{A,0} - Y_{B,0}}{D_{AB}^{calc}} \right) & \left( \frac{X_{B,0} - X_{A,0}}{D_{AB}^{calc}} \right) & \left( \frac{Y_{B,0} - Y_{A,0}}{D_{AB}^{calc}} \right) & 0 & 0 \\

\left( \frac{X_{A,0} - X_{C,0}}{D_{CA}^{calc}} \right) & \left( \frac{Y_{A,0} - Y_{C,0}}{D_{CA}^{calc}} \right) & 0 & 0 & \left( \frac{X_{C,0} - X_{A,0}}{D_{CA}^{calc}} \right) & \left( \frac{Y_{C,0} - Y_{A,0}}{D_{CA}^{calc}} \right)

$$

### Conclusión

La construcción del vector gradiente y la matriz de diseño $ \mathbf{A} $ requiere calcular las derivadas parciales de las funciones residuales con respecto a las coordenadas involucradas en cada observación. Estas derivadas se posicionan en la matriz de diseño de acuerdo con la estructura del vector de parámetros. Al seguir este proceso, se garantiza que el sistema de ecuaciones resultante sea coherente y se pueda resolver para obtener las correcciones de las coordenadas en el ajuste de mínimos cuadrados.

\left( \frac{X_{A,0} - X_{C,0}}{D_{CA}^{calc}} \right) & \left( \frac{Y_{A,0} - Y_{C,0}}{D_{CA}^{calc}} \right) & 0 & 0 & \left( \frac{X_{C,0} - X_{A,0}}{D_{CA}^{calc}} \right) & \left( \frac{Y_{C,0} - Y_{A,0}}{D_{CA}^{calc}} \right)
\end{bmatrix}
$$

### Ejemplo Detallado

Para la observación $ AB $:

$$
D_{AB}^{calc} = \sqrt{(X_{B,0} - X_{A,0})^2 + (Y_{B,0} - Y_{A,0})^2}
$$

Las derivadas parciales son:

$$
\frac{\partial r_{AB}}{\partial X_A} = \left( \frac{X_{A,0} - X_{B,0}}{D_{AB}^{calc}} \right), \quad \frac{\partial r_{AB}}{\partial Y_A} = \left( \frac{Y_{A,0} - Y_{B,0}}{D_{AB}^{calc}} \right)

$$
\frac{\partial r_{AB}}{\partial X_B} = \left( \frac{X_{B,0} - X_{A,0}}{D_{AB}^{calc}} \right), \quad \frac{\partial r_{AB}}{\partial Y_B} = \left( \frac{Y_{B,0} - Y_{A,0}}{D_{AB}^{calc}} \right)

\frac{\partial r_{BC}}{\partial X_C} = \left( \frac{X_{C,0} - X_{B,0}}{D_{BC}^{calc}} \right), \quad \frac{\partial r_{BC}}{\partial Y_C} = \left( \frac{Y_{C,0} - Y_{B,0}}{D_{BC}^{calc}} \right)

$$
\frac{\partial r_{CA}}{\partial X_A} = \left( \frac{X_{A,0} - X_{C,0}}{D_{CA}^{calc}} \right), \quad \frac{\partial r_{CA}}{\partial Y_A} = \left( \frac{Y_{A,0} - Y_{C,0}}{D_{CA}^{calc}} \right)

Combinando todas las filas de las observaciones $ AB $, $ AC $, $ BC $, y $ CA $, obtenemos la matriz de diseño $ \mathbf{A} $:

\mathbf{A} = \begin{bmatrix}

\end{bmatrix}
$$

### Conclusión

La matriz de diseño $ \mathbf{A} $ y el vector de residuos $ \mathbf{v} $ forman el sistema de ecuaciones linealizado que se resuelve iterativamente en el ajuste de mínimos cuadrados. Al seguir este proceso, podemos ajustar las coordenadas de la red topográfica de manera precisa, minimizando las diferencias entre las observaciones y el modelo.

### Relación del Vector Parámetro, el Vector Gradiente y el Vector de Correcciones

Para entender cómo los vectores de parámetros, gradiente y correcciones interactúan en el ajuste de mínimos cuadrados, es esencial explorar la estructura y función de cada uno dentro del contexto de la optimización.

#### Vector de Parámetros

El vector de parámetros $ \boldsymbol{\beta} $ contiene las coordenadas de los puntos que estamos ajustando. En un ajuste de redes topográficas, estos parámetros corresponden a las coordenadas $ X $ e $ Y $ de los puntos de la red.

$$
\boldsymbol{\beta} = \begin{bmatrix}
X_A \\
Y_A \\
X_B \\
Y_B \\
X_C \\
Y_C \\
\vdots
\end{bmatrix}
$$

#### Vector Gradiente

El vector gradiente es una representación de las derivadas parciales de las funciones de observación (en este caso, distancias) respecto a las coordenadas de los puntos. Este vector muestra cómo cambia la función de observación con cambios infinitesimales en los parámetros.

Para una observación de distancia $ D_{IJ} $ entre dos puntos $ I $ y $ J $, el vector gradiente $ \mathbf{g}_{IJ} $ es:

$$
\mathbf{g}_{IJ} = \begin{bmatrix}
\frac{\partial r_{IJ}}{\partial X_I} \\
\frac{\partial r_{IJ}}{\partial Y_I} \\
\frac{\partial r_{IJ}}{\partial X_J} \\
\frac{\partial r_{IJ}}{\partial Y_J} \\
\vdots
\end{bmatrix}
$$

#### Vector de Correcciones

El vector de correcciones $ \Delta \boldsymbol{\beta} $ representa los cambios que necesitamos aplicar a las coordenadas iniciales para minimizar la suma de los cuadrados de los residuos. Cada iteración del método de mínimos cuadrados ajusta las coordenadas según las correcciones calculadas.

$$
\Delta \boldsymbol{\beta} = \begin{bmatrix}
\Delta X_A \\
\Delta Y_A \\
\Delta X_B \\
\Delta Y_B \\
\Delta X_C \\
\Delta Y_C \\
\vdots
\end{bmatrix}
$$

### Sistema de Ecuaciones Linealizadas

El objetivo del ajuste de mínimos cuadrados es encontrar $ \Delta \boldsymbol{\beta} $ tal que las distancias calculadas se acerquen lo más posible a las distancias observadas. Esto se logra resolviendo el siguiente sistema de ecuaciones linealizadas:

$$
\mathbf{A} \Delta \boldsymbol{\beta} = \mathbf{v}
$$

donde:
- $ \mathbf{A} $ es la matriz de diseño que contiene las derivadas parciales organizadas de tal manera que cada fila corresponde a una observación y cada columna corresponde a un parámetro.
- $ \Delta \boldsymbol{\beta} $ es el vector de corrección de coordenadas.
- $ \mathbf{v} $ es el vector de residuos.

#### Estructura Común

La estructura de $ \boldsymbol{\beta} $, $ \mathbf{g} $ y $ \Delta \boldsymbol{\beta} $ es similar porque todos ellos representan coordenadas de los puntos de la red, ya sea en su estado inicial, en términos de gradientes o en términos de correcciones necesarias. Esta consistencia en la estructura permite que las matrices y los vectores se alineen correctamente para formar un sistema de ecuaciones que puede resolverse iterativamente.

### Ejemplo de Integración

Consideremos tres puntos $ A $, $ B $ y $ C $ con las siguientes observaciones de distancia: $ AB $, $ AC $ y $ BC $.

1. **Vector de Parámetros Inicial**:

   $$
   \boldsymbol{\beta}_0 = \begin{bmatrix}
   X_{A,0} \\
   Y_{A,0} \\
   X_{B,0} \\
   Y_{B,0} \\
   X_{C,0} \\
   Y_{C,0}
   \end{bmatrix}
   $$

2. **Vector Gradiente para Observación $ AB $**:

   $$
   \mathbf{g}_{AB} = \begin{bmatrix}
   \left( \frac{X_{A,0} - X_{B,0}}{D_{AB}^{calc}} \right) \\
   \left( \frac{Y_{A,0} - Y_{B,0}}{D_{AB}^{calc}} \right) \\
   \left( \frac{X_{B,0} - X_{A,0}}{D_{AB}^{calc}} \right) \\
   \left( \frac{Y_{B,0} - Y_{A,0}}{D_{AB}^{calc}} \right) \\
   0 \\
   0
   \end{bmatrix}
   $$

3. **Vector de Correcciones**:

   $$
   \Delta \boldsymbol{\beta} = \begin{bmatrix}
   \Delta X_A \\
   \Delta Y_A \\
   \Delta X_B \\
   \Delta Y_B \\
   \Delta X_C \\
   \Delta Y_C
   \end{bmatrix}
   $$

4. **Matriz de Diseño $ \mathbf{A} $** (combinando todas las observaciones):

   $$
   \mathbf{A} = \begin{bmatrix}
   \left( \frac{X_{A,0} - X_{B,0}}{D_{AB}^{calc}} \right) & \left( \frac{Y_{A,0} - Y_{B,0}}{D_{AB}^{calc}} \right) & \left( \frac{X_{B,0} - X_{A,0}}{D_{AB}^{calc}} \right) & \left( \frac{Y_{B,0} - Y_{A,0}}{D_{AB}^{calc}} \right) & 0 & 0 \\
   \left( \frac{X_{A,0} - X_{C,0}}{D_{AC}^{calc}} \right) & \left( \frac{Y_{A,0} - Y_{C,0}}{D_{AC}^{calc}} \right) & 0 & 0 & \left( \frac{X_{C,0} - X_{A,0}}{D_{AC}^{calc}} \right) & \left( \frac{Y_{C,0} - Y_{A,0}}{D_{AC}^{calc}} \right) \\
   0 & 0 & \left( \frac{X_{B,0} - X_{C,0}}{D_{BC}^{calc}} \right) & \left( \frac{Y_{B,0} - Y_{C,0}}{D_{BC}^{calc}} \right) & \left( \frac{X_{C,0} - X_{B,0}}{D_{BC}^{calc}} \right) & \left( \frac{Y_{C,0} - Y_{B,0}}{D_{BC}^{calc}} \right)
   \end{bmatrix}
   $$

### Conclusión

El ajuste de mínimos cuadrados se basa en la interacción entre el vector de parámetros, el vector gradiente y el vector de correcciones. Todos ellos comparten una estructura común, reflejando las coordenadas de los puntos de la red. Esta estructura coherente permite formar un sistema de ecuaciones linealizado que puede resolverse iterativamente para obtener las mejores estimaciones de las coordenadas, minimizando la suma de los cuadrados de los residuos. La correcta alineación y relación entre estos vectores y matrices es esencial para el éxito del ajuste de la red topográfica.

### Relación entre la Cantidad de Filas en la Matriz $ \mathbf{A} $, el Vector de Residuos y las Observaciones

En el ajuste de mínimos cuadrados, la relación entre la matriz de diseño $ \mathbf{A} $, el vector de residuos $ \mathbf{v} $ y las observaciones es fundamental para entender cómo se estructura y resuelve el sistema de ecuaciones.

#### Observaciones

Cada observación proporciona información sobre la relación entre diferentes puntos en la red topográfica. Las observaciones pueden ser distancias, ángulos, diferencias de elevación, etc. En este ejemplo, nos centramos en observaciones de distancias entre puntos.

Supongamos que tenemos $ m $ observaciones de distancias. Cada observación se representa como una ecuación residual que describe la diferencia entre la distancia observada y la distancia calculada a partir de las coordenadas actuales de los puntos.

#### Vector de Residuos $ \mathbf{v} $

El vector de residuos $ \mathbf{v} $ contiene los residuos de cada observación. Si tenemos $ m $ observaciones, entonces el vector de residuos tendrá $ m $ elementos:

$$
\mathbf{v} = \begin{bmatrix}
v_1 \\
v_2 \\
\vdots \\
v_m
\end{bmatrix}
$$

Cada residuo $ v_i $ es la diferencia entre la distancia observada y la distancia calculada para la $ i $-ésima observación:

$$
v_i = D_{ij}^{obs} - D_{ij}^{calc}
$$

#### Matriz de Diseño $ \mathbf{A} $

La matriz de diseño $ \mathbf{A} $ tiene una fila por cada observación, es decir, $ m $ filas. Las columnas de $ \mathbf{A} $ corresponden a las derivadas parciales de las funciones de observación respecto a los parámetros (coordenadas de los puntos). Si tenemos $ n $ puntos en la red, cada uno con dos coordenadas $ X $ e $ Y $, entonces hay $ 2n $ columnas en $ \mathbf{A} $.

Para $ m $ observaciones y $ n $ puntos:

$$
\mathbf{A} = \begin{bmatrix}
\frac{\partial r_1}{\partial X_1} & \frac{\partial r_1}{\partial Y_1} & \frac{\partial r_1}{\partial X_2} & \frac{\partial r_1}{\partial Y_2} & \cdots & \frac{\partial r_1}{\partial X_n} & \frac{\partial r_1}{\partial Y_n} \\
\frac{\partial r_2}{\partial X_1} & \frac{\partial r_2}{\partial Y_1} & \frac{\partial r_2}{\partial X_2} & \frac{\partial r_2}{\partial Y_2} & \cdots & \frac{\partial r_2}{\partial X_n} & \frac{\partial r_2}{\partial Y_n} \\
\vdots & \vdots & \vdots & \vdots & \cdots & \vdots & \vdots \\
\frac{\partial r_m}{\partial X_1} & \frac{\partial r_m}{\partial Y_1} & \frac{\partial r_m}{\partial X_2} & \frac{\partial r_m}{\partial Y_2} & \cdots & \frac{\partial r_m}{\partial X_n} & \frac{\partial r_m}{\partial Y_n}
\end{bmatrix}
$$

Cada fila de $ \mathbf{A} $ corresponde a una observación y contiene las derivadas parciales de la función residual de esa observación respecto a cada parámetro (coordenadas).

#### Sistema de Ecuaciones Linealizadas

El sistema de ecuaciones linealizadas para una iteración del ajuste de mínimos cuadrados se puede escribir como:

$$
\mathbf{A} \Delta \boldsymbol{\beta} = \mathbf{v}
$$

donde:
- $ \mathbf{A} $ es una matriz $ m \times 2n $.
- $ \Delta \boldsymbol{\beta} $ es el vector de correcciones de coordenadas de tamaño $ 2n \times 1 $.
- $ \mathbf{v} $ es el vector de residuos de tamaño $ m \times 1 $.

### Ejemplo con Tres Puntos y Cuatro Observaciones

Supongamos que tenemos tres puntos $ A $, $ B $, y $ C $ con cuatro observaciones de distancias: $ AB $, $ AC $, $ BC $, y $ CA $. Esto significa que $ m = 4 $ y $ n = 3 $.

#### Vector de Parámetros

El vector de parámetros inicial $ \boldsymbol{\beta}_0 $ es:

$$
\boldsymbol{\beta}_0 = \begin{bmatrix}
X_{A,0} \\
Y_{A,0} \\
X_{B,0} \\
Y_{B,0} \\
X_{C,0} \\
Y_{C,0}
\end{bmatrix}
$$

#### Vector de Residuos

El vector de residuos $ \mathbf{v} $ para las cuatro observaciones es:

$$
\mathbf{v} = \begin{bmatrix}
v_{AB} \\
v_{AC} \\
v_{BC} \\
v_{CA}
\end{bmatrix}
$$

#### Matriz de Diseño

La matriz de diseño $ \mathbf{A} $ tendrá $ 4 $ filas (una por cada observación) y $ 6 $ columnas (dos por cada uno de los tres puntos):

$$
\mathbf{A} = \begin{bmatrix}
\left( \frac{X_{A,0} - X_{B,0}}{D_{AB}^{calc}} \right) & \left( \frac{Y_{A,0} - Y_{B,0}}{D_{AB}^{calc}} \right) & \left( \frac{X_{B,0} - X_{A,0}}{D_{AB}^{calc}} \right) & \left( \frac{Y_{B,0} - Y_{A,0}}{D_{AB}^{calc}} \right) & 0 & 0 \\
\left( \frac{X_{A,0} - X_{C,0}}{D_{AC}^{calc}} \right) & \left( \frac{Y_{A,0} - Y_{C,0}}{D_{AC}^{calc}} \right) & 0 & 0 & \left( \frac{X_{C,0} - X_{A,0}}{D_{AC}^{calc}} \right) & \left( \frac{Y_{C,0} - Y_{A,0}}{D_{AC}^{calc}} \right) \\
0 & 0 & \left( \frac{X_{B,0} - X_{C,0}}{D_{BC}^{calc}} \right) & \left( \frac{Y_{B,0} - Y_{C,0}}{D_{BC}^{calc}} \right) & \left( \frac{X_{C,0} - X_{B,0}}{D_{BC}^{calc}} \right) & \left( \frac{Y_{C,0} - Y_{B,0}}{D_{BC}^{calc}} \right) \\
\left( \frac{X_{A,0} - X_{C,0}}{D_{CA}^{calc}} \right) & \left( \frac{Y_{A,0} - Y_{C,0}}{D_{CA}^{calc}} \right) & 0 & 0 & \left( \frac{X_{C,0} - X_{A,0}}{D_{CA}^{calc}} \right) & \left( \frac{Y_{C,0} - Y_{A,0}}{D_{CA}^{calc}} \right)
\end{bmatrix}
$$

### Conclusión

La cantidad de filas en la matriz $ \mathbf{A} $ y el vector de residuos $ \mathbf{v} $ está directamente relacionada con el número de observaciones $ m $. Cada observación proporciona una ecuación residual, lo que se traduce en una fila en $ \mathbf{A} $ y un elemento en $ \mathbf{v} $. La estructura coherente de estas matrices y vectores es crucial para formar un sistema de ecuaciones que pueda resolverse iterativamente para ajustar las coordenadas de la red topográfica.

### Matriz Diagonal de Pesos y el Vector de Parámetros

En el ajuste de mínimos cuadrados ponderados, cada observación puede tener una diferente confiabilidad o precisión. Para tomar esto en cuenta, se utiliza una matriz diagonal de pesos $ \mathbf{P} $. Esta matriz de pesos permite ponderar las observaciones de manera que aquellas con mayor precisión tengan mayor influencia en la solución del ajuste.

#### Vector de Parámetros y Observaciones

El vector de parámetros $ \boldsymbol{\beta} $ contiene las coordenadas de los puntos de la red que estamos ajustando:

$$
\boldsymbol{\beta} = \begin{bmatrix}
X_A \\
Y_A \\
X_B \\
Y_B \\
X_C \\
Y_C \\
\vdots
\end{bmatrix}
$$

La longitud del vector de parámetros está relacionada con el número de puntos y sus respectivas coordenadas en la red.

#### Relación con la Matriz de Pesos

La matriz de pesos $ \mathbf{P} $ se utiliza para ponderar el impacto de cada observación en el ajuste. Aunque la longitud del vector de parámetros no se utiliza directamente para crear la matriz de pesos, la matriz de pesos debe ser compatible en dimensiones con la matriz de diseño $ \mathbf{A} $ y el vector de residuos $ \mathbf{v} $.

#### Construcción de la Matriz de Pesos

Si tenemos $ m $ observaciones, la matriz de pesos $ \mathbf{P} $ será una matriz diagonal de tamaño $ m \times m $, donde cada elemento diagonal $ p_{ii} $ representa el peso asociado a la $ i $-ésima observación.

$$
\mathbf{P} = \begin{bmatrix}
p_{11} & 0 & 0 & \cdots & 0 \\
0 & p_{22} & 0 & \cdots & 0 \\
0 & 0 & p_{33} & \cdots & 0 \\
\vdots & \vdots & \vdots & \ddots & \vdots \\
0 & 0 & 0 & \cdots & p_{mm}
\end{bmatrix}
$$

#### Ajuste de Mínimos Cuadrados Ponderados

En el ajuste de mínimos cuadrados ponderados, el sistema de ecuaciones se modifica para incluir la matriz de pesos:

$$
\mathbf{A}^T \mathbf{P} \mathbf{A} \Delta \boldsymbol{\beta} = \mathbf{A}^T \mathbf{P} \mathbf{v}
$$

donde:
- $ \mathbf{A} $ es la matriz de diseño.
- $ \mathbf{P} $ es la matriz de pesos diagonal.
- $ \Delta \boldsymbol{\beta} $ es el vector de corrección de coordenadas.
- $ \mathbf{v} $ es el vector de residuos.

#### Ejemplo con Tres Puntos y Cuatro Observaciones

Supongamos que tenemos tres puntos $ A $, $ B $, y $ C $ con cuatro observaciones de distancias: $ AB $, $ AC $, $ BC $, y $ CA $.

1. **Vector de Parámetros Inicial**:

   $$
   \boldsymbol{\beta}_0 = \begin{bmatrix}
   X_{A,0} \\
   Y_{A,0} \\
   X_{B,0} \\
   Y_{B,0} \\
   X_{C,0} \\
   Y_{C,0}
   \end{bmatrix}
   $$

2. **Vector de Residuos**:

   $$
   \mathbf{v} = \begin{bmatrix}
   v_{AB} \\
   v_{AC} \\
   v_{BC} \\
   v_{CA}
   \end{bmatrix}
   $$

3. **Matriz de Diseño**:

   $$
   \mathbf{A} = \begin{bmatrix}
   \left( \frac{X_{A,0} - X_{B,0}}{D_{AB}^{calc}} \right) & \left( \frac{Y_{A,0} - Y_{B,0}}{D_{AB}^{calc}} \right) & \left( \frac{X_{B,0} - X_{A,0}}{D_{AB}^{calc}} \right) & \left( \frac{Y_{B,0} - Y_{A,0}}{D_{AB}^{calc}} \right) & 0 & 0 \\
   \left( \frac{X_{A,0} - X_{C,0}}{D_{AC}^{calc}} \right) & \left( \frac{Y_{A,0} - Y_{C,0}}{D_{AC}^{calc}} \right) & 0 & 0 & \left( \frac{X_{C,0} - X_{A,0}}{D_{AC}^{calc}} \right) & \left( \frac{Y_{C,0} - Y_{A,0}}{D_{AC}^{calc}} \right) \\
   0 & 0 & \left( \frac{X_{B,0} - X_{C,0}}{D_{BC}^{calc}} \right) & \left( \frac{Y_{B,0} - Y_{C,0}}{D_{BC}^{calc}} \right) & \left( \frac{X_{C,0} - X_{B,0}}{D_{BC}^{calc}} \right) & \left( \frac{Y_{C,0} - Y_{B,0}}{D_{BC}^{calc}} \right) \\
   \left( \frac{X_{A,0} - X_{C,0}}{D_{CA}^{calc}} \right) & \left( \frac{Y_{A,0} - Y_{C,0}}{D_{CA}^{calc}} \right) & 0 & 0 & \left( \frac{X_{C,0} - X_{A,0}}{D_{CA}^{calc}} \right) & \left( \frac{Y_{C,0} - Y_{A,0}}{D_{CA}^{calc}} \right)
   \end{bmatrix}
   $$

4. **Matriz de Pesos**:

   La matriz de pesos $ \mathbf{P} $ tendrá una dimensión de $ 4 \times 4 $ (ya que hay 4 observaciones):

   $$
   \mathbf{P} = \begin{bmatrix}
   p_{11} & 0 & 0 & 0 \\
   0 & p_{22} & 0 & 0 \\
   0 & 0 & p_{33} & 0 \\
   0 & 0 & 0 & p_{44}
   \end{bmatrix}
   $$

### Conclusión

La matriz diagonal de pesos $ \mathbf{P} $ no depende directamente de la longitud del vector de parámetros $ \boldsymbol{\beta} $, sino del número de observaciones $ m $. Sin embargo, la estructura del sistema de ecuaciones ajustadas, que incluye $ \mathbf{A} $, $ \mathbf{v} $ y $ \mathbf{P} $, requiere que la matriz de pesos sea compatible en dimensiones con la matriz de diseño y el vector de residuos. La longitud del vector de parámetros influye en las dimensiones de la matriz de diseño $ \mathbf{A} $, y la cantidad de observaciones determina la dimensión de $ \mathbf{P} $ y $ \mathbf{v} $. Todos estos elementos deben alinearse correctamente para resolver el sistema de ecuaciones y obtener un ajuste preciso de las coordenadas en la red topográfica.
