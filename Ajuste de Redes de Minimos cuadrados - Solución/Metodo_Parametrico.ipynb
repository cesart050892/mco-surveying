### Métodos para Mínimos Cuadrados No Lineales

Los métodos de mínimos cuadrados no lineales se utilizan para ajustar modelos no lineales a datos observados. Existen varios métodos para realizar estos ajustes, y a continuación se explicará el método paramétrico, utilizando los vectores y matrices previamente mencionados.

### Método Paramétrico

El método paramétrico de mínimos cuadrados no lineales implica iterar sobre las estimaciones iniciales de los parámetros para minimizar la suma de los cuadrados de los residuos. Este proceso utiliza la linearización del modelo no lineal mediante la serie de Taylor y la construcción de un sistema de ecuaciones linealizadas.

#### Pasos del Método Paramétrico

1. **Estimaciones Iniciales:**
   Se comienza con un conjunto de estimaciones iniciales para los parámetros $ \boldsymbol{\beta}_0 $.

2. **Cálculo de la Distancia Calculada:**
   Para cada observación de distancia, se calcula la distancia inicial $ D_{ij}^{calc} $ utilizando las coordenadas iniciales.

3. **Cálculo de Residuos Iniciales:**
   Se calcula el residuo inicial para cada observación como la diferencia entre la distancia observada y la distancia calculada:
   $$
   v_{ij} = D_{ij}^{obs} - D_{ij}^{calc}
   $$

4. **Construcción de la Matriz de Diseño:**
   Se construye la matriz de diseño $ \mathbf{A} $, que contiene las derivadas parciales de los residuos respecto a cada parámetro:
   $$
   \mathbf{A}_{ij} = \begin{bmatrix}
   \frac{\partial r_{ij}}{\partial X_i} & \frac{\partial r_{ij}}{\partial Y_i} & \frac{\partial r_{ij}}{\partial X_j} & \frac{\partial r_{ij}}{\partial Y_j} & \cdots
   \end{bmatrix}
   $$

5. **Sistema de Ecuaciones Linealizado:**
   Se forma el sistema de ecuaciones linealizadas utilizando la matriz de diseño, el vector de residuos y la matriz de pesos (si se utiliza):
   $$
   \mathbf{A}^T \mathbf{P} \mathbf{A} \Delta \boldsymbol{\beta} = \mathbf{A}^T \mathbf{P} \mathbf{v}
   $$

6. **Solución del Sistema:**
   Se resuelve el sistema de ecuaciones para encontrar el vector de correcciones $ \Delta \boldsymbol{\beta} $:
   $$
   \Delta \boldsymbol{\beta} = (\mathbf{A}^T \mathbf{P} \mathbf{A})^{-1} \mathbf{A}^T \mathbf{P} \mathbf{v}
   $$

7. **Actualización de los Parámetros:**
   Los parámetros se actualizan sumando las correcciones obtenidas:
   $$
   \boldsymbol{\beta}_{k+1} = \boldsymbol{\beta}_k + \Delta \boldsymbol{\beta}
   $$

8. **Iteración:**
   Se repiten los pasos 2 a 7 hasta que la magnitud de las correcciones $ \Delta \boldsymbol{\beta} $ sea menor que un umbral predefinido, indicando que se ha alcanzado la convergencia.

#### Ejemplo Detallado

Consideremos el mismo ejemplo de tres puntos $ A $, $ B $, y $ C $ con cuatro observaciones de distancia $ AB $, $ AC $, $ BC $, y $ CA $.

1. **Estimaciones Iniciales:**
   $$
   \boldsymbol{\beta}_0 = \begin{bmatrix}
   X_{A,0} \\
   Y_{A,0} \\
   X_{B,0} \\
   Y_{B,0} \\
   X_{C,0} \\
   Y_{C,0}
   \end{bmatrix}
   $$

2. **Cálculo de Distancias Calculadas:**
   $$
   D_{AB}^{calc} = \sqrt{(X_{B,0} - X_{A,0})^2 + (Y_{B,0} - Y_{A,0})^2}
   $$
   $$
   D_{AC}^{calc} = \sqrt{(X_{C,0} - X_{A,0})^2 + (Y_{C,0} - Y_{A,0})^2}
   $$
   $$
   D_{BC}^{calc} = \sqrt{(X_{C,0} - X_{B,0})^2 + (Y_{C,0} - Y_{B,0})^2}
   $$
   $$
   D_{CA}^{calc} = \sqrt{(X_{A,0} - X_{C,0})^2 + (Y_{A,0} - Y_{C,0})^2}
   $$

3. **Cálculo de Residuos Iniciales:**
   $$
   v_{AB} = D_{AB}^{obs} - D_{AB}^{calc}
   $$
   $$
   v_{AC} = D_{AC}^{obs} - D_{AC}^{calc}
   $$
   $$
   v_{BC} = D_{BC}^{obs} - D_{BC}^{calc}
   $$
   $$
   v_{CA} = D_{CA}^{obs} - D_{CA}^{calc}
   $$

4. **Construcción de la Matriz de Diseño:**
   $$
   \mathbf{A} = \begin{bmatrix}
   \left( \frac{X_{A,0} - X_{B,0}}{D_{AB}^{calc}} \right) & \left( \frac{Y_{A,0} - Y_{B,0}}{D_{AB}^{calc}} \right) & \left( \frac{X_{B,0} - X_{A,0}}{D_{AB}^{calc}} \right) & \left( \frac{Y_{B,0} - Y_{A,0}}{D_{AB}^{calc}} \right) & 0 & 0 \\
   \left( \frac{X_{A,0} - X_{C,0}}{D_{AC}^{calc}} \right) & \left( \frac{Y_{A,0} - Y_{C,0}}{D_{AC}^{calc}} \right) & 0 & 0 & \left( \frac{X_{C,0} - X_{A,0}}{D_{AC}^{calc}} \right) & \left( \frac{Y_{C,0} - Y_{A,0}}{D_{AC}^{calc}} \right) \\
   0 & 0 & \left( \frac{X_{B,0} - X_{C,0}}{D_{BC}^{calc}} \right) & \left( \frac{Y_{B,0} - Y_{C,0}}{D_{BC}^{calc}} \right) & \left( \frac{X_{C,0} - X_{B,0}}{D_{BC}^{calc}} \right) & \left( \frac{Y_{C,0} - Y_{B,0}}{D_{BC}^{calc}} \right) \\
   \left( \frac{X_{A,0} - X_{C,0}}{D_{CA}^{calc}} \right) & \left( \frac{Y_{A,0} - Y_{C,0}}{D_{CA}^{calc}} \right) & 0 & 0 & \left( \frac{X_{C,0} - X_{A,0}}{D_{CA}^{calc}} \right) & \left( \frac{Y_{C,0} - Y_{A,0}}{D_{CA}^{calc}} \right)
   \end{bmatrix}
   $$

5. **Construcción de la Matriz de Pesos (opcional):**
   Si se utilizan pesos, se construye la matriz diagonal de pesos $ \mathbf{P} $:
   $$
   \mathbf{P} = \begin{bmatrix}
   p_{AB} & 0 & 0 & 0 \\
   0 & p_{AC} & 0 & 0 \\
   0 & 0 & p_{BC} & 0 \\
   0 & 0 & 0 & p_{CA}
   \end{bmatrix}
   $$

6. **Sistema de Ecuaciones Linealizado:**
   Se forma el sistema de ecuaciones:
   $$
   \mathbf{A}^T \mathbf{P} \mathbf{A} \Delta \boldsymbol{\beta} = \mathbf{A}^T \mathbf{P} \mathbf{v}
   $$

7. **Solución del Sistema:**
   Se resuelve para $ \Delta \boldsymbol{\beta} $:
   $$
   \Delta \boldsymbol{\beta} = (\mathbf{A}^T \mathbf{P} \mathbf{A})^{-1} \mathbf{A}^T \mathbf{P} \mathbf{v}
   $$

8. **Actualización de los Parámetros:**
   $$
   \boldsymbol{\beta}_{k+1} = \boldsymbol{\beta}_k + \Delta \boldsymbol{\beta}
   $$

9. **Iteración:**
   Se repiten los pasos hasta alcanzar la convergencia.

### Conclusión

El método paramétrico de mínimos cuadrados no lineales utiliza los vectores y matrices previamente mencionados para iterativamente ajustar los parámetros del modelo, minimizando la suma de los cuadrados de los residuos. La matriz de diseño $ \mathbf{A} $, el vector de residuos $ \mathbf{v} $, la matriz de pesos $ \mathbf{P} $ y el vector de correcciones $ \Delta \boldsymbol{\beta} $ son elementos fundamentales en este proceso. El ajuste se realiza mediante iteraciones sucesivas hasta que los cambios en los parámetros sean insignificantes, indicando que se ha alcanzado una solución óptima.
