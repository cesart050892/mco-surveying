# Módulo 1: Introducción al Ajuste de Redes con Mínimos Cuadrados

## Objetivos del Módulo
- Comprender qué es el ajuste de redes utilizando mínimos cuadrados.
- Conocer la importancia del ajuste de redes en la topografía y GNSS.

## Conceptos Básicos
### ¿Qué es el ajuste de redes con mínimos cuadrados?
El ajuste de redes utilizando mínimos cuadrados es un método matemático y estadístico que se utiliza para estimar las coordenadas más probables de una serie de puntos \
interconectados en una red geodésica o topográfica. Este método minimiza la suma de los cuadrados de las diferencias (residuos) entre las observaciones medidas y las \
estimaciones calculadas, proporcionando así una solución óptima que es la mejor aproximación posible a la realidad basada en los datos disponibles.

Matemáticamente, se representa la relación entre las observaciones $ \mathbf{l} $ y las incógnitas $ \mathbf{x} $ mediante un modelo linealizado:
$$ \mathbf{l} = \mathbf{A}\mathbf{x} + \mathbf{v} $$
donde $ \mathbf{A} $ es la matriz de diseño, $ \mathbf{v} $ es el vector de residuos, y $ \mathbf{x} $ es el vector de parámetros desconocidos que se desea estimar.

El objetivo es minimizar la función objetivo:
$$ Q(\mathbf{x}) = \mathbf{v}^T\mathbf{P}\mathbf{v} $$
donde $ \mathbf{P} $ es la matriz de pesos.

### Importancia en Topografía y GNSS
El ajuste de redes es crucial porque permite obtener soluciones precisas y consistentes para las coordenadas de los puntos medidos, lo que es esencial para garantizar la \
exactitud y la fiabilidad de los datos topográficos y GNSS. En la topografía, esto se traduce en la capacidad de realizar levantamientos precisos y detallados, y en el \
ámbito GNSS, asegura que las posiciones obtenidas sean lo más exactas posible, lo cual es fundamental para aplicaciones como la navegación, la cartografía y los estudios \
geodésicos.

## Resumen
- El ajuste de redes es un método matemático que minimiza la suma de los cuadrados de las diferencias entre las observaciones medidas y las estimaciones calculadas.
- En topografía es crucial porque permite obtener soluciones precisas y consistentes, asegurando la exactitud y la fiabilidad de los datos levantados.
- En GNSS, asegura que las posiciones obtenidas sean exactas, fundamentales para aplicaciones en navegación, cartografía y estudios geodésicos.

# Módulo 2: Ajuste de Redes No Lineales

## Objetivos del Módulo
- Entender la diferencia entre ajustes lineales y no lineales.
- Aprender cómo aplicar métodos de mínimos cuadrados a problemas no lineales.
- Conocer los métodos no lineales paramétrico, condicional y combinado.

## Conceptos Básicos

### ¿Qué son los mínimos cuadrados no lineales?
Los mínimos cuadrados no lineales se aplican cuando el modelo matemático que relaciona las observaciones con las incógnitas no es lineal. Este modelo es común en problemas geodésicos y topográficos, donde las relaciones geométricas entre los puntos no pueden ser representadas adecuadamente mediante ecuaciones lineales.

En este caso, el modelo general es de la forma:
$$ \mathbf{l} = f(\mathbf{x}) + \mathbf{v} $$
donde $ f(\mathbf{x}) $ es una función no lineal de los parámetros $ \mathbf{x} $.

Para resolver este problema, se utiliza un proceso iterativo, linealizando la función $ f(\mathbf{x}) $ en torno a una estimación inicial y aplicando el método de mínimos cuadrados a la forma linealizada:
$$ \mathbf{l} = \mathbf{A}\mathbf{x} + \mathbf{v} $$
El proceso se repite hasta que las correcciones a los parámetros se vuelvan suficientemente pequeñas.

### Métodos de Resolución
Los métodos más comunes para resolver problemas de mínimos cuadrados no lineales incluyen:
- **Método de Gauss-Newton**: Utiliza la aproximación lineal de la función en cada iteración para minimizar los residuos.
- **Método de Levenberg-Marquardt**: Combina el método de Gauss-Newton con un término de regularización para mejorar la convergencia.

## Métodos No Lineales Paramétrico, Condicional y Combinado

### Método Paramétrico
El método paramétrico se utiliza cuando las observaciones pueden ser expresadas como funciones no lineales de los parámetros a estimar. En este método, todas las variables son tratadas como parámetros desconocidos que deben ser estimados.

#### Ecuaciones
$$ \mathbf{l} = f(\mathbf{x}) + \mathbf{v} $$
donde $ f(\mathbf{x}) $ es una función no lineal que describe la relación entre las observaciones $ \mathbf{l} $ y los parámetros $ \mathbf{x} $.

#### Uso Práctico
El método paramétrico es útil en situaciones donde todas las observaciones dependen directamente de los parámetros a estimar, como en la calibración de instrumentos y en ciertos tipos de levantamientos topográficos donde todas las medidas están relacionadas con coordenadas de puntos.

### Método Condicional
El método condicional se aplica cuando algunas de las observaciones están relacionadas con parámetros conocidos y otras con parámetros desconocidos. Este método introduce condiciones adicionales que deben ser satisfechas por las soluciones.

#### Ecuaciones
$$ \mathbf{l} = f(\mathbf{x}, \mathbf{y}) + \mathbf{v} $$
donde $ \mathbf{x} $ son los parámetros desconocidos y $ \mathbf{y} $ son parámetros conocidos o valores condicionados.

#### Uso Práctico
El método condicional es útil en aplicaciones donde algunas medidas son fijas o conocidas y otras deben ser determinadas. Un ejemplo práctico es la nivelación de una red donde algunos puntos tienen elevaciones conocidas y otros necesitan ser ajustados.

### Método Combinado
El método combinado integra tanto el enfoque paramétrico como el condicional. Se utiliza cuando se tienen observaciones que dependen de parámetros conocidos y desconocidos, y se desean ajustar todos simultáneamente.

#### Ecuaciones
$$ \mathbf{l} = f(\mathbf{x}, \mathbf{y}) + \mathbf{v} $$
donde $ \mathbf{x} $ son parámetros desconocidos que se desean estimar y $ \mathbf{y} $ son parámetros conocidos que se utilizan para condicionar las estimaciones.

#### Uso Práctico
El método combinado se aplica en situaciones complejas donde se tienen múltiples fuentes de datos y relaciones entre observaciones y parámetros. Un uso práctico es en el ajuste de redes GNSS, donde se pueden tener coordenadas conocidas de estaciones de referencia y se necesita ajustar la red completa incluyendo estaciones móviles.

## Resumen
- Los métodos de mínimos cuadrados no lineales son esenciales para resolver problemas donde las relaciones entre observaciones y parámetros no son lineales.
- **Método Paramétrico**: Todas las variables son tratadas como parámetros desconocidos a estimar.
- **Método Condicional**: Introduce condiciones adicionales relacionadas con parámetros conocidos y desconocidos.
- **Método Combinado**: Integra enfoques paramétricos y condicionales para ajustar todos los parámetros simultáneamente.
- Cada método tiene aplicaciones prácticas en la topografía y el análisis de datos GNSS, dependiendo de la naturaleza de las observaciones y los parámetros involucrados.

Estos métodos proporcionan herramientas poderosas para resolver problemas complejos en la topografía y el GNSS, mejorando la precisión y la fiabilidad de las soluciones obtenidas.

# Módulo 3: Resección y Ajuste de Redes Topográficas

## Objetivos del Módulo
- Entender el concepto de resección en topografía.
- Aplicar técnicas de ajuste de redes a problemas de resección.

## Conceptos Básicos
### ¿Qué es la resección?
La resección es un método de posicionamiento en topografía que consiste en determinar las coordenadas de un punto desconocido a partir de la observación \
de varios puntos conocidos. Este método es ampliamente utilizado en levantamientos topográficos y en el posicionamiento de estaciones GNSS.

### Métodos de Resección
Los métodos más comunes de resección incluyen:
- **Resección por ángulos**: Se mide el ángulo entre líneas visuales a varios puntos conocidos desde el punto desconocido.
- **Resección por distancias**: Se miden las distancias entre el punto desconocido y varios puntos conocidos.

### Aplicación de Mínimos Cuadrados a la Resección
El problema de resección puede ser formulado como un problema de mínimos cuadrados, donde se minimizan las diferencias entre las medidas observadas y las calculadas. \
Para una resección por ángulos, el modelo sería:
$$ \theta_i = f(\mathbf{x}) + \mathbf{v}_i $$
donde $ \theta_i $ es el ángulo medido y $ f(\mathbf{x}) $ es la función que relaciona las coordenadas del punto desconocido $ \mathbf{x} $ con los ángulos observados.

Para una resección por distancias, el modelo sería:
$$ d_i = g(\mathbf{x}) + \mathbf{v}_i $$
donde $ d_i $ es la distancia medida y $ g(\mathbf{x}) $ es la función que relaciona las coordenadas del punto desconocido $ \mathbf{x} $ con las distancias observadas.

## Importancia en Topografía
La resección es una técnica fundamental en topografía, ya que permite la determinación precisa de puntos desconocidos utilizando puntos de referencia conocidos. \
La aplicación de métodos de mínimos cuadrados en la resección mejora la precisión y la confiabilidad de los resultados, lo que es crucial para \
levantamientos detallados y precisos.

## Resumen
- La resección es un método de posicionamiento utilizado para determinar las coordenadas de un punto desconocido a partir de puntos conocidos.
- Los métodos comunes incluyen resección por ángulos y resección por distancias.
- Aplicando métodos de mínimos cuadrados, se pueden obtener soluciones precisas y confiables en problemas de resección en topografía.

Estos módulos proporcionan una base sólida para comprender y aplicar métodos de mínimos cuadrados en el ajuste de redes, tanto lineales como no lineales, \
y en problemas específicos como la resección, esenciales para la práctica topográfica y el análisis de datos GNSS.

### Ecuaciones de Distancia, Azimut y Ángulos

#### Distancia
$$ L_{IJ} + v_{IJ} = \sqrt{(X_J - X_I)^2 + (Y_J - Y_I)^2} $$

Expansión linealizada:
$$ \left( \frac{X_I - X_J}{IJ} \right)_0 dX_I + \left( \frac{Y_I - Y_J}{IJ} \right)_0 dY_I + \left( \frac{X_J - X_I}{IJ} \right)_0 dX_J + \left( \frac{Y_J - Y_I}{IJ} \right)_0 dY_J = k_{IJ} + v_{IJ} $$

Donde:
$$ (IJ)_0 = \sqrt{(X_{J0} - X_{I0})^2 + (Y_{J0} - Y_{I0})^2} $$

Para otro par de puntos:
$$ \left( \frac{X_A - X_B}{AB} \right)_0 dX_A + \left( \frac{Y_A - Y_B}{AB} \right)_0 dY_A + \left( \frac{X_B - X_A}{AB} \right)_0 dX_B + \left( \frac{Y_B - Y_A}{AB} \right)_0 dY_B = k_{AB} + v_{AB} $$

#### Azimut
$$ Az_{IJ} + v_{IJ} = \tan^{-1} \left( \frac{X_J - X_I}{Y_J - Y_I} \right) + C $$

Expansión linealizada:
$$ \rho \left( \frac{Y_I - Y_J}{IJ^2} \right)_0 dX_I + \rho \left( \frac{X_I - X_J}{IJ^2} \right)_0 dY_I + \rho \left( \frac{Y_J - Y_I}{IJ^2} \right)_0 dX_J + \rho \left( \frac{X_J - X_I}{IJ^2} \right)_0 dY_J = k_{IJ} + v_{IJ} $$

Donde:
$$ k_{IJ} = Az_{IJ} - \tan^{-1} \left( \frac{X_{J0} - X_{I0}}{Y_{J0} - Y_{I0}} \right) + C $$

#### Ángulos
$$ \theta_{BIF} + v_{BIF} = \tan^{-1} \left( \frac{X_F - X_I}{Y_F - Y_I} \right) - \tan^{-1} \left( \frac{X_B - X_I}{Y_B - Y_I} \right) + D $$

Expansión linealizada:
$$ \rho \left( \frac{Y_I - Y_B}{IB^2} \right)_0 dX_B + \rho \left( \frac{X_B - X_I}{IB^2} \right)_0 dY_B + \rho \left( \frac{Y_B - Y_I - Y_F + Y_I}{IB^2 - IF^2} \right)_0 dX_I + \rho \left( \frac{X_I - X_B - X_I + X_F}{IB^2 - IF^2} \right)_0 dY_I + \rho \left( \frac{X_I - X_F}{IF^2} \right)_0 dY_F = k_{BIF} + v_{BIF} $$

Donde:
$$ k_{BIF} = \theta_{BIF} - \left[ \tan^{-1} \left( \frac{Y_F - Y_I}{X_F - X_I} \right) - \tan^{-1} \left( \frac{X_B - X_I}{Y_B - Y_I} \right) \right] + D $$

Para otro par de puntos:
$$ \rho \left( \frac{Y_A - Y_G}{AG^2} \right)_0 dX_G + \rho \left( \frac{X_G - X_A}{AG^2} \right)_0 dY_G + \rho \left( \frac{Y_G - Y_A - Y_B + Y_A}{AG^2 - AB^2} \right)_0 dX_A + \rho \left( \frac{X_A - X_G - X_A + X_B}{AG^2 - AB^2} \right)_0 dY_A + \rho \left( \frac{X_A - X_B}{AB^2} \right)_0 dY_B = k_{GAB} + v_{GAB} $$

### Explicación del Desarrollo de las Ecuaciones

#### Distancia
1. **Ecuación Inicial**: La ecuación de distancia $ L_{IJ} + v_{IJ} = \sqrt{(X_J - X_I)^2 + (Y_J - Y_I)^2} $ es una expresión básica que describe la distancia entre dos puntos en un plano cartesiano.
2. **Linealización**: Para facilitar el cálculo en ajustes de mínimos cuadrados, la ecuación no lineal se expande linealmente alrededor de valores iniciales $ (X_{I0}, Y_{I0}) $ y $ (X_{J0}, Y_{J0}) $.
3. **Descomposición en Diferenciales**: Los términos diferenciales representan cambios en las coordenadas que se deben minimizar para obtener la mejor aproximación lineal a la distancia medida.

#### Azimut
1. **Ecuación Inicial**: El azimut es el ángulo entre la línea que conecta dos puntos y una referencia (normalmente el norte), dado por $ Az_{IJ} + v_{IJ} = \tan^{-1} \left( \frac{X_J - X_I}{Y_J - Y_I} \right) + C $.
2. **Linealización**: Se realiza una expansión de Taylor de primer orden para linealizar la ecuación alrededor de las coordenadas iniciales.
3. **Constantes de Linealización**: Las constantes $ k_{IJ} $ y $ C $ se utilizan para ajustar la linealización, asegurando que la diferencia entre la medición y el modelo sea mínima.

#### Ángulos
1. **Ecuación Inicial**: La ecuación para el ángulo entre tres puntos $ \theta_{BIF} + v_{BIF} = \tan^{-1} \left( \frac{X_F - X_I}{Y_F - Y_I} \right) - \tan^{-1} \left( \frac{X_B - X_I}{Y_B - Y_I} \right) + D $ representa la diferencia de dos azimuts.
2. **Linealización**: Similar a los casos anteriores, la ecuación se expande linealmente para facilitar su uso en cálculos de ajuste.
3. **Constantes de Ajuste**: Las constantes $ k_{BIF} $ y $ D $ ajustan la expansión lineal para mejorar la aproximación.

Estas ecuaciones y sus expansiones linealizadas son esenciales en la topografía y el ajuste de redes, ya que permiten transformar problemas no lineales en formas más manejables para optimizar las soluciones usando métodos de mínimos cuadrados.

### Explicación de $ C $ y $ \rho $ en las Ecuaciones de Observación de Ángulos y Azimut

#### Constante $ C $

En la ecuación de azimut:

$$ Az_{IJ} + v_{IJ} = \tan^{-1} \left( \frac{X_J - X_I}{Y_J - Y_I} \right) + C $$

$ C $ es una constante que depende de la dirección de la línea entre los puntos $ I $ y $ J $. Su valor varía según el rango de ángulos que define la dirección de la línea, específicamente:

- Si el acimut de la línea está entre $ 0^\circ $ y $ 90^\circ $, el valor de $ C $ es $ 0^\circ $.
- Si el acimut de la línea está entre $ 90^\circ $ y $ 270^\circ $, el valor de $ C $ es $ 180^\circ $.
- Si el acimut de la línea está entre $ 270^\circ $ y $ 360^\circ $, el valor de $ C $ es $ 360^\circ $.

Esta constante asegura que el ángulo calculado se ajuste correctamente dentro del rango esperado de 0° a 360°, dependiendo de la dirección en la que se mide el acimut.

#### Coeficiente $ \rho $

El coeficiente $ \rho $ aparece en la expansión linealizada de las ecuaciones de ángulos y azimut para convertir los términos diferenciales en segundos de arco (arcseconds). Dado que las unidades angulares a menudo se expresan en segundos de arco para una mayor precisión, $ \rho $ se utiliza para esta conversión. El valor de $ \rho $ es:

$$ \rho = \frac{206,265}{\text{rad}} $$

Este factor convierte radianes a segundos de arco, lo cual es importante para mantener la coherencia de las unidades en las ecuaciones linealizadas. La inclusión de $ \rho $ asegura que los coeficientes de las correcciones $ dX $ y $ dY $ se midan en segundos de arco cuando se aplican a las correcciones de las coordenadas en las ecuaciones de observación.
