## 12 Ajuste por Mínimos Cuadrados de Modelos Condicionales

### 12.1 Introducción

Como se discutió en la Sección 4.3, las ecuaciones matemáticas involucradas en el campo de la geomática no siempre se formulan solo como ecuaciones de modelos paramétricos ($\hat{\ell} = f(\hat{x})$, donde $\hat{\ell}$ es un vector de observaciones y $\hat{x}$ es un vector de parámetros ajustados). A veces, uno puede no estar inmediatamente interesado en usar las mediciones de campo para resolver algún parámetro (como las coordenadas de los puntos de la red); pero uno puede estar interesado solo en evaluar las mediciones para su consistencia entre sí sin tener que resolver ningún parámetro. Por ejemplo, considere un caso en el que los tres lados de un triángulo y sus tres ángulos interiores se miden independientemente; se espera que los ángulos interiores medidos sumen 180° con los ángulos interiores y sus correspondientes lados medidos del triángulo satisfaciendo algunas reglas de seno y coseno. Es muy probable que los ángulos interiores medidos no sumen 180° y que los ángulos y los lados correspondientes del triángulo no satisfagan las reglas de seno y coseno perfectamente. Este es un problema típico de ajuste por mínimos cuadrados que no involucra la resolución de ningún parámetro (como las coordenadas de los puntos), pero que implica la formulación de ecuaciones que conectan matemáticamente las observaciones. Las ecuaciones formuladas de esta forma se refieren como ecuaciones de modelos condicionales. Por lo tanto, un modelo condicional es un modelo que contiene solo observaciones ($\hat{\ell}$) y no parámetros, y puede expresarse matemáticamente como:

$f(\hat{\ell}) = 0$ (12.1)

En otras palabras, las observaciones son las mismas que los parámetros en los modelos condicionales. Ejemplos de ecuaciones de modelos condicionales se dan en la Sección 12.2.

### 12.2 Ecuaciones de Modelos Condicionales

Las ecuaciones de modelos condicionales se formulan de tal manera que todas las mediciones y las condiciones geométricas impuestas se representan y acomodan de manera única. Una vez que se han satisfecho estas, no se deben agregar ecuaciones adicionales o innecesarias. Una ecuación se considerará innecesaria si cualquier combinación de ecuaciones ya formadas puede reproducir esa ecuación. Formar más ecuaciones de las necesarias puede resultar en ecuaciones que se vuelvan imposibles de resolver. Dado que no hay parámetros involucrados en los modelos condicionales, el número de ecuaciones condicionales es el mismo que el número de redundancias del problema. Esto significa que la redundancia o el número de grados de libertad de las ecuaciones de modelos condicionales se determina como el número de ecuaciones de modelos independientes formadas. Formar ecuaciones únicas es usualmente el aspecto más difícil del ajuste por mínimos cuadrados de modelos condicionales. Los pasos involucrados en la formulación de ecuaciones de modelos condicionales son:

1) Identificar qué constituye las observaciones a partir de la pregunta. Note que los datos dados se identifican como observaciones cuando se asocian desviaciones estándar con ellos o cuando los datos se refieren como medidos u observados.
   
2) A partir de la comprensión del tipo de modelo ($f(\hat{\ell}) = 0$) a formular, las ecuaciones deben incluir solo las observaciones y algunos valores constantes (si es necesario); no deben estar involucrados parámetros desconocidos. Cada ecuación también debe ser dispuesta de manera que se iguale a cero; y debe asegurarse que cada ecuación sea única (basada en una relación matemática sensible que ocurre entre las mediciones) y no puede ser formulada combinando (sumando o restando o haciendo ambos) cualquiera de las ecuaciones ya formadas.

Un sistema lineal de ecuaciones se llama sistema linealmente dependiente de ecuaciones si una o más de las ecuaciones es redundante, es decir, si al menos una de las ecuaciones es una combinación lineal de las otras. Un sistema es linealmente dependiente si una de las ecuaciones se puede eliminar sin afectar las soluciones generales. Si ninguna de las ecuaciones se puede obtener como una combinación lineal de las otras, entonces el sistema se llama sistema linealmente independiente de ecuaciones. Los siguientes ejemplos ilustran cómo formar ecuaciones condicionales únicas e independientes, que excluyen ecuaciones innecesarias que harían que el problema sea irresoluble.

### 12.2.1 Ejemplos de Ecuaciones de Modelos

Los siguientes ejemplos se dan para ilustrar cómo se formulan las ecuaciones de modelos condicionales.

**Ejemplo 12.1** Se realizaron mediciones de distancia entre los puntos A, B y C como se muestra en la Figura 12.1, donde $\hat{\ell}_1$, $\hat{\ell}_2$ y $\hat{\ell}_3$ son las distancias ajustadas a determinarse, y las distancias medidas son 190.40, 203.16, y 393.65 m, respectivamente. Forme el modelo condicional $f(\hat{\ell}) = 0$ para este problema.

**Solución:** 

Las ecuaciones del modelo condicional deberían verse como $f(\hat{\ell}) = 0$. Siguiendo los pasos dados en la Sección 12.2, se forma la siguiente ecuación:

$$
\hat{\ell}_3 - (\hat{\ell}_1 + \hat{\ell}_2) = 0 \quad (12.2)
$$

donde los sombreros en las mediciones indican que las mediciones deben ajustarse; estos son necesarios para la consistencia de la ecuación, es decir, para que la ecuación no tenga ningún cierre.

**Ejemplo 12.2** En la Figura 12.2, los tres ángulos horizontales medidos alrededor del horizonte son $\alpha = 22° 12' 13''$, $\beta = 59° 56' 15''$, $\gamma = 277° 51' 35''$. Forme una ecuación apropiada $f(\hat{\ell}) = 0$ para ajustar las observaciones.

**Solución:**

Las ecuaciones del modelo condicional deberían verse como $f(\hat{\ell}) = 0$. Siguiendo los pasos dados en el Ejemplo 12.1, la relación matemática sensible entre las observaciones es que todos los ángulos en un círculo deben sumar 360. Esto se representa matemáticamente como:

$$
\hat{\alpha} + \hat{\beta} + \hat{\gamma} - 360° = 0 \quad (12.3)
$$

**Ejemplo 12.3** Dado un circuito de nivelación cerrado en la Figura 12.3 con diferencias de elevación medidas para cuatro secciones como $\Delta h_1$, $\Delta h_2$, $\Delta h_3$, $\Delta h_4$, dé el modelo condicional para este problema.

**Solución:**

En un circuito de nivelación cerrado, la suma de las diferencias de elevación (siguiendo la dirección de nivelación indicada por las flechas) debe ser igual a cero. La dirección de la flecha para la diferencia de elevación $\Delta h_1$ se usará como dirección de referencia (cualquier dirección se puede usar como referencia mientras se mantenga la consistencia), cualquier diferencia de elevación en la misma dirección se representará como positiva, y cualquier dirección opuesta se representará como negativa. El modelo condicional $f(\hat{\ell}) = 0$ se da como:

$$
\Delta h_1 - \Delta h_2 + \Delta h_3 + \Delta h_4 = 0 \quad (12.4)
$$

Ángulos: $\ell_1, \ell_2, \ell_3$ Distancias: $\ell_4, \ell_5$

Formule el modelo condicional para la red, asumiendo que la distancia $b$ es conocida y fija.

**Solución (considerando la condición de redundancia en el paso 2 en la Sección 12.2):**

Usando las leyes del seno en las mediciones del triángulo y reorganizando para que parezca $f(\hat{\ell}) = 0$ (notando que cualquier otra ecuación será redundante),

$$
\hat{\ell}_1 + \hat{\ell}_2 + \hat{\ell}_3 - 180 = 0 \quad (12.5)
$$

$$
b \sin \hat{\ell}_2 - \hat{\ell}_4 \sin \hat{\ell}_3 = 0 \quad (12.6)
$$

$$
b \sin \hat{\ell}_1 - \hat{\ell}_5 \sin \hat{\ell}_3 = 0 \quad (12.7)
$$

**Ejemplo 12.5** En la red de nivelación mostrada en la Figura 12.5, las elevaciones $h_A$ y $h_B$ de los puntos A y B, respectivamente, deben determinarse. Las elevaciones de referencia (en m) se muestran entre paréntesis, y las mediciones de campo se muestran en la figura como $\text{dh}_1$, $\text{dh}_2$, $\text{dh}_3$, $\text{dh}_4$ y $\text{dh}_5$ con las flechas que indican las direcciones de nivelación.

Dé el modelo matemático $f(\hat{\ell}) = 0$ para este problema, donde $\hat{\ell}$ es un vector de observaciones ajustadas.

**Solución:**

Tome nota de lo siguiente al formular las ecuaciones:

- Se puede ver en la red de nivelación en la Figura 12.5 que solo hay dos incógnitas (elevaciones de los puntos A y B), necesitando solo dos mediciones (diferencia de elevación hacia A y hacia B) de las cinco mediciones para determinar de manera única las dos incógnitas. Las tres mediciones restantes son redundantes, lo que significa que se deben formular tres ecuaciones condicionales.

- La suma de las diferencias de elevación entre dos referencias conocidas debe resultar en la diferencia entre las elevaciones de las dos referencias (siguiendo una dirección como valor positivo y la dirección opuesta como valor negativo).

- Siguiendo la ruta de nivelación BM1-A-B-BM3 ($\text{dh}_3$ tiene una flecha opuesta, por lo que se le asigna un valor negativo), se obtiene lo siguiente:

$$
\hat{\text{dh}}_1 - \hat{\text{dh}}_3 + \hat{\text{dh}}_4 - (794.881 - 785.232) = 0 \quad (12.8)
$$

- Siguiendo la ruta de nivelación BM4-B-A-BM2 (todas las flechas están en la misma dirección), se obtiene lo siguiente:

$$
\hat{\text{dh}}_2 + \hat{\text{dh}}_3 + \hat{\text{dh}}_5 - (805.410 - 801.930) = 0 \quad (12.9)
$$

- Aunque todas las mediciones se han tomado en cuenta en las Ecuaciones (12.8) y (12.9), aún es necesario asegurar que los circuitos BM2-A-BM1 y BM4-B-BM3 se cierren después del ajuste; si uno de estos circuitos se satisface junto con las condiciones establecidas en las Ecuaciones (12.8) y (12.9), el otro circuito se satisfará automáticamente. La condición para fijar el circuito BM2-A-BM1 se puede dar de la siguiente manera:

$$
\hat{\text{dh}}_1 + \hat{\text{dh}}_2 - (805.410 - 785.232) = 0 \quad (12.10)
$$

o la condición para fijar el circuito BM4-B-BM3 como

$$
\hat{\text{dh}}_5 + \hat{\text{dh}}_4 - (794.881 - 801.930) = 0 \quad (12.11)
$$

Solo hay tres ecuaciones únicas que acomodan todas las mediciones. El número de ecuaciones condicionales únicas se puede determinar aproximadamente como el número de mediciones menos el número de cantidades no medidas pero cuyos valores se están determinando implícitamente. En este problema, hay cinco mediciones ($\text{dh}_1$, $\text{dh}_2$, $\text{dh}_3$, $\text{dh}_4$, $\text{dh}_5$) y dos cantidades desconocidas ($h_A$ y $h_B$) no medidas y cuyos valores se están determinando implícitamente, dando cinco menos dos ecuaciones condicionales (o tres ecuaciones condicionales). Las tres ecuaciones de condición simplificadas de las Ecuaciones (12.8) a (12.10) se pueden dar como sigue:

$$
\hat{\text{dh}}_1 - \hat{\text{dh}}_3 + \hat{\text{dh}}_4 - 9.649 = 0 \quad (12.12)
$$
$$
\hat{\text{dh}}_2 + \hat{\text{dh}}_3 - 3.480 = 0 \quad (12.13)
$$
$$
\hat{\text{dh}}_1 + \hat{\text{dh}}_2 - 20.178 = 0 \quad (12.14)
$$

o de las Ecuaciones (12.8), (12.9) y (12.11):

$$
\hat{\text{dh}}_1 - \hat{\text{dh}}_3 + \hat{\text{dh}}_4 - 9.649 = 0 \quad (12.15)
$$
$$
\hat{\text{dh}}_2 + \hat{\text{dh}}_3 - 3.480 = 0 \quad (12.16)
$$
$$
\hat{\text{dh}}_5 + \hat{\text{dh}}_4 - 7.049 = 0 \quad (12.17)
$$

Como puede ver arriba, solo las Ecuaciones (12.14) y (12.17) son diferentes en las ecuaciones de condición; las dos ecuaciones no deben aparecer juntas para no crear problemas al resolver las ecuaciones. Las ecuaciones de condición deben ser únicas (independientes entre sí). Por ejemplo, se puede ver que la Ecuación (12.12) más la Ecuación (12.13) menos la Ecuación (12.14) da exactamente la Ecuación (12.17); por lo tanto, no es necesario agregar la Ecuación (12.17) nuevamente a las tres ecuaciones. De manera similar, la Ecuación (12.15) más la Ecuación (12.16) menos la Ecuación (12.17) da exactamente la Ecuación (12.14). Considerando las Ecuaciones (12.8)–(12.11), se dice que la Ecuación (12.11) es redundante en el sentido de que proporciona información que ya está contenida en las otras ecuaciones.

La otra forma de formular las ecuaciones de modelos condicionales para el problema en la Figura 12.5 es la siguiente:

- Siga la ruta de nivelación BM1-A-BM2 para obtener la siguiente ecuación:

$$
\hat{\text{dh}}_1 + \hat{\text{dh}}_2 - (805.410 - 785.232) = 0 \quad (12.18)
$$

- Siga la ruta de nivelación BM4-B-BM3 para obtener la siguiente ecuación:

$$
\hat{\text{dh}}_5 + \hat{\text{dh}}_4 - (794.881 - 801.930) = 0 \quad (12.19)
$$

- Dado que solo hay una observación más ($\text{dh}_3$) que debe satisfacerse, cualquier ruta que acomode esta observación será aceptable. Siguiendo la ruta BM1-A-B-BM3 se obtiene la siguiente ecuación:

$$
\hat{\text{dh}}_1 - \hat{\text{dh}}_3 + \hat{\text{dh}}_4 - (794.881 - 785.232) = 0 \quad (12.20)
$$

Las Ecuaciones (12.18)–(12.20) son las ecuaciones de condición linealmente independientes que se pueden usar en lugar de las dadas en las Ecuaciones (12.12)–(12.17). Tenga en cuenta que cualquiera de estos tres conjuntos de ecuaciones producirá resultados idénticos en el ajuste por mínimos cuadrados de las ecuaciones.

**Ejemplo 12.6** En un polígono cerrado mostrado en la Figura 12.6, se midieron los siguientes valores: ángulos $\theta_1$ y $\theta_2$; distancias $d_1$, $d_2$, y $d_3$; y rumbos $\alpha_1$ y $\alpha_f$. Si las coordenadas del punto 1 $(x_1, y_1)$ y del punto 4 $(x_4, y_4)$ están fijas, dé todas las ecuaciones de modelo condicional necesarias $f(\hat{\ell}) = 0$ para las mediciones.

**Solución:**

Todos los ajustes del polígono tienen un máximo de tres condiciones a satisfacer de la siguiente manera:

1) La diferencia entre el rumbo hacia adelante $\alpha_f$ y el rumbo hacia adelante calculado $\alpha'_f$ basado en el rumbo inicial $\alpha_1$ y los ángulos medidos en los puntos del polígono debe ser igual a cero. Esto es para verificar la precisión de las mediciones angulares y se puede expresar matemáticamente como

$$
\hat{\alpha}_f - (\hat{\alpha}_1 + \hat{\theta}_1 + \hat{\theta}_2) = 0 \quad (12.21)
$$

2) La diferencia entre el cambio en las coordenadas este (easting) de los puntos de inicio y final del polígono y la suma de los cambios en las coordenadas este de las diferentes patas del polígono debe ser igual a cero. Esto es para verificar tanto los errores en las distancias como los errores residuales en las mediciones angulares, a lo largo de la dirección este. Esto se puede expresar matemáticamente como:

$$
(x_4 - x_1) - \hat{d}_1 \sin \hat{\alpha}_1 - \hat{d}_2 \sin (\hat{\alpha}_1 + \hat{\theta}_1 + 180) - \hat{d}_3 \sin (\hat{\alpha}_1 + \hat{\theta}_1 + \hat{\theta}_2) = 0 \quad (12.22)
$$

3) La diferencia entre el cambio en las coordenadas norte (northing) de los puntos de inicio y final del polígono y la suma de los cambios en las coordenadas norte de las diferentes patas del polígono debe ser igual a cero. Al igual que en el caso este, esto también es para verificar tanto los errores en las distancias como los errores residuales en las mediciones angulares, a lo largo de la dirección norte. Esto también se puede expresar matemáticamente como:

$$
(y_4 - y_1) - \hat{d}_1 \cos \hat{\alpha}_1 - \hat{d}_2 \cos (\hat{\alpha}_1 + \hat{\theta}_1 + 180) - \hat{d}_3 \cos (\hat{\alpha}_1 + \hat{\theta}_1 + \hat{\theta}_2) = 0 \quad (12.23)
$$

### 12.3 Formulación de Ajuste de Modelos Condicionales

Los pasos involucrados en el ajuste de modelos condicionales se derivan en esta sección. Por ejemplo, considere el siguiente modelo condicional:

$$
f(\hat{\ell}) = 0 \quad P \quad (12.24)
$$

donde $f$ es el vector del modelo matemático, $\hat{\ell}$ es el vector de observaciones ajustadas, y $P$ es la matriz de pesos del vector de observación original $\ell$. Los pasos lógicos involucrados en la obtención del vector de solución para este modelo se pueden dar de la siguiente manera:

- Linearizar el modelo matemático si es no lineal.
- Formular la función de variación, que implica imponer el criterio de mínimos cuadrados del problema en el modelo linearizado.
- Derivar las ecuaciones normales de mínimos cuadrados a partir de la función de variación.
- Resolver las ecuaciones normales de mínimos cuadrados para obtener las correcciones que se aplicarán a los valores aproximados de las cantidades desconocidas.

Para linearizar el modelo matemático en la Ecuación (12.24), asuma lo siguiente:

$$
\hat{\ell} = \ell + v \quad (12.25)
$$

donde $v$ es un vector de residuos de observación que se aplicará al vector de observación original, $\ell$. A partir de las Ecuaciones (12.24) a (12.25), se puede obtener lo siguiente:

$$
f(\ell + v) = 0 \quad (12.26)
$$

Mediante la linearización por expansión en serie de Taylor, la Ecuación (12.26) en forma linearizada se puede dar como:

$$
f(\ell) + \frac{\partial f}{\partial \ell}v = 0 \quad (12.27)
$$

La Ecuación linearizada (12.27) se puede expresar de forma más simple como:

$$
w + Bv = 0 \quad (12.28)
$$

donde $B = \frac{\partial f}{\partial \ell}$ se conoce comúnmente como la segunda matriz de diseño y $w = f(\ell)$ es un vector de cierres. La Ecuación (12.27) o Ecuación (12.28) es una forma linearizada de la Ecuación (12.24).

La función de variación se puede formular de la siguiente manera. Dado que la matriz de pesos $P$ está directamente asociada con el modelo condicional en la Ecuación (12.24), el criterio de mínimos cuadrados para este problema será:

$$
v^T P v = \text{mínimo} \quad (12.29)
$$

Imponiendo el criterio de mínimos cuadrados en el modelo linearizado en la Ecuación (12.28) basado en el enfoque Lagrangiano para dar la siguiente función de variación:

$$
\varphi = v^T P v - 2k^T (Bv + w) = \text{mínimo} \quad (12.30)
$$

donde $k$ es un vector de correlatos o un vector de multiplicadores de Lagrange. El sistema de ecuaciones normales de mínimos cuadrados se puede derivar encontrando las derivadas parciales de la función de variación en la Ecuación (12.30) con respecto a las incógnitas (es decir, $v$ y $k$) y estableciéndolas en cero de la siguiente manera:

$$
\frac{\partial \varphi}{\partial v} = 2v^T P - 2k^T B = 0 \quad (12.31)
$$
$$
\frac{\partial \varphi}{\partial k^T} = -2(Bv + w) = 0 \quad (12.32)
$$

El resumen de la forma más expandida de las Ecuaciones normales (12.31) y (12.32) es:

$$
v^T P - k^T B = 0 \quad (12.33)
$$
$$
Bv + w = 0 \quad (12.34)
$$

Las Ecuaciones (12.33) y (12.34) se pueden resolver simultáneamente para las cantidades desconocidas $k$ y $v$ de la siguiente manera. A partir de la Ecuación (12.33), se puede obtener lo siguiente:

$$
v = P^{-1} B^T k \quad (12.35)
$$

Ponga la Ecuación (12.35) en la Ecuación (12.34) para obtener:

$$
BP^{-1} B^T k + w = 0 \quad (12.36)
$$

Si $Q = P^{-1}$ y $M = BQB^T$, se pueden derivar las siguientes funciones de solución a partir de las Ecuaciones (12.36) y (12.35):

$$
k = -M^{-1}w \quad (12.37)
$$
$$
v = QB^T k \quad (12.38)
$$

A partir de las Ecuaciones (12.25) y (12.38), se puede determinar la siguiente solución para la observación ajustada ($\hat{\ell}$):

$$
\hat{\ell} = \ell + v \quad (12.39)
$$

### 12.3.1 Pasos de Ajuste del Modelo Condicional

El resumen de los pasos para el ajuste por mínimos cuadrados condicional se da de la siguiente manera:

1) Formular las ecuaciones del modelo condicional:

$$
f(\hat{\ell}) = 0 \quad (12.40)
$$

donde

$$
\hat{\ell} = \ell + v \quad (12.41)
$$

2) Formar la segunda matriz de diseño ($B$) y el vector de cierres ($w$):

$$
B = \frac{\partial f}{\partial \ell} \quad \text{y} \quad w = f(\ell) \quad (12.42)
$$

3) Formar las ecuaciones normales:

$$
BQB^T k + w = 0 \quad (12.43)
$$

o

$$
Mk + w = 0 \quad (12.44)
$$

donde

$$
M = BQB^T \quad (12.45)
$$

$$
Q = P^{-1} \quad (12.46)
$$

o

$$
Q = C_\ell \ \text{(asumiendo} \ \sigma_0^2 = 1\text{)} \quad (12.47)
$$

$Q$ es la matriz de cofactores, $C_\ell$ es la matriz de covarianza de las observaciones, $P$ es la matriz de pesos de las observaciones, y $k$ es un vector de correlatos o vector de multiplicadores de Lagrange.

4) Resolver para $k$ en las ecuaciones normales en la Ecuación (12.44):

$$
k = -M^{-1}w \quad (12.48)
$$

5) Calcular el vector de residuos ($v$):

$$
v = QB^T k \quad (12.49)
$$

6) Calcular las observaciones ajustadas como:

$$
\hat{\ell} = \ell + v \quad (12.50)
$$

Los pasos de ajuste del modelo condicional dados anteriormente se ilustran numéricamente en los siguientes ejemplos.

### 12.4 Modelo Estocástico del Ajuste Condicional

El modelo estocástico para los ajustes condicionales se deriva de la misma manera que en el caso del ajuste del modelo paramétrico mencionado en la Sección 5.8. Recordando de la Sección 1.7 y el Capítulo 5 la relación entre la matriz de varianza-covarianza ($C_{\ell}$) del vector de observación $\ell$ y la matriz de cofactores ($Q_{\ell}$) del vector de observación, se tiene:

$$
C_{\ell} = \sigma_0^2 Q_{\ell} \quad (12.55)
$$

donde $\sigma_0^2$ es un factor de varianza a priori de unidad de peso (que siempre se toma como la varianza poblacional general con un valor de 1.00). Dado que el vector de observaciones $\ell$ y su matriz de varianza-covarianza $C_{\ell}$ se utilizan para calcular las cantidades desconocidas, como las observaciones ajustadas $\hat{\ell}$ y los residuos de observación ($v$), es evidente que los errores en esas observaciones se propagarán a estas cantidades también. Los conceptos de propagación de varianza-covarianza (Capítulo 2) se utilizarán a continuación para derivar la matriz de cofactores de las observaciones ajustadas ($Q_{\hat{\ell}}$) y la matriz de cofactores de los residuos de observación ($Q_v$).

#### 12.4.1 Derivación de la Matriz de Cofactores de las Observaciones Ajustadas

El vector de observaciones ajustadas por mínimos cuadrados se da a partir de las Ecuaciones (12.42), (12.48), (12.49), y (12.50) como:

$$
\hat{\ell} = \ell - Q_{\ell} B^T M^{-1} f(\ell) \quad (12.56)
$$

La Ecuación (12.56) es una forma de $\hat{\ell} = f(\ell)$ con la matriz de cofactores propagada usual (basada en las reglas de propagación de la varianza-covarianza) dada como:

$$
Q_{\hat{\ell}} = JQ_{\ell} J^T \quad (12.57)
$$

donde $Q_{\ell}$ es la matriz de cofactores de las observaciones y $J$ es la matriz Jacobiana dada como sigue:

$$
J = \frac{\partial \hat{\ell}}{\partial \ell} = I - Q_{\ell} B^T M^{-1} B \quad (12.58)
$$

Teniendo en cuenta que la derivada parcial del vector $\ell$ con respecto al vector $\ell$ da una matriz identidad $I$ y la derivada parcial de $f(\ell)$ con respecto a $\ell$ da la segunda matriz de diseño $B$ (basada en la Ecuación 12.42). Sustituyendo la Ecuación (12.58) en la Ecuación (12.57) se obtiene lo siguiente:

$$
Q_{\hat{\ell}} = (I - Q_{\ell} B^T M^{-1} B) Q_{\ell} (I - Q_{\ell} B^T M^{-1} B)^T \quad (12.59)
$$

Al multiplicar post el primer paréntesis por $Q_{\ell}$ y transponer los elementos en el otro paréntesis se obtiene la siguiente ecuación:

$$
Q_{\hat{\ell}} = (I - Q_{\ell} B^T M^{-1} B) Q_{\ell} (I - B^T M^{-1} B Q_{\ell})^T \quad (12.60)
$$

Expandiendo la Ecuación (12.60) se obtiene lo siguiente:

$$
Q_{\hat{\ell}} = Q_{\ell} - Q_{\ell} B^T M^{-1} B Q_{\ell} - Q_{\ell} B^T M^{-1} B Q_{\ell} + Q_{\ell} B^T M^{-1} (B^T Q_{\ell} B) M^{-1} B Q_{\ell} \quad (12.61)
$$

Recordando que $M = (B^T Q_{\ell} B)$ y $(B^T Q_{\ell} B) M^{-1} = I$, esto reduce la Ecuación (12.61) a lo siguiente:

$$
Q_{\hat{\ell}} = Q_{\ell} - Q_{\ell} B^T M^{-1} B Q_{\ell} \quad (12.62)
$$

Los últimos tres términos en la Ecuación (12.62) son idénticos, por lo que los últimos dos términos se cancelan, dando la forma reducida como sigue:

$$
Q_{\hat{\ell}} = Q_{\ell} - Q_{\ell} B^T M^{-1} B Q_{\ell} \quad (12.63)
$$

La Ecuación (12.63) es la matriz de cofactores de las observaciones ajustadas $\hat{\ell}$.

### 12.4.2 Derivación de la Matriz de Cofactores de los Residuos de Observación

El vector de residuos de observación se obtiene a partir de las Ecuaciones (12.49), (12.48) y (12.42) como:

$$
v = -Q_{\ell} B^T M^{-1} f(\ell) \quad (12.64)
$$

La Ecuación (12.64) es una forma de $v = f(\ell)$ con la matriz de cofactores propagada usual (basada en las reglas de propagación de la varianza-covarianza) dada como:

$$
Q_v = J Q_{\ell} J^T \quad (12.65)
$$

donde $Q_{\ell}$ es la matriz de cofactores de las observaciones y $J$ es la matriz Jacobiana dada como sigue:

$$
J = \frac{\partial v}{\partial \ell} = -Q_{\ell} B^T M^{-1} B \quad (12.66)
$$

Sustituyendo la Ecuación (12.66) en la Ecuación (12.65) se obtiene lo siguiente:

$$
Q_v = \left(-Q_{\ell} B^T M^{-1} B\right) Q_{\ell} \left(-Q_{\ell} B^T M^{-1} B\right)^T \quad (12.67)
$$

Aplicando la transposición a los elementos del segundo paréntesis en la Ecuación (12.67) y dejando que los dos negativos se multipliquen, se obtiene lo siguiente:

$$
Q_v = \left(Q_{\ell} B^T M^{-1} B\right) Q_{\ell} \left(B^T M^{-1} B Q_{\ell}\right) \quad (12.68)
$$

La Ecuación (12.68) en su forma expandida se puede dar como:

$$
Q_v = Q_{\ell} B^T M^{-1} \left(B Q_{\ell} B^T\right) M^{-1} B Q_{\ell} \quad (12.69)
$$

Tomando $M = \left(B^T Q_{\ell} B\right)$ y $\left(B^T Q_{\ell} B\right) M^{-1} = I$, la Ecuación (12.69) se reduce a lo siguiente:

$$
Q_v = Q_{\ell} B^T M^{-1} B Q_{\ell} \quad (12.70)
$$

La Ecuación (12.70) es la matriz de cofactores de los residuos de observación ($v$).

### 12.4.3 Matrices de Covarianza de las Observaciones Ajustadas y Residuos

Las matrices de varianza-covarianza de las cantidades ajustadas se pueden obtener a partir de las Ecuaciones (12.63) y (12.70), respectivamente, como sigue:

$$
C_{\hat{\ell}} = s_0^2 Q_{\ell} - s_0^2 Q_{\ell} B^T M^{-1} B Q_{\ell} \quad (12.71)
$$

$$
C_v = s_0^2 Q_{\ell} B^T M^{-1} B Q_{\ell} \quad (12.72)
$$

donde $C_{\hat{\ell}}$ es la matriz de covarianza de las observaciones ajustadas, $C_v$ es la matriz de covarianza de los residuos de observación, y $s_0^2$ es el factor de varianza a posteriori (APVF) de unidad de peso, que se calcula después del ajuste por mínimos cuadrados como sigue:

$$
s_0^2 = \frac{v^T Pv}{r} \quad (12.73)
$$

donde $v$ es el vector de residuos, $P$ es la matriz de pesos de las observaciones, y $r$ es la redundancia (o los grados de libertad). Los grados de libertad $r$ son el número de ecuaciones de condición independientes formuladas.

De las Ecuaciones (12.71) a (12.72), se puede observar que la matriz de covarianza de la observación ajustada se puede dar como:

$$
C_{\hat{\ell}} = \hat{C}_{\ell} - C_v \quad (12.74)
$$

donde $\hat{C}_{\ell} = s_0^2 Q_{\ell}$ es la matriz de covarianza escalada de las observaciones originales $\ell$, que no debe confundirse con $C_{\ell} = \sigma_0^2 Q_{\ell}$ (con $\sigma_0^2 = 1$), que es la matriz de covarianza no escalada de las observaciones originales, utilizada al comienzo del ajuste por mínimos cuadrados. La Ecuación (12.74) se puede interpretar como que la matriz de covarianza de la observación ajustada es siempre menor que la matriz de covarianza de la observación original.

### 12.5 Evaluación de Observaciones y del Modelo Condicional

Los cierres en el ajuste del modelo condicional pueden ser probados para verificar si tienen una función de densidad de probabilidad normal, siguiendo la idea de pruebas globales y locales en el Capítulo 11. Esto se hace probando las formas cuadráticas de los cierres de la siguiente manera:

i) Si la matriz de covarianza ($C_w$) del vector de cierres ($w$) es bien conocida y el factor de varianza a priori de unidad de peso es bien conocido, la forma cuadrática de los cierres se considerará como teniendo una función de densidad de probabilidad normal si se cumple lo siguiente:

$$ \Large{
\chi^2_{\alpha/2, df = m} < w^T C_w^{-1} w < \chi^2_{1-\alpha/2, df = m}} \quad (12.75)
$$

donde $\chi^2_{\alpha/2, df = m}$ es la función de densidad de chi-cuadrado de cola baja con el número de grados de libertad $df = m$ para $m$ número de cierres. Esto es similar a la prueba global discutida en el Capítulo 11.

ii) Si se utiliza la matriz de covarianza escalada ($\hat{C}_w$) del vector de cierres ($w$) porque no es bien conocida y el factor de varianza a priori de unidad de peso es desconocido pero se utiliza el factor de varianza a posteriori de unidad de peso ($s_0^2$), la forma cuadrática de los cierres se considerará como teniendo una función de densidad de probabilidad normal si se cumple lo siguiente:

$$ \Large{
F_{\alpha/2, df_1 = m, df_2 = m-u} < \frac{w^T \hat{C}_w^{-1} w}{m} < F_{1-\alpha/2, df_1 = m, df_2 = m-u}} \quad (12.76)
$$

donde $F_{\alpha/2, df_1 = m, df_2 = m-u}$ es la función de densidad $F$ (o de Fisher) de cola baja con el primer grado de libertad $df_1 = m$ (número de cierres) y $df_2 = m-u$ con $u$ como el número de parámetros desconocidos en el ajuste por mínimos cuadrados.

La discrepancia individual también puede evaluarse para verificar si tiene una función de densidad de probabilidad normal de la siguiente manera:

a) Si la matriz de covarianza ($C_w$) del vector de discrepancias ($w$) es bien conocida y el factor de varianza a priori de unidad de peso es bien conocido, la discrepancia estandarizada se considerará como teniendo una función de densidad de probabilidad normal si se cumple lo siguiente (lo que significa que la discrepancia no es un valor atípico):

$$
\left|\frac{w_i}{\sigma_{w_i}}\right| < z_{1-\alpha/2} \quad (12.77)
$$

donde $\sigma_{w_i}$ es la desviación estándar de la discrepancia $w_i$.

b) Si se utiliza la matriz de covarianza escalada ($\hat{C}_w$) del vector de discrepancias ($w$) porque no es bien conocida y el factor de varianza a priori de unidad de peso es desconocido pero se utiliza el factor de varianza a posteriori de unidad de peso ($s_0^2$), la discrepancia estandarizada se considerará como teniendo una función de densidad de probabilidad normal si se cumple lo siguiente (lo que significa que la discrepancia no es un valor atípico):

$$
\left|\frac{w_i}{\hat{\sigma}_{w_i}}\right| < t_{1-\alpha/2, m-u} \quad (12.78)
$$

donde $\hat{\sigma}_{w_i}$ es la desviación estándar escalada de la discrepancia $w_i$, $t_{1-\alpha/2, m-u}$ es la densidad $t$ de Student (asumiendo que las observaciones utilizadas en el cálculo de $w$ y $s_0^2$ son diferentes), $m$ es el número de discrepancias, y $u$ es el número de parámetros desconocidos en el ajuste por mínimos cuadrados. El $t_{1-\alpha/2, m-u}$ será reemplazado por la densidad $\tau$ con los grados de libertad $m-u$ si las observaciones utilizadas en el cálculo de $w$ también se utilizan para determinar $s_0^2$.

### 12.6 Propagación de la Varianza-Covarianza para Parámetros Derivados del Ajuste Condicional

La relación entre los parámetros ajustados ($\hat{x}$) y las observaciones ajustadas ($\hat{\ell}$) del ajuste condicional por mínimos cuadrados se puede dar de la siguiente manera:

$$
\hat{x} = f(\hat{\ell}) \quad (12.79)
$$

donde las observaciones ajustadas ($\hat{\ell}$) se utilizan para derivar los parámetros desconocidos ($\hat{x}$). Según las leyes de la propagación de la varianza-covarianza, la matriz de covarianza ($C_{\hat{x}}$) de los parámetros ajustados se puede dar como:

$$
C_{\hat{x}} = J C_{\hat{\ell}} J^T \quad (12.80)
$$

donde $J$ es la matriz Jacobiana de la Ecuación (12.79) con respecto a todas las observaciones ajustadas del ajuste condicional por mínimos cuadrados, y $C_{\hat{\ell}}$ es la matriz de covarianza de las observaciones ajustadas. El paso más importante en el proceso computacional es poder formular la Ecuación (12.79), que es un conjunto de ecuaciones utilizadas para determinar el vector de parámetros ajustados a partir de las observaciones ajustadas.

Los siguientes ejemplos se proporcionan para ilustrar los pasos computacionales involucrados en la derivación de vectores de parámetros ajustados y sus matrices de covarianza basadas en los resultados del ajuste por mínimos cuadrados de modelos condicionales.
