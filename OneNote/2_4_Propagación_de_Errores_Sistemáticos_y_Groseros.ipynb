### 2.4 Propagación de Errores Sistemáticos y Groseros

Usualmente en topografía, las mediciones se corrigen por errores sistemáticos antes de que se utilicen en cualquier cálculo. Si esto no se hace, los errores sistemáticos se propagarán en las cantidades calculadas de cierta manera, lo cual se puede derivar utilizando el enfoque de linearización de series de Taylor. De manera similar, si no se detectan errores groseros en la observación, los errores se propagarán de cierta manera como en la propagación de errores sistemáticos. Al evaluar cuidadosamente el modelo lineal en la Ecuación (2.12), se puede ver que el término de primer orden de la serie de Taylor constituye las correcciones sistemáticas o errores sistemáticos (o groseros) que se aplicarán a los parámetros iniciales ($x$) estimados a partir de las observaciones ($\ell$). Las correcciones sistemáticas o errores sistemáticos propagados desde las mediciones a las estimaciones iniciales de los parámetros pueden, por lo tanto, expresarse a partir de la Ecuación (2.12) como:

$$
\Delta x = J (\Delta \ell) \tag{2.34}
$$

donde $\Delta x$ es un vector de correcciones sistemáticas o errores sistemáticos que se aplicarán a los parámetros iniciales. Para ilustrar el procedimiento de propagación del error sistemático (corrección), refiérase a la Ecuación (2.1) para el caso en el que un parámetro $x = (x_1)$ y tres observaciones $\ell = (\ell_1, \ell_2, \ell_3)^T$ se miden; el siguiente modelo matemático general se puede formular:

$$
x_1 = f(\ell_1, \ell_2, \ell_3) \tag{2.35}
$$

El error sistemático ($dx_1$) en el parámetro $x_1$ debido a los errores sistemáticos $d\ell_1, d\ell_2, d\ell_3$ de las mediciones $\ell_1, \ell_2, \ell_3$ respectivamente, puede darse como:

$$
dx_1 = \left( \frac{\partial f}{\partial \ell_1} \right) d\ell_1 + \left( \frac{\partial f}{\partial \ell_2} \right) d\ell_2 + \left( \frac{\partial f}{\partial \ell_3} \right) d\ell_3 \tag{2.36}
$$

Para simplificar la Ecuación (2.36), las notaciones matriciales son comúnmente empleadas utilizando el enfoque de Jacobiana; en este caso, la Ecuación (2.36) se puede expresar como la Ecuación (2.34) de la siguiente manera:

$$
dx_1 = J (\Delta \ell) \tag{2.37}
$$

donde

$$ \Large{
\Delta \ell = [d\ell_1 \quad d\ell_2 \quad d\ell_3]^T \tag{2.38}}
$$

$$\Large{
J = \begin{bmatrix}
\frac{\partial f}{\partial \ell_1} & \frac{\partial f}{\partial \ell_2} & \frac{\partial f}{\partial \ell_3}
\end{bmatrix} \tag{2.39}}
$$

Las Ecuaciones (2.36) y (2.37) representan las leyes de propagación del error sistemático (corrección).

**Ejemplo 2.9**  
Considera el terreno mostrado en la Figura 2.8 con las dimensiones medidas como $\ell_1 = 50 \, \text{m}, \ell_2 = 20 \, \text{m}, \ell_3 = 30 \, \text{m}$. El terreno se va a dividir en dos partes a lo largo de la línea quebrada que conecta los puntos C y F. Las dos áreas $A_1$ y $A_2$ se expresan mediante las siguientes funciones, respectivamente:

$$
A_1 = \frac{\pi \ell_1^2}{4} + \frac{\ell_1 \ell_3}{2} + \frac{\ell_2 \ell_3}{2}
$$

$$
A_2 = \frac{\ell_1 \ell_3}{2} + \frac{\ell_2 \ell_3}{2}
$$

Responde lo siguiente:

a) Evalúa la matriz Jacobiana $J$.

**Solución:**

La matriz Jacobiana se forma considerando que hay tres mediciones ($\ell_1, \ell_2, \ell_3$) involucradas en las dos ecuaciones que representan las dos áreas ($A_1, A_2$). Las tres mediciones dan tres columnas, mientras que las dos ecuaciones dan dos filas de manera que la matriz Jacobiana es una matriz $2 \times 3$, dada de la siguiente manera:

$$ \Large{
J = \begin{bmatrix}
\frac{\partial A_1}{\partial \ell_1} & \frac{\partial A_1}{\partial \ell_2} & \frac{\partial A_1}{\partial \ell_3} \\
\frac{\partial A_2}{\partial \ell_1} & \frac{\partial A_2}{\partial \ell_2} & \frac{\partial A_2}{\partial \ell_3}
\end{bmatrix} \rightarrow J = \begin{bmatrix}
\left( \frac{\pi \ell_1}{2} + \frac{\ell_3}{2} \right) & \frac{\ell_3}{2} & \frac{\ell_1}{2} + \frac{\ell_2}{2} \\
\frac{\ell_3}{2} & \frac{\ell_3}{2} & \frac{\ell_1}{2} + \frac{\ell_2}{2}
\end{bmatrix}}
$$

b) Si las correcciones que se aplican a las dimensiones ($\ell_1 = 50 \, \text{m}, \ell_2 = 20 \, \text{m}, \ell_3 = 30 \, \text{m}$) son 0.02, -0.04 y 0.03 m, respectivamente, evalúa las correcciones a las áreas $A_1$ y $A_2$ aplicando las leyes de propagación de errores sistemáticos.

**Solución:**

Las correcciones a las áreas se determinarán en este problema siguiendo la ley de propagación de correcciones sistemáticas en la Ecuación (2.34):

$$
\begin{bmatrix}
dA_1 \\
dA_2
\end{bmatrix} = J \begin{bmatrix}
\Delta \ell_1 \\
\Delta \ell_2 \\
\Delta \ell_3
\end{bmatrix} \rightarrow \begin{bmatrix}
dA_1 \\
dA_2
\end{bmatrix} = \begin{bmatrix}
54.3 & 15 & 35 \\
15 & 15 & 35
\end{bmatrix} \begin{bmatrix}
0.02 \\
-0.04 \\
0.03
\end{bmatrix}
$$

$$
\begin{bmatrix}
dA_1 \\
dA_2
\end{bmatrix} = \begin{bmatrix}
1.536 \\
0.75
\end{bmatrix} \, \text{m}^2
$$

**Ejemplo 2.10**
Un terreno rectangular de tierra ABCD en la Figura 2.9 mide 50.170 m x 61.090 m. Si se usa la misma cinta de 30 m (0.030 m demasiado larga, es decir, la cinta se ha estirado pero aún muestra el valor nominal de 30 m) para hacer las mediciones, evalúe el error en el área calculada del terreno:

$$ A = \ell_1 \ell_2 $$

**Solución:**

La Matriz Jacobiana $ J $:

$$ J = \begin{bmatrix} \frac{\partial A}{\partial \ell_1} & \frac{\partial A}{\partial \ell_2} \end{bmatrix} \rightarrow J = \begin{bmatrix} \ell_2 & \ell_1 \end{bmatrix} $$

Los errores se calculan sobre la base de que la cinta de 30 m es demasiado larga en 0.030 m, de la siguiente manera:

$$
\Delta \ell_1 = \frac{-0.030}{30} (50.170) = -0.050 \, \text{m}
$$

$$
\Delta \ell_2 = \frac{-0.030}{30} (61.090) = -0.061 \, \text{m}
$$

Aplicando la regla de propagación de errores sistemáticos en la Ecuación (2.34):

$$
\text{Error del Área: } dA = J \begin{bmatrix} \Delta \ell_1 \\ \Delta \ell_2 \end{bmatrix} \rightarrow dA = \begin{bmatrix} 61.090 & 50.170 \end{bmatrix} \begin{bmatrix} -0.050 \\ -0.061 \end{bmatrix}
$$

$$
\begin{bmatrix} dA \end{bmatrix} = -6.1 \, \text{m}^2
$$

El error sistemático propagado en el área es $ dA = -6.1 \, \text{m}^2 $.
