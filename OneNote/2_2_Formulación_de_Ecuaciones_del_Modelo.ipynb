## 2.2 Formulación de Ecuaciones del Modelo

Las ecuaciones de los modelos típicos de geomática se formulan en esta sección. Para evaluar las propiedades de error de los parámetros, primero se deben formular los parámetros en términos de las observaciones, como se indica en la Ecuación (2.1). Esto es importante ya que la propagación de errores en las mediciones se realizará matemáticamente. Los pasos generales para formular ecuaciones de modelos para la propagación de errores son los siguientes:

1. Los elementos importantes a manipular en la Ecuación (2.1) son los parámetros ($x$) y las observaciones ($\ell$). Es necesario poder identificar a partir de una pregunta qué constituye los parámetros ($x$) y qué constituye las observaciones ($\ell$) y usar símbolos adecuados para representar cada uno de ellos; sin esto, no se podrá formular las ecuaciones necesarias.
2. Organizar los elementos que representan los parámetros ($x$) y las observaciones ($\ell$) en formas de ecuaciones tales que cada ecuación tenga un parámetro ($x$) como sujeto. Cada ecuación involucrará algunas o todas las observaciones (con la posibilidad de que algunos valores constantes estén presentes), pero no se deben encontrar otros parámetros desconocidos en la ecuación para que satisfaga el modelo $x = f(\ell)$. Cada uno de los parámetros identificados y las observaciones deben representarse en las ecuaciones como símbolos, con la sustitución de valores numéricos en las ecuaciones realizada más tarde, en la etapa de solución.

Sobre la base de los pasos anteriores, se dan los siguientes ejemplos para ilustrar algunos de los modelos geomáticos comunes que eventualmente pueden requerir una aplicación de reglas de propagación de errores.

**Ejemplo 2.1**  
Dos lados de un área triangular rectángula se muestran en la Figura 2.1 y se miden como $\ell_1$ y $\ell_2$; se requiere que se determine el área y la hipotenusa del triángulo. Formule las ecuaciones ($x = f(\ell)$) para calcular el área y la hipotenusa.

**Solución:**

El modelo esperado debe tener la forma $x = f(\ell)$. Los pasos proporcionados anteriormente en esta sección se tomarán de la siguiente manera:

1) Dado que la pregunta indica que $\ell_1$ y $\ell_2$ fueron medidos, representan las observaciones ($\ell$) en forma simbólica (no en forma numérica) y se pueden usar directamente para formular las ecuaciones de área e hipotenusa del triángulo, que son los parámetros ($x$). Estos parámetros se pueden representar como símbolos: $A$ para el área y $H$ para la hipotenusa. Los elementos de las ecuaciones se dan en forma de vector de la siguiente manera:

$$ x = \begin{bmatrix} A \\ H \end{bmatrix} \quad \ell = \begin{bmatrix} \ell_1 \\ \ell_2 \end{bmatrix} $$

2) Considerando el tipo de modelo a formular ($x = f(\ell)$), se entiende que cada parámetro es un sujeto y debe formar una ecuación como función de las observaciones sin que se encuentren otros parámetros desconocidos en esa ecuación. Dado que hay dos parámetros, se formulan las siguientes dos ecuaciones:

$$ A = \frac{1}{2} \ell_1 \ell_2 \tag{2.2} $$

$$ H = \sqrt{\ell_1^2 + \ell_2^2} \tag{2.3} $$

**Ejemplo 2.2**  
En un levantamiento horizontal de una área triangular BAC mostrado en la Figura 2.2, las mediciones son el ángulo ($\theta$) en el punto fijo A, el rumbo ($\alpha_{AB}$) y la distancia ($S_{AC}$). Si las coordenadas ($E_A, N_A$) del punto A están fijadas, da las ecuaciones ($x = f(\ell)$) para calcular las coordenadas ($E_C, N_C$) del punto C.

**Solución:**

Las ecuaciones a formular deben tener la forma $x = f(\ell)$. Los pasos de solución tomados en el Ejemplo 2.1 se seguirán en esta pregunta:

1) Las mediciones indicadas en la pregunta son el ángulo $\theta$, el rumbo $\alpha_{AB}$ y la distancia $S_{AC}$, que constituyen las observaciones ($\ell$). Las coordenadas (el este ($E_C$) y el norte ($N_C$) del punto C son los parámetros a calcular. Dado que las observaciones ya están en forma simbólica (no en forma numérica), se pueden usar directamente para formular las ecuaciones de la siguiente manera. Los parámetros y observaciones se dan respectivamente como sigue:

$$ x = \begin{bmatrix} E_C \\ N_C \end{bmatrix} \quad \ell = \begin{bmatrix} \theta \\ \alpha_{AB} \\ S_{AC} \end{bmatrix} $$

2) Las ecuaciones del modelo formulado ($x = f(\ell)$) para el recorrido se pueden dar como sigue:

$$ E_C = E_A + S_{AC} \sin(\alpha_{AB} + \theta) \tag{2.4} $$

$$ N_C = N_A + S_{AC} \cos(\alpha_{AB} + \theta) \tag{2.5} $$

Se puede ver en las ecuaciones anteriores que cada parámetro forma una ecuación y solo las constantes ($E_A, N_A$) y las observaciones se encuentran en los lados derechos de las ecuaciones, dando un modelo en la forma de $x = f(\ell)$. Nótese que las constantes son necesarias (pero no funcionales) partes de las ecuaciones.

**Ejemplo 2.3**  
Los límites de dos lotes rectangulares se determinaron con mediciones de tres distancias, como se muestra en la Figura 2.3. Da las ecuaciones para calcular las áreas ($A_1, A_2$) de los dos lotes, expresándolas como $x = f(\ell)$.

**Solución:**

Las ecuaciones a formular deben tener la forma $x = f(\ell)$. Los pasos en el Ejemplo 2.1 se aplican de manera similar en este problema de la siguiente manera. Los parámetros ($x$) y observaciones ($\ell$) son los siguientes:

$$ x = \begin{bmatrix} A_1 \\ A_2 \end{bmatrix} \quad \ell = \begin{bmatrix} \ell_1 \\ \ell_2 \\ \ell_3 \end{bmatrix} $$

Cada parámetro forma una ecuación como una función de las observaciones de la siguiente manera:

$$ A_1 = \ell_1 \ell_3 \tag{2.6} $$

$$ A_2 = \ell_2 \ell_3 \tag{2.7} $$

**Ejemplo 2.4**  
Para determinar la elevación ($h_B$) del punto B, se instaló una estación total en el punto A y se midió el ángulo cenital $z$ y la distancia en pendiente $s$ hacia un objetivo en el punto B. La altura del instrumento se midió como $h_i$, la altura del objetivo se midió como $h_t$ y la elevación del punto A se fijó en 30.0 m. Da el modelo para este problema en la forma de $x = f(\ell)$.

**Solución:**

La ecuación a formular debe tener la forma $x = f(\ell)$. Aplicando los mismos pasos que en el Ejemplo 2.1, se obtienen los siguientes:

1) Cantidades medidas (las observaciones): ángulo cenital $z$, distancia en pendiente $s$, altura del instrumento $h_i$ y altura del objetivo $h_t$. Las observaciones se pueden usar directamente para determinar la elevación ($h_B$) del punto B, que es el único parámetro en este problema. El parámetro ($x$) y las observaciones ($\ell$) son, respectivamente, los siguientes:

$$ x = [h_B] \quad \ell = [z \quad s \quad h_i \quad h_t]^T $$

2) La ecuación del modelo ($\hat{x} = f(\ell)$) se puede formular como un problema de nivelación trigonométrica. La elevación del punto A está fijada en 30.0 como un valor constante. Este valor constante no es una observación ya que no se asocia una desviación estándar con él, y tampoco es un parámetro ya que su valor no se va a determinar. El único parámetro en este problema se puede expresar de la siguiente manera:

$$ h_B = 30.0 + s \cos z + h_i - h_t \tag{2.8} $$

Se puede ver en la Ecuación (2.8) que el valor numérico de la constante (elevación fija del punto A) se usa directamente sin intentar representarlo como un símbolo, ya que no es un parámetro y no es una observación.
