## 6. Ajuste Paramétrico de Mínimos Cuadrados
### Aplicaciones

### 6.1 Introducción

El método de mínimos cuadrados es significativo si el número ($ n $) de ecuaciones de modelo paramétrico es mayor que el número ($ u $) de parámetros. Esto significa que el ajuste de mínimos cuadrados por método paramétrico es posible si la redundancia (grados de libertad) dada como ($ n - u $) es mayor que cero. Matemáticamente, esto significa que las ecuaciones normales de mínimos cuadrados son singulares si la redundancia es menor que cero, asumiendo que el datum ya está definido para el ajuste como se discutió en el Capítulo 4. La singularidad de las ecuaciones normales en el ajuste de mínimos cuadrados también significa que el inverso de la matriz de los coeficientes de las ecuaciones normales ($ N $) en los pasos de solución dados en la Sección 5.7 será imposible. Tenga en cuenta que cuanto mayor es la redundancia, más fiables son los resultados del ajuste.

En este capítulo, se dan algunos ejemplos de ajuste de mínimos cuadrados por el método paramétrico para ilustrar los pasos involucrados en el ajuste paramétrico de mínimos cuadrados. Continuando con las derivaciones matemáticas dadas en el Capítulo 5, el resumen de los pasos de mínimos cuadrados para el ajuste de ecuaciones de modelo paramétrico se puede dar como sigue:

1. Formular las ecuaciones del modelo paramétrico ($ \hat{\ell} = f(\hat{x}) $). En esta formulación del modelo, cada observación ($ \ell_i $) se traducirá en una ecuación involucrando solo algunos o todos (dependiendo del tipo de observable) de los parámetros desconocidos ($ x_i $) y posibles valores constantes usando símbolos para observaciones y parámetros (referirse al Capítulo 5 sobre cómo formular las ecuaciones).
2. Formar la primera matriz de diseño ($ A $), que es la matriz de las derivadas parciales de las ecuaciones paramétricas con respecto a los parámetros desconocidos. El tamaño de esta matriz será igual al número de ecuaciones o número de observaciones (como el número de filas) por el número de parámetros (como el número de columnas).
3. Elegir valores aproximados para los parámetros ($ x^0 $); estos valores son valores iniciales que se consideran lo suficientemente cercanos a lo que deberían ser sus valores finales. Por lo general, si este no es el caso, los valores ajustados basados en estos valores iniciales se pueden usar nuevamente como valores aproximados en otra ronda de ajuste, de manera iterativa, hasta que los valores ajustados finales no sean significativamente diferentes de los valores aproximados (o iniciales) anteriores. En este caso, se dice que la solución del ajuste ha convergido. Es posible que el ajuste de mínimos cuadrados tenga que repetirse tantas veces como sea posible hasta que la diferencia entre la solución final y la anterior converja a una tolerancia específica. La mejor manera de obtener valores aproximados de los parámetros de coordenadas es calcular las coordenadas a partir de observaciones seleccionadas.
4. Sustituir los valores aproximados ($ x^0 $) en las ecuaciones paramétricas formuladas en el paso 1 para obtener los llamados valores simulados de las observaciones ($ \ell^0 = f(x^0) $); usar estos valores simulados para obtener el vector de cierre ($ w = f(x^0) - \ell $) como la diferencia entre las observaciones simuladas ($ \ell^0 $) y las observaciones reales ($ \ell $).
5. Formar la matriz cofactor ($ Q $) de las observaciones (asumiendo que el factor de varianza, $ \sigma_0^2 = 1 $) posicionando cuadrados de desviaciones estándar de las observaciones a lo largo del elemento diagonal principal de la matriz (en el orden en el que se formulan las ecuaciones paramétricas en el paso 1). El inverso de esta matriz cofactor es la matriz de pesos de las observaciones ($ P $).
6. Formar los siguientes productos de matrices usando las cantidades ya determinadas: $ N = A^T PA $ y $ u = A^T Pw $. Tenga en cuenta que la matriz A se evalúa utilizando los valores aproximados ($ x^0 $) de los parámetros y $ N $ es una matriz cuadrada con su tamaño igual al número de parámetros desconocidos.
7. Resolver el sistema de ecuaciones normales para obtener las correcciones ($ \delta $) a los parámetros aproximados ($ x^0 $) como $ \delta = -N^{-1}u $.
8. Determinar los valores de los parámetros ajustados como $ \hat{x} = x^0 + \delta $.
9. Si es necesario, determinar los residuos ($ v $) de las observaciones y las observaciones ajustadas ($ \hat{\ell} $) como $ v = A\delta + w $ y $ \hat{\ell} = \ell + v $, respectivamente.

## 6.2.1 Ajuste de Nivelación

Determine las elevaciones ajustadas $ h_A $ y $ h_B $ de los puntos A y B, respectivamente, mediante el ajuste paramétrico de mínimos cuadrados de las notas de nivelación de campo en la Tabla 6.1, que se relaciona con la red de nivelación en la Figura 6.1. Suponga que las elevaciones de los puntos FH1 y FH2 son conocidas y fijas como $ h_1 = 100.000 \, \text{m} $ y $ h_2 = 99.729 \, \text{m} $, respectivamente.

**Solución:**

Las elevaciones conocidas de FH1 y FH2 definen el datum.

Parámetros ajustados: $ \hat{x} = \begin{bmatrix} \hat{h}_A \\ \hat{h}_B \end{bmatrix} $; observaciones ajustadas: $ \hat{\ell} = \begin{bmatrix} \hat{d}h_1 \\ \hat{d}h_2 \\ \hat{d}h_3 \end{bmatrix} $.

#### Tabla 6.1 Notas de campo de nivelación diferencial.

| Tramo $ i $ | BS  | FS  | Diferencia de elevación $ dh_i $ (m) | Desviación estándar (m) |
|---------------|-----|-----|----------------------------------------|--------------------------|
| 1             | FH1 | BMA | -7.341                                | 0.008                    |
| 2             | BMA | BMB |  2.495                                | 0.005                    |
| 3             | BMB | FH2 |  5.107                                | 0.004                    |

Determine las elevaciones aproximadas de los puntos A y B usando cualquier dirección de nivelación:

$$ x^0 = \begin{bmatrix} h_A^0 = 92.659 \\ h_B^0 = 95.154 \end{bmatrix} $$

Forme el modelo paramétrico (recuerde del Capítulo 5):

$$ d\hat{h}_1 = \hat{h}_A - 100 \quad (6.1) $$

$$ d\hat{h}_2 = \hat{h}_B - \hat{h}_A \quad (6.2) $$

$$ d\hat{h}_3 = 99.729 - \hat{h}_B \quad (6.3) $$

Encuentre la matriz $ A $ a partir de las Ecuaciones (6.1) a (6.3) (con los sombreros ignorados por brevedad):

$$ A = \frac{\partial \ell}{\partial x} = \begin{bmatrix} \frac{\partial d h_1}{\partial h_A} & \frac{\partial d h_1}{\partial h_B} \\ \frac{\partial d h_2}{\partial h_A} & \frac{\partial d h_2}{\partial h_B} \\ \frac{\partial d h_3}{\partial h_A} & \frac{\partial d h_3}{\partial h_B} \end{bmatrix} \rightarrow A = \begin{bmatrix} 1 & 0 \\ -1 & 1 \\ 0 & -1 \end{bmatrix} $$

Como se puede ver en la matriz $ A $ anterior, el número de columnas corresponde al número de parámetros desconocidos ($ h_A $ y $ h_B $) en ese orden, y el número de filas corresponde al número de ecuaciones. El tamaño de la matriz $ A $, por lo tanto, es siempre el mismo que el número de ecuaciones (filas) por el número de parámetros (columnas). Forme el vector de cierres $ w $ sustituyendo los valores aproximados de los parámetros $ x^0 $ y las observaciones originales $ \ell $ en las Ecuaciones (6.1)–(6.3):

$$ w = f(x^0) - \ell = \begin{bmatrix} d h_1 = -7.341 \\ d h_2 = 2.495 \\ d h_3 = 5.107 \end{bmatrix} $$

La función $ f(x^0) $ de arriba da lo siguiente:

$$ d h_1^0 = 92.659 - 100 \rightarrow -7.341 $$

$$ d h_2^0 = 95.154 - 92.659 \rightarrow 2.495 $$

$$ d h_3^0 = 99.729 - 95.154 \rightarrow 4.575 $$

$$ f(x^0) = \begin{bmatrix} -7.341 \\ 2.495 \\ 4.575 \end{bmatrix} $$

$$ w = f(x^0) - \ell \rightarrow \begin{bmatrix} -7.341 \\ 2.495 \\ 4.575 \end{bmatrix} - \begin{bmatrix} -7.341 \\ 2.495 \\ 5.107 \end{bmatrix} \rightarrow w = \begin{bmatrix} 0.000 \\ 0.000 \\ -0.532 \end{bmatrix} $$

Forme la matriz de pesos ($ P $) de las diferencias de elevación usando las desviaciones estándar dadas de las diferencias de elevación:

$$ P = \begin{bmatrix} 1 / (0.008)^2 & 0 & 0 \\ 0 & 1 / (0.005)^2 & 0 \\ 0 & 0 & 1 / (0.004)^2 \end{bmatrix} $$

Forme los siguientes productos:

$$ u = A^T P w = \begin{bmatrix} 0.000 \\ 33.250 \end{bmatrix} $$

$$ N = A^T P A = \begin{bmatrix} 55,625 & -40,000 \\ -40,000 & 102,500 \end{bmatrix} $$

Encuentre el vector de solución:

$$ \delta = \begin{bmatrix} \delta_{h_A} \\ \delta_{h_B} \end{bmatrix} = -N^{-1} u $$

$$ \delta = \begin{bmatrix} \delta_{h_A} \\ \delta_{h_B} \end{bmatrix} = \begin{bmatrix} -0.324 \\ -0.451 \end{bmatrix} $$

Las elevaciones ajustadas de BMA y BMB se dan como sigue: $ \hat{x} = x^0 + \delta $

$$ \hat{h}_A = h_A^0 + \delta_{h_A} = 92.659 - 0.324 \rightarrow 92.335 \, \text{m} $$

$$ \hat{h}_B = h_B^0 + \delta_{h_B} = 95.154 - 0.451 \rightarrow 94.703 \, \text{m} $$

Calcule los residuos de las observaciones: $ v = A\delta + w $

$$ v = A\delta + w = \begin{bmatrix} 1 & 0 \\ -1 & 1 \\ 0 & -1 \end{bmatrix} \begin{bmatrix} -0.324 \\ -0.451 \end{bmatrix} + \begin{bmatrix} 0.000 \\ 0.000 \\ -0.532 \end{bmatrix} = \begin{bmatrix} -0.324 \\ -0.127 \\ -0.081 \end{bmatrix} $$

Calcule las observaciones ajustadas como sigue: $ \hat{\ell} = \ell + v $

$$ \hat{d}h_1 = d h_1 + v_1 \rightarrow -7.341 - 0.324 = -7.665 \, \text{m} $$

$$ \hat{d}h_2 = d h_2 + v_2 \rightarrow 2.495 - 0.127 = 2.368 \, \text{m} $$

$$ \hat{d}h_3 = d h_3 + v_3 \rightarrow 5.107 - 0.081 = 4.994 \, \text{m} $$

Las observaciones ajustadas son $ \hat{\ell} = \begin{bmatrix} \hat{d}h_1 \\ \hat{d}h_2 \\ \hat{d}h_3 \end{bmatrix} = \begin{bmatrix} -7.665 \\ 2.368 \\ 4.994 \end{bmatrix} \, \text{m} $.

```matlab
% use syms to define the variables hA and hB
syms hA hB

% create the parametric Equations (6.1) to (6.3)
dh1=hA-100;
dh2=hB-hA;
dh3=99.729-hB;

% Determine the Jacobian of equations wrt hA and hB
J=jacobian([dh1;dh2;dh3], [hA hB]);
hA=92.659;hB=95.154;

% evaluate Jacobian as the A matrix, computed observations as L0 and given observations as L
A=eval(J);
L0=eval([dh1;dh2;dh3]);
L=[-7.341;2.495;5.107];

% determine misclosure w, weight matrix P and solution d
w=L0-L;
P=diag([1/0.008^2 1/0.005^2 1/0.004^2]);
u=A'*P*w;
N=A'*P*A;
d=-inv(N)*u
d =
  -0.3243
  -0.4509

% determine adjusted parameters
hAa=hA+d(1)
hAa =
   92.3347
hBa=hB+d(2)
hBa =
   94.7031

% determine residuals v and adjusted observations La
v=A*d+w;
La=L0+v
La =
  -7.6653
   2.3683
   4.4939
```

## 6.2.2 Ajuste de Estación

El ajuste de estación es un procedimiento que consiste en encontrar los valores ajustados de los ángulos medidos en una estación de levantamiento para que sean consistentes con la geometría asociada. El siguiente ejemplo se da para ilustrar la forma más simple de ajuste de estación, que se puede extender a otras formas de ajustes de estación como se ilustra en los siguientes ejemplos. Por ejemplo, un ángulo se midió repetidamente durante dos noches. Las medias aritméticas de las mediciones realizadas en las dos noches ($ \bar{\ell}_1 $, $ \bar{\ell}_2 $) con los segundos dados únicamente y el número de repeticiones ($ n_1 $, $ n_2 $) de mediciones para cada noche son los siguientes:

Noche 1: $ \bar{\ell}_1 = 40'' $, $ n_1 = 3 $

Noche 2: $ \bar{\ell}_2 = 45'' $, $ n_2 = 4 $

Determine el ángulo ajustado por mínimos cuadrados para las dos noches utilizando el método de ajuste de mínimos cuadrados paramétricos con los números de repeticiones considerados como pesos.

Solución:**

Considere la Figura 6.3 para la ilustración del problema en este ejemplo. Las medias de las mediciones del mismo ángulo ($ \theta $) se dan como $ \bar{\ell}_1 $ y $ \bar{\ell}_2 $; por lo tanto, ahora es necesario identificar qué se refiere a los parámetros, ya que el problema no se formulará como ecuaciones de modelo condicionales (ecuaciones que involucran solo observaciones). Como se utilizan dos direcciones en la obtención y ajuste de ángulos, sea la primera dirección involucrada para este ángulo $ d_1 = 0° $ (la dirección trasera) y $ d_2 $ (la dirección delantera). Como la lectura de la dirección trasera generalmente se establece en cero por conveniencia, se establece $ d_1 = 0° $ y $ d_2 $ como la dirección desconocida (parámetro) a determinar mediante el método de ajuste de mínimos cuadrados paramétricos.

Tenga en cuenta que este enfoque se puede extender para involucrar muchos ángulos medidos alrededor de un punto, como se muestra en los siguientes ejemplos.

Sea el parámetro $ x = [d_2] $:

Observaciones: $ \ell = \begin{bmatrix} \bar{\ell}_1 \\ \bar{\ell}_2 \end{bmatrix} $

Formule las ecuaciones paramétricas ($ \ell = f(x) $) como sigue. Dado que hay dos observaciones independientes involucradas, habrá dos ecuaciones formuladas como sigue (una ecuación para cada observación, con los sombreros ignorados por brevedad):

$$ \bar{\ell}_1 = d_2 - 0.0 \quad \text{(6.4)} $$

$$ \bar{\ell}_2 = d_2 - 0.0 \quad \text{(6.5)} $$

La primera matriz de diseño ($ A $) se puede formar encontrando las derivadas parciales de las Ecuaciones (6.4) y (6.5) con respecto al parámetro desconocido $ d_2 $ como sigue:

$$ A = \frac{\partial \ell}{\partial x} = \begin{bmatrix} 1 \\ 1 \end{bmatrix} $$

Sea $ x^0 $ (valor aproximado del parámetro) = 40''; sustituya este valor en las Ecuaciones (6.4) y (6.5) para obtener:

$$ f(x^0) = \begin{bmatrix} 40 - 0 \\ 40 - 0 \end{bmatrix} \rightarrow \begin{bmatrix} 40 \\ 40 \end{bmatrix} $$

Forme el vector de cierre $ w $:

$$ w = f(x^0) - \ell = \begin{bmatrix} 40 - 40 \\ 40 - 45 \end{bmatrix} \rightarrow w = \begin{bmatrix} 0 \\ -5 \end{bmatrix} $$

Dado que la pregunta requiere que el número de repeticiones se tome como correspondiente al peso de cada medición promedio, la matriz de pesos ($ P $) de las observaciones se puede dar como:

$$ P = \begin{bmatrix} 3 & 0 \\ 0 & 4 \end{bmatrix} $$

Otros productos son los siguientes:

$$ N = A^T P A \rightarrow 7 $$

$$ u = A^T P w \rightarrow -20 $$

Vector de correcciones al parámetro aproximado:

$$ \delta = - N^{-1} u \rightarrow \left( \frac{1}{7} \right) \times 20 = +2.86'' $$

Dirección ajustada $ \hat{d}_2 $:

$$ \hat{d}_2 = x^0 + \delta $$

$$ \hat{d}_2 = 40 + 2.86 \rightarrow 42.86'' $$

El vector residual $ v $: $ v = A\delta + w $

$$ v = \begin{bmatrix} 1 \\ 1 \end{bmatrix} \begin{bmatrix} 2.86 \end{bmatrix} + \begin{bmatrix} 0 \\ -5 \end{bmatrix} \rightarrow v = \begin{bmatrix} 2.86 \\ -2.14 \end{bmatrix} $$

Ángulo ajustado:

$$ \hat{\ell} = \ell + v $$

$$ \hat{\ell} = \begin{bmatrix} 40 \\ 45 \end{bmatrix} + \begin{bmatrix} 2.86 \\ -2.14 \end{bmatrix} \rightarrow \begin{bmatrix} 42.86 \\ 42.86 \end{bmatrix} $$

Ángulo ajustado $ \theta = 42.86'' $.

### Ejemplo 6.2

Tres ángulos horizontales se miden con un instrumento de estación total alrededor del horizonte en el punto A como se muestra en la Figura 6.4. Las mediciones y sus desviaciones estándar se dan en la Tabla 6.4. Ajuste las mediciones utilizando el ajuste de mínimos cuadrados paramétricos.

|--------|------------------|---------------------|
| $ \alpha $ | 42°12'13''      | 3''                 |

**Solución:**

Si la dirección de la estación total A–B se toma como valor fijo de 0°00'00'', las otras direcciones (A–C y A–D) se consideran los parámetros desconocidos ($ x $), y los ángulos horizontales medidos son las observaciones ($ \ell $). Sea $ d_1 $ el parámetro desconocido para A–C y $ d_2 $ el parámetro desconocido para A–D. Las ecuaciones paramétricas ($ \hat{\ell} = f(\hat{x}) $) para el problema se pueden dar como sigue:

$$ \hat{\alpha} = d_1 - 0 \quad (6.6) $$

$$ \hat{\beta} = d_2 - \hat{d}_1 \quad (6.7) $$

**6.3 Propiedades Estocásticas del Ajuste Paramétrico**

El resumen de los modelos estocásticos para el ajuste paramétrico por mínimos cuadrados discutido en el Capítulo 5 se presenta en esta sección como sigue. Los modelos se utilizan en los ejemplos que siguen.

1) Al inicio de un ajuste paramétrico por mínimos cuadrados, las desviaciones estándar dadas de las mediciones se utilizan para formar la matriz de varianza-covarianza ($C_{\ell}$) del vector de observación $\ell$. La matriz de varianza-covarianza formada y su matriz de cofactores ($Q_{\ell}$) están relacionadas por la Ecuación (6.39) como:

$$
C_{\ell} = \sigma_0^2 Q_{\ell}
$$

donde $\sigma_0^2 = 1$ es el factor de varianza a priori de unidad de peso. En esencia, esto significa que la matriz de varianza-covarianza de las observaciones y su matriz de cofactores son las mismas al inicio de cada ajuste.

2) Durante o después del ajuste por mínimos cuadrados, las matrices de cofactores de los parámetros ajustados ($Q_{\hat{x}}$), las observaciones ajustadas ($Q_{\hat{\ell}}$), y los residuos de observación ($Q_v$) se derivan. Estas cantidades derivadas se resumen mediante las Ecuaciones (6.40), (6.41) y (6.42), respectivamente:

$$
Q_{\hat{x}} = (A^T P A)^{-1}
$$

$$
Q_{\hat{\ell}} = A(A^T P A)^{-1}A^T
$$

$$
Q_v = Q_{\ell} - A(A^T P A)^{-1}A^T
$$

donde $A$ es la primera matriz de diseño y $P$ es la matriz de pesos de las observaciones dada como $P = Q_{\ell}^{-1}$.

3) Las siguientes matrices de varianza-covarianza de las cantidades ajustadas (Ecuaciones (6.43)–(6.45)) se pueden obtener a partir de las Ecuaciones (6.40), (6.41) y (6.42), respectivamente:

$$
C_{\hat{x}} = \sigma_0^2 (A^T P A)^{-1} \quad \text{(6.43)}
$$

$$
C_{\hat{\ell}} = \sigma_0^2 A(A^T P A)^{-1}A^T \quad \text{(6.44)}
$$

$$
C_v = \sigma_0^2 Q_{\ell} - \sigma_0^2 A(A^T P A)^{-1}A^T \quad \text{(6.45)}
$$

donde $C_{\hat{x}}$ es la matriz de covarianza de los parámetros ajustados, $C_{\hat{\ell}}$ es la matriz de covarianza de las observaciones ajustadas, $C_v$ es la matriz de covarianza de los residuos de observación, y $\sigma_0^2$ es el factor de varianza a posteriori (APVF) de unidad de peso, que se calcula después del ajuste por mínimos cuadrados como:

$$
\sigma_0^2 = \frac{v^T P v}{n - u} \quad \text{(6.46)}
$$

con la redundancia expresada como $n - u$, $n$ el número de ecuaciones paramétricas, $u$ el número de parámetros independientes, $v$ el vector de residuos de observación, y $P$ la matriz de pesos de las observaciones.

4) La matriz de covarianza ($C_v$) de los residuos de observación se puede reescribir a partir de la Ecuación (6.45) como:

$$
C_v = \hat{C}_{\ell} - C_{\hat{\ell}} \quad \text{(6.47)}
$$

donde $\hat{C}_{\ell}$ es la matriz de covarianza escalada de las observaciones originales $\ell$, expresada como:

$$
\hat{C}_{\ell} = \sigma_0^2 Q_{\ell} \quad \text{(6.48)}
$$

con todos los símbolos definidos en el paso (3).

La matriz de covarianza escalada $\hat{C}_{\ell}$ no debe confundirse con $C_{\ell} = \sigma_0^2 Q_{\ell}$ (con $\sigma_0^2 = 1$), que es la matriz de covarianza no escalada de las observaciones originales, utilizada al inicio del ajuste por mínimos cuadrados. En realidad, $\hat{C}_{\ell} = s_0^2 Q_{\ell}$ y $C_{\ell} = \sigma_0^2 Q_{\ell}$ serán idénticas cuando $\sigma_0^2 = s_0^2$.

## **6.7 Pasos para el Ajuste con Restricción de Peso**

Los pasos para el método de ajuste paramétrico de mínimos cuadrados con restricción de peso discutido en el Capítulo 5 se resumen en esta sección de la siguiente manera:

1) Proporcione el modelo paramétrico ($\hat{\ell} = f(\hat{x})$) con $P$ como la matriz de pesos de las observaciones y $C_{x}$ como la matriz de covarianza a priori del vector de todos los parámetros o de un sub-vector de los parámetros.
2) Forme la primera matriz de diseño ($A$), que es la matriz que contiene las derivadas parciales de las ecuaciones paramétricas con respecto a los parámetros desconocidos. El tamaño de esta matriz será igual al número de ecuaciones (o número de observaciones) formando el tamaño de la fila por el tamaño de la columna que representa el número de parámetros, incluidos aquellos parámetros asociados con la matriz de covarianza dada $C_{x}$.
3) Elija valores aproximados para los parámetros ($x_0$); los valores de aquellos parámetros con matriz de covarianza dada pueden usarse como valores aproximados para el ajuste. Generalmente, si los valores elegidos no están lo suficientemente cerca de lo que deberían ser los valores finales, los valores finales calculados pueden usarse nuevamente como valores aproximados en otra ronda de ajuste de manera iterativa hasta que los valores finales no sean demasiado diferentes de los valores aproximados (o iniciales). En este caso, el ajuste de mínimos cuadrados puede tener que repetirse tantas veces como sea posible hasta que la solución final y la anterior converjan a un límite de tolerancia específico. Sustituya estos valores aproximados en las ecuaciones paramétricas formadas en el paso 1 para obtener los valores derivados o simulados de las observaciones ($\ell_0 = f(x_0)$).
4) Encuentre el vector de cierre ($w = f(x_0) − \ell$) como la diferencia entre las observaciones derivadas ($\ell_0 = f(x_0)$) calculadas en el paso 3 y las observaciones reales ($\mathbf{\ell}$).
5) Forme la matriz de cofactores ($Q$) de las observaciones (asumiendo que el factor de varianza a priori $\sigma_0^2$ es 1) utilizando las varianzas de las observaciones a lo largo de la diagonal de la matriz (en el orden en que se forman las ecuaciones paramétricas en el paso 1). La inversa de esta matriz de cofactores es la matriz de pesos ($P$) de las observaciones.
6) Tome la inversa ($C_{x}^{-1}$) de la matriz de covarianza de los parámetros como el peso de los parámetros, $P_{x} = C_{x}^{-1}$; alinéelos a lo largo de la diagonal de la matriz (en el orden en que se forman las ecuaciones paramétricas en el paso 1); y asigne pesos cero a aquellos parámetros con matriz de covarianza desconocida.
7) Determine la solución de restricción de peso de las correcciones a los parámetros aproximados de la siguiente manera:
$$\delta = −(P_{x} + A^T P A)^{-1} A^T P w \quad \text{(6.57)}$$
8) Determine los parámetros ajustados: $\hat{x} = x_0 + \delta$.
9) Si es necesario, determine los residuos ($v$) de las observaciones y las observaciones ajustadas ($\hat{\ell}$): $v = A\delta + w$ y $\hat{\ell} = \ell + v$.
10) Determine la matriz de cofactores ($Q_{\hat{x}}$) y la matriz de covarianza ($C_{\hat{x}}$) de los parámetros ajustados, respectivamente, de la siguiente manera:
$$Q_{\hat{x}} = (P_{x} + A^T P A)^{-1} \quad \text{(6.58)}$$
$$C_{\hat{x}} = s_0^2 Q_{\hat{x}} \quad \text{(6.59)}$$
donde el APVF de peso unitario $s_0^2$ se puede dar de la siguiente manera:
$$s_0^2 = \frac{v^T P v + \delta^T P_{x} \delta}{n_d − u_d} \quad \text{(6.60)}$$
con $n_d$ como el número de observaciones directas (o reales) y $u_d$ como el número de parámetros directos (cuyos valores no tienen desviaciones estándar asociadas).
11) Determine la matriz de cofactores ($Q_{\hat{\ell}}$) y la matriz de covarianza ($C_{\hat{\ell}}$) de las observaciones ajustadas, respectivamente, de la siguiente manera:
$$Q_{\hat{\ell}} = A (P_{x} + A^T P A)^{-1} A^T \quad \text{(6.61)}$$
$$C_{\hat{\ell}} = s_0^2 Q_{\hat{\ell}} \quad \text{(6.62)}$$
donde el factor de varianza $s_0^2$ se determina a partir de la Ecuación (6.60).
