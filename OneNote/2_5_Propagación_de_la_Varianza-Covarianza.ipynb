### 2.5 Propagación de la Varianza-Covarianza

La ecuación que muestra la relación entre las varianzas de las mediciones y las varianzas y covarianzas de las cantidades a resolver se conoce como **propagación de la varianza-covarianza** o **propagación de errores aleatorios**. En la propagación de la varianza-covarianza, los valores específicos de los errores son desconocidos. Sobre esta base, los errores se expresan, por ejemplo, como ±5 unidades, de modo que es imposible corregir este tipo de error o aplicar un enfoque de propagación de errores sistemáticos. En lugar de trabajar con los valores reales de error, se utilizan los valores probables de los errores en la propagación de varianzas y covarianzas. Para ilustrar el procedimiento de propagación de errores aleatorios, refiérase a la Ecuación (2.1) para un caso en el que dos parámetros $x = \left( x_1 \, x_2 \right)^T$ y dos observables $\ell = \left( \ell_1 \, \ell_2 \right)^T$ están involucrados, por simplicidad; el siguiente modelo matemático general se puede formular:

$$ x = f(\ell) \tag{2.40} $$

Por ejemplo, si la Ecuación (2.40) se descompone en dos funciones $f_1$ y $f_2$ que operan en ambas observables $\ell_1$ y $\ell_2$, se pueden obtener las siguientes relaciones:

$$ x_1 = f_1(\ell_1, \ell_2) \tag{2.41} $$

$$ x_2 = f_2(\ell_1, \ell_2) \tag{2.42} $$

Y las varianzas $\sigma_{x_1}^2$ y $\sigma_{x_2}^2$ de $x_1$ y $x_2$ respectivamente, se pueden dar como:

$$ \sigma_{x_1}^2 = \left( \frac{\partial f_1}{\partial \ell_1} \right)^2 \sigma_{\ell_1}^2 + \left( \frac{\partial f_1}{\partial \ell_2} \right)^2 \sigma_{\ell_2}^2 + 2 \left( \frac{\partial f_1}{\partial \ell_1} \right) \left( \frac{\partial f_1}{\partial \ell_2} \right) \sigma_{\ell_1 \ell_2} \tag{2.43} $$

$$ \sigma_{x_2}^2 = \left( \frac{\partial f_2}{\partial \ell_1} \right)^2 \sigma_{\ell_1}^2 + \left( \frac{\partial f_2}{\partial \ell_2} \right)^2 \sigma_{\ell_2}^2 + 2 \left( \frac{\partial f_2}{\partial \ell_1} \right) \left( \frac{\partial f_2}{\partial \ell_2} \right) \sigma_{\ell_1 \ell_2} \tag{2.44} $$

Si $\ell_1$ y $\ell_2$ se miden independientemente, entonces la covarianza $\sigma_{\ell_1 \ell_2}$ entre ellas será cero. La covarianza $\sigma_{x_1 x_2}$ entre las dos cantidades $x_1$ y $x_2$ se puede dar como:

$$ \sigma_{x_1 x_2} = \left( \frac{\partial f_1}{\partial \ell_1} \right) \left( \frac{\partial f_2}{\partial \ell_1} \right) \sigma_{\ell_1}^2 + \left( \frac{\partial f_1}{\partial \ell_2} \right) \left( \frac{\partial f_2}{\partial \ell_2} \right) \sigma_{\ell_2}^2 $$

$$ + \left( \frac{\partial f_1}{\partial \ell_1} \right) \left( \frac{\partial f_2}{\partial \ell_2} \right) \sigma_{\ell_1 \ell_2} + \left( \frac{\partial f_1}{\partial \ell_2} \right) \left( \frac{\partial f_2}{\partial \ell_1} \right) \sigma_{\ell_1 \ell_2} \tag{2.45} $$

Para simplificar las Ecuaciones (2.43)–(2.45), las notaciones matriciales se emplean comúnmente utilizando el enfoque de Jacobiana; en este caso, las Ecuaciones (2.43)–(2.45) se pueden dar como:

$$ C_x = JCJ^T \tag{2.46} $$

donde

$$
C_x = \begin{bmatrix}
\sigma^2_{x_1} & \sigma_{x_1 x_2} \\
\sigma_{x_1 x_2} & \sigma^2_{x_2}
\end{bmatrix} \tag{2.47}
$$

$$ J = \begin{bmatrix}
\frac{\partial f_1}{\partial \ell_1} & \frac{\partial f_1}{\partial \ell_2} \\
\frac{\partial f_2}{\partial \ell_1} & \frac{\partial f_2}{\partial \ell_2}
\end{bmatrix} \tag{2.48} $$

$$ C = \begin{bmatrix}
\sigma_{\ell_1}^2 & \sigma_{\ell_1 \ell_2} \\
\sigma_{\ell_1 \ell_2} & \sigma_{\ell_2}^2
\end{bmatrix} \tag{2.49} $$

$ C_x $ es la matriz de covarianza de $ x_1 $ y $ x_2 $; $ C $ es la matriz de covarianza de $ \ell_1 $ y $ \ell_2 $; y $ J $ es la matriz Jacobiana de las Ecuaciones (2.41)–(2.42). De manera similar, para un vector de parámetros $ x = (x_1, x_2, x_3, ..., x_m)^T $ expresado explícitamente como una función de las mediciones de campo $ \ell = (\ell_1, \ell_2, ..., \ell_n)^T $, se puede derivar la siguiente matriz Jacobiana:

$$ J = \begin{bmatrix}
\frac{\partial x_1}{\partial \ell_1} & \frac{\partial x_1}{\partial \ell_2} & \cdots & \frac{\partial x_1}{\partial \ell_n} \\
\frac{\partial x_2}{\partial \ell_1} & \frac{\partial x_2}{\partial \ell_2} & \cdots & \frac{\partial x_2}{\partial \ell_n} \\
\vdots & \vdots & \ddots & \vdots \\
\frac{\partial x_m}{\partial \ell_1} & \frac{\partial x_m}{\partial \ell_2} & \cdots & \frac{\partial x_m}{\partial \ell_n}
\end{bmatrix} \tag{2.50} $$

Aplicando las leyes de propagación de la varianza-covarianza a la Ecuación (2.40) se obtiene:

$$
\Sigma_x = \Sigma_\ell J^T \tag{2.51}
$$

donde $\Sigma_x$ es la verdadera matriz de varianza-covarianza de los parámetros $x$ y $\Sigma_\ell$ es la matriz de varianza-covarianza de las observaciones $\ell$. Dado que la verdadera matriz de varianza-covarianza de las observaciones no es conocida, $\Sigma_\ell$ no puede ser calculada. En cambio, se calcula la estimación de la matriz de varianza-covarianza de los parámetros ($C_x$):

$$
C_x = JC_eJ^T \tag{2.52}
$$

donde $C_e$ es la matriz estimada de varianza-covarianza de las observaciones. Las Ecuaciones (2.43)–(2.45) y (2.52) constituyen lo que se conoce como las leyes de propagación de la varianza-covarianza. Los siguientes ejemplos ilustran cómo se aplican las leyes de propagación de la varianza-covarianza.

**Ejemplo 2.14**

Dadas $ n $ mediciones independientes de una única cantidad como $\ell_1, \ell_2, ..., \ell_n$ y sus correspondientes desviaciones estándar $\sigma_1, \sigma_2, ..., \sigma_n$, derive la expresión para la desviación estándar del valor medio de las mediciones utilizando la ley de propagación de varianzas.

**Solución:**

La ecuación para encontrar la media puede expresarse como

$$
\bar{x} = \frac{1}{n} (\ell_1 + \ell_2 + \cdots + \ell_n)
$$

Basado en las leyes de propagación de la matriz de varianza-covarianza, la matriz de covarianza de la media puede ser dada como

$$
C_{\bar{x}} = J C_\ell J^T
$$

donde la matriz Jacobiana ($J$) puede ser dada como

$$
J = \begin{pmatrix}
\frac{\partial \bar{x}}{\partial \ell_1} & \frac{\partial \bar{x}}{\partial \ell_2} & \cdots & \frac{\partial \bar{x}}{\partial \ell_n}
\end{pmatrix}
\rightarrow J = \begin{pmatrix}
\frac{1}{n} & \frac{1}{n} & \cdots & \frac{1}{n}
\end{pmatrix}
$$

y la matriz de covarianza de las observaciones es una matriz diagonal, la cual puede ser dada como

$$
C_\ell = \begin{pmatrix}
\sigma^2_1 & \cdots & 0 \\
\cdots & \ddots & \cdots \\
0 & \cdots & \sigma^2_n
\end{pmatrix}
$$

La matriz de covarianza de la media puede ser dada como

$$
C_{\bar{x}} = \sigma^2_{\bar{x}} J \begin{pmatrix}
\sigma^2_1 & \cdots & 0 \\
\cdots & \ddots & \cdots \\
0 & \cdots & \sigma^2_n
\end{pmatrix} J^T \quad \text{o} \quad \sigma^2_{\bar{x}} = \left( \frac{1}{n} \right)^2 (\sigma^2_1 + \sigma^2_2 + \cdots + \sigma^2_n)
$$

Asumiendo que cada observación tiene la misma varianza (i.e. $\sigma^2_1 = \sigma^2_2 = \cdots = \sigma^2_n = \sigma^2_\ell$), la varianza de la media puede ser dada como sigue:

$$
\sigma^2_{\bar{x}} = \left( \frac{1}{n} \right)^2 (n \sigma^2_\ell) \quad \text{o} \quad \sigma^2_{\bar{x}} = \frac{1}{n} \sigma^2_\ell
$$

La desviación estándar de la media se da entonces como

$$
\sigma_{\bar{x}} = \frac{\sigma_\ell}{\sqrt{n}}
$$

**2.6 Propagación de Errores Basada en las Especificaciones del Equipo**

El principal equipo moderno de topografía consiste en estación total e instrumentos de nivelación. Los instrumentos de estación total están diseñados para medir distancias y direcciones (o ángulos), mientras que los instrumentos de nivelación están diseñados para medir diferencias de altura. La desviación estándar para una medición de distancia (S) usando un equipo de estación total generalmente se especifica de acuerdo con la Organización Internacional de Normalización (ISO 17123-4) en forma de $a + b$ ppm, y la desviación estándar de una medición de dirección (ángulo) (horizontal y vertical) se especifica de acuerdo con la ISO 17123-3 como $c \cdot s$. La desviación estándar de mediciones de altura generalmente se especifica de acuerdo con la ISO 17123-2 como $td \, \text{mm km}^{-1} \, \text{double run}$ o simplemente $td \, \text{mm}$.

## 2.6.1 Propagación para Distancia Basada en la Especificación de Precisión

Las especificaciones de precisión para equipos de estación total generalmente se indican de la siguiente manera por los fabricantes:

$$
SE_s = \pm (a + bS) \tag{2.58}
$$

$$
SE_s = a + b \tag{2.59}
$$

$$
SE_s = \pm \sqrt{a^2 + (b \times S)^2} \tag{2.60}
$$

donde $SE_s$ es el error estándar (o desviación estándar) de la medición de distancia (S), "a" es la parte constante en mm, y "b" es el factor de escala en ppm. Las especificaciones generalmente se interpretan de las siguientes dos maneras:

a) Determine el error de escala en mm, y añádalo directamente al error constante para obtener el error total. Este enfoque de interpretar la precisión generalmente está respaldado por los fabricantes de instrumentos. Esto será válido si uno quiere ser pesimista (asumiendo errores mayores de lo esperado) y estar en el lado seguro del presupuesto de error. En este caso, el error constante y el error debido a la escala se consideran aditivos como sigue:

$$
SE_s = \pm (a + b \times 10^{-6} \times S) \tag{2.61}
$$

b) Determine el error de escala en mm, elévelo al cuadrado, añádalo al cuadrado del error constante, y encuentre la raíz cuadrada de la suma. Este enfoque parece ser respaldado por la teoría de propagación de errores por la estación total si "a" y "b" se consideran como dos errores aleatorios independientes. Los errores aleatorios se propagan como sigue:

$$
SE_s = \pm \sqrt{a^2 + (b \times 10^{-6} \times S)^2} \tag{2.62}
$$

Por ejemplo, considere una estación total que tiene las especificaciones de precisión para la medición de distancia de acuerdo con ISO17123-4 como 2 mm + 2 ppm (para modo de campo fino) y 5 mm + 2 ppm (para modo rápido IR). Para el modo rápido IR, para una distancia medida de 1000 m, el error estándar para la distancia será:

$$
SE_s = \pm (5 + 2 \times 10^{-6} \times 1000 \, 000) \rightarrow \pm 7 \, \text{mm}
$$

Usando el ejemplo en la Ecuación (2.62) se obtiene lo siguiente:

$$
SE_s = \pm \sqrt{5^2 + (2 \times 10^{-6} \times 1000 \, 000)^2} \rightarrow \pm 5.4 \, \text{mm}
$$

Se puede ver a partir de los ejemplos anteriores que la Ecuación (2.61) da una estimación pesimista; esto significa que usar este enfoque siempre dará una estimación pesimista que es siempre mayor que el valor de la Ecuación (2.62). Sin embargo, la solución precisa se da mediante la Ecuación (2.62), y se sugiere su uso en lugar de usar la pesimista. Cabe señalar también que la ISO (17123-4) generalmente no descompone el error de distancia en componentes constantes y dependientes de la distancia; más bien, se especifica el error promedio de medir una distancia, por ejemplo, el error promedio de medir una distancia se puede expresar simplemente como 5 mm. Para fines educativos, se aconseja a los estudiantes que usen la Ecuación (2.62) para determinar las desviaciones estándar de las mediciones de distancia. Los otros errores aleatorios que pueden considerarse al propagar el error para una medición de distancia son los errores de centrado del instrumento y el objetivo sobre las marcas de topografía. Por ejemplo, si el instrumento y los errores de centrado del objetivo son $\sigma_i$ y $\sigma_t$ (en mm), respectivamente, la desviación estándar total (SE_total) de medir una distancia puede ser dada como:

$$
SE_{\text{total}} = \pm \sqrt{SE_s^2 + \sigma_i^2 + \sigma_t^2} \tag{2.63}
$$

## 2.6.2 Propagación para Dirección (Ángulo) Basada en la Especificación de Precisión

Si un instrumento de estación total tiene una desviación estándar para la medición de ángulo horizontal (o vertical) de $2''$ de acuerdo con ISO 17123-3, la desviación estándar para el promedio de un conjunto de mediciones de dirección con la desviación estándar de una dirección medida en una sola cara es $SE_d = 2\sqrt{2''}$ o $2.8''$. Este error generalmente se considera como el error combinado de apuntar y leer. Los otros errores que se pueden considerar con respecto a la medición de dirección (o ángulo) se deben al centrado y nivelación del instrumento y los objetivos correspondientes. Por ejemplo, si los errores de centrado del instrumento y del objetivo son $\sigma_i$ y $\sigma_t$ (en arcosegundos), respectivamente, la desviación estándar total (SE_total) de medir una dirección de una línea en una sola cara del telescopio del instrumento puede darse como

$$
SE_{\text{Total}} = \pm \sqrt{SE_d^2 + \sigma_i^2 + \sigma_t^2} \tag{2.64}
$$

Para la medición de ángulo, se involucrarán dos objetivos de modo que la Ecuación (2.64) tendrá dos $\sigma_t$ en lugar de solo $\sigma_t^2$ si los errores de los dos objetivos son los mismos o el cuadrado del error del segundo objetivo se agrega si es diferente del primer objetivo.

## 2.6.3 Propagación para la Diferencia de Altura Basada en la Especificación de Precisión

Si la desviación estándar de la diferencia de elevación sobre 1 km basada en el promedio de nivelación de doble recorrido es ±0.5 mm (según ISO 17123-3), la desviación estándar de la diferencia de elevación sobre 1 km basada en nivelación de un solo recorrido será ±0.5 mm/$\sqrt{2}$. La desviación estándar sobre 1 km (doble recorrido) puede determinarse sobre L km como ±0.5 mm $\sqrt{L}$ con L como la distancia de un solo recorrido expresada en km; la desviación estándar sobre 1 km (un solo recorrido) puede determinarse sobre L km como ±0.5 mm $\sqrt{2L}$ con L como la distancia de un solo recorrido expresada en km.

## 2.7 Regla Heurística para la Propagación de la Covarianza

La regla heurística es una regla que fomenta el uso de la observación personal de la solución de un problema para resolver otro problema. El método se refiere a la multiplicación simbólica en Mikhail (1976), quien acreditó a Tienstra (1966) como el inventor de la regla. Por ejemplo, dadas las siguientes dos ecuaciones:

$$
x = \ell_1 + \ell_2 \tag{2.65}
$$

$$
y = \ell_1 + \ell_2 \tag{2.66}
$$

donde $\ell_1$ y $\ell_2$ son variables aleatorias, la covarianza $s_{xy}$ de las dos variables $x$ y $y$ se puede obtener basándose en el patrón mostrado por el producto de las dos variables $x$ y $y$ como sigue:

$$
\text{Producto ordinario} : xy = (\ell_1 + \ell_2)(\ell_1 + \ell_2) \tag{2.67}
$$

o

$$
\text{Producto ordinario} : xy = \ell_1^2 + \ell_2^2 + 2\ell_1\ell_2 \tag{2.68}
$$

Usando la regla heurística en la Ecuación (2.68), la covarianza ($s_{xy}$) de $x$ y $y$ se puede dar como sigue:

$$
\text{Producto heurístico} : s_{xy} = s_{\ell_1}^2 + s_{\ell_2}^2 + 2s_{\ell_1\ell_2} \tag{2.69}
$$

donde $s_{\ell_1}^2$ y $s_{\ell_2}^2$ son las varianzas de $\ell_1$ y $\ell_2$, respectivamente, y $s_{\ell_1\ell_2}$ es la covarianza entre $\ell_1$ y $\ell_2$. Sin embargo, si $\ell_1$, $\ell_2$, $x$ e $y$ son vectores aleatorios, la misma regla se puede aplicar con las varianzas convirtiéndose en matrices de covarianza. En este caso, la Ecuación (2.69) se escribirá, dando la matriz de covarianza ($C_{xy}$) de $x$ y $y$ como sigue:

$$
\text{Producto heurístico} : C_{xy} = C_{\ell_1} + C_{\ell_2} + 2C_{\ell_1\ell_2} \tag{2.70}
$$

donde $C_{\ell_1}$ y $C_{\ell_2}$ son las matrices de covarianza de $\ell_1$ y $\ell_2$, respectivamente, y $C_{\ell_1\ell_2}$ es la matriz de covarianza de $\ell_1$ y $\ell_2$, ya que la matriz de covarianza ya indica covarianzas y varianzas. De manera similar, dada la siguiente función en la cual $A$ es una constante y $x$, $y$ y $\ell$ son variables aleatorias, dando la matriz de covarianza ($C_{xy}$) de $x$ y $y$ como sigue:

$$
X = A\ell \tag{2.71}
$$

$$
Y = \ell \tag{2.72}
$$

El producto $XY$ se puede dar como sigue:

$$
\text{Producto ordinario} : XY = A\ell^2 \tag{2.73}
$$

Aplicando la regla heurística a la Ecuación (2.73), la covarianza de $X$ y $Y$ se puede dar como sigue:

$$
\text{Producto heurístico} : s_{xy} = As_{\ell}^2 \tag{2.74}
$$

donde $s_{\ell}^2$ es la varianza de $\ell$. La misma regla se puede aplicar para un caso en el que $A$ es una matriz cuyos elementos son valores constantes, y $X$, $Y$ y $\ell$ son vectores de variables aleatorias, dando la matriz de covarianza ($C_{xy}$) de $X$ y $Y$ como sigue:

$$
\text{Producto heurístico} : C_{xy} = AC_\ell \tag{2.75}
$$

donde $C_\ell$ es la matriz de covarianza de $\ell$.

**Ejemplo 2.17**

Dada la matriz de varianza-covarianza del vector de medición $\lambda = \begin{bmatrix}

y dos funciones de λ, $ x = \lambda_1 + \lambda_2 $ y $ y = 3\lambda_1 $, determine lo siguiente:

i) La covarianza entre $ x $ e $ y $ (o $ C_{xy} $) utilizando enfoques de propagación de varianza-covarianza y regla heurística.

### Solución (i):

Covarianza entre $ x $ e $ y $ (o $ C_{xy} $) utilizando el enfoque de propagación de varianza-covarianza.

Sea $ x = \begin{bmatrix} x \\ y \end{bmatrix} $ y $ \lambda = \begin{bmatrix} \lambda_1 \\ \lambda_2 \end{bmatrix} $.

Determine la Jacobiana ($ J $) de las siguientes ecuaciones:

$$
x = \lambda_1 + \lambda_2 \tag{2.76}
$$

$$
y = 3\lambda_1 \tag{2.77}
$$

$$
J = \begin{bmatrix} 1 & 1 \\ 3 & 0 \end{bmatrix}
$$

De las leyes de propagación de varianza-covarianza,

$$
C = JC_\lambda J^T
$$

$$
C = \begin{bmatrix} 1 & 1 \\ 3 & 0 \end{bmatrix}
\begin{bmatrix} \frac{2}{3} & \frac{1}{3} \\ \frac{1}{3} & \frac{2}{3} \end{bmatrix}
\begin{bmatrix} 1 & 3 \\ 1 & 0 \end{bmatrix}
\rightarrow C = \begin{bmatrix} 2 & 3 \\ 3 & 6 \end{bmatrix}

donde la covarianza entre $ x $ e $ y $ de la matriz $ C $ anterior es 3. La covarianza entre $ x $ e $ y $ también se puede encontrar usando la regla heurística en las Ecuaciones (2.76) y (2.77) como sigue. Dé el producto ordinario de las dos ecuaciones como

\text{Producto ordinario} : xy = 3\lambda_1^2 + 3\lambda_1\lambda_2 \tag{2.78}
$$

lo cual se traduce en la covarianza ($ s_{xy} $) entre $ x $ e $ y $ en la siguiente regla heurística:

$$
\text{Producto heurístico} : s_{xy} = 3s_{\lambda_1}^2 + 3s_{\lambda_1\lambda_2} \tag{2.79}
$$

De la matriz de covarianza dada $ C_\lambda $, $ s_{\lambda_1}^2 = \frac{2}{3} $ y $ s_{\lambda_1\lambda_2} = \frac{1}{3} $ sustituyendo en la Ecuación (2.79) da $ s_{xy} = 3 $ (el mismo resultado obtenido previamente).

ii) La covarianza entre $ x $ y $ \lambda $ (o $ C_{x\lambda} $) utilizando el enfoque de la regla heurística.
