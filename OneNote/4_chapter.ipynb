### 4.2 Métodos Tradicionales de Ajuste

Los levantamientos de propiedades para localizar o establecer límites de propiedades se basan tradicionalmente en levantamientos de travesía. Son comunes tres clases de levantamientos de travesía: travesía abierta, en la cual se conoce la posición del punto de inicio pero se desconoce la del punto final; travesía cerrada, en la cual se conocen las posiciones de los puntos de inicio y final; y travesía cerrada en bucle, que es una forma de travesía cerrada en la cual los puntos de inicio y final son los mismos. Generalmente, los levantamientos de travesía requieren ajustes; de lo contrario, los errores en la travesía se acumularán y se asignarán a la última etapa de la travesía de la red, lo que será injustificable.

Hay tres métodos tradicionales de ajuste utilizados en geomática: regla de tránsito, regla de brújula o regla de Bowditch, y regla de Crandall. Estos métodos se utilizan comúnmente con respecto a las travesías cerradas que representan parcelas de tierra en planes de subdivisión. Cada uno de estos métodos requiere que el componente de mediciones angulares de la travesía se ajuste geométricamente primero antes de ajustar toda la travesía. Sin embargo, estos métodos no lograrán los resultados de ajuste deseados cuando se ajusta una red de travesías cerradas con estaciones comunes. Además, los métodos generalmente ignoran la calidad de las observaciones en sus procedimientos de ajuste. Esto significa que el topógrafo que compra un instrumento de levantamiento preciso tendrá que ignorar la precisión lograda con dicho instrumento al ajustar las mediciones adquiridas con el instrumento, lo cual será irrazonable.

En los métodos tradicionales de ajuste, el término **latitud** se usa para referirse al cambio en el valor de norte ($\Delta N$), que puede expresarse como

$$
\Delta N = S \times \cos \beta \quad \text{(4.1)}
$$

donde $S$ es la longitud de un tramo dado de la travesía y $\beta$ es la dirección del tramo de la travesía (o el rumbo de la travesía). El término **salida** se usa para referirse al cambio en el valor de este ($\Delta E$), que puede expresarse como

$$
\Delta E = S \times \sin \beta \quad \text{(4.2)}
$$

El error de cierre para toda la travesía es la longitud (error lineal de cierre) (LC) y la dirección ($\theta$) que conecta las estaciones finales de la travesía calculadas y conocidas, lo que se puede dar como

$$
LC = \sqrt{(\epsilon_N)^2 + (\epsilon_E)^2} \quad \text{(4.3)}
$$

$$
\theta = \arctan \left( \frac{\epsilon_E}{\epsilon_N} \right) \quad \text{(4.4)}
$$

$$
\epsilon_N = \sum_{i=1}^{n} \Delta N_i \quad \text{(4.5)}
$$

$$
\epsilon_E = \sum_{i=1}^{n} \Delta E_i \quad \text{(4.6)}
$$

donde $\epsilon_N$ y $\epsilon_E$ son los errores totales de cierre de la travesía en las direcciones norte y este, respectivamente; LC es el error lineal de cierre; y $n$ es el número de tramos de travesía medidos en la travesía. La relación de LC a la longitud total (D) de la travesía se puede dar como

$$
RC = \frac{LC}{D} \quad \text{(4.7)}
$$

Las ecuaciones (4.1) - (4.7) se aplican de manera similar en los tres métodos tradicionales de ajuste. Las principales diferencias están en cómo se calculan las correcciones a los cambios en las coordenadas de norte y este. Estas diferencias se discuten en las subsecciones siguientes.

### 4.2.1 Método de Ajuste por Regla de Tránsito

El método de la regla de tránsito es un método empírico de ajuste que no tiene una base teórica. Se desarrolló inicialmente para ajustar errores aleatorios de cierre en levantamientos de travesías cerradas basándose en la suposición de que se utilizan equipos de tránsito y cinta. El método asume que los ángulos medidos con tránsito son más precisos que las distancias correspondientes medidas con cinta. Este método de ajuste, aunque no se usa comúnmente, calcula un ajuste o corrección ($c_{\Delta N_i}$) al cambio en el valor de norte ($\Delta N_i$) para un tramo de travesía dado $i$, como

$$
c_{\Delta N_i} = - \epsilon_N \left( \frac{|\Delta N_i|}{\sum_{i=1}^{n} |\Delta N_i|} \right) \quad \text{(4.8)}
$$

donde $|\Delta N_i|$ es el valor absoluto del cambio en el valor de norte para el tramo de travesía dado $i$. De manera similar, la corrección ($c_{\Delta E_i}$) a aplicar al cambio en el valor de este ($\Delta E_i$) para el tramo de travesía $i$ se puede dar como

$$
c_{\Delta E_i} = - \epsilon_E \left( \frac{|\Delta E_i|}{\sum_{i=1}^{n} |\Delta E_i|} \right) \quad \text{(4.9)}
$$

donde $|\Delta E_i|$ es el valor absoluto del cambio en el valor de este para el tramo de travesía $i$. Los cambios de coordenadas ajustados para un tramo de travesía dado $i$ se dan como

$$
\Delta \hat{N_i} = \Delta N_i + c_{\Delta N_i} \quad \text{(4.10)}
$$

$$
\Delta \hat{E_i} = \Delta E_i + c_{\Delta E_i} \quad \text{(4.11)}
$$

donde $\Delta \hat{N_i}$ y $\Delta \hat{E_i}$ son los cambios ajustados en los valores de norte y este para el tramo de travesía dado $i$. Las coordenadas ajustadas para un punto de travesía dado $k$ se dan como

$$
\hat{N_k} = \hat{N_{k-1}} + \Delta \hat{N_{k-1 \rightarrow k}} \quad \text{(4.12)}
$$

$$
\hat{E_k} = \hat{E_{k-1}} + \Delta \hat{E_{k-1 \rightarrow k}} \quad \text{(4.13)}
$$

donde $\hat{N_k}$ y $\hat{E_k}$ son los valores ajustados de las coordenadas de norte y este, respectivamente, para el punto de travesía actual $k$; $\hat{N_{k-1}}$ y $\hat{E_{k-1}}$ son los valores ajustados previamente de las coordenadas de norte y este, respectivamente, para el punto de travesía $k-1$ precedente al punto de travesía actual $k$; y $\Delta \hat{N_{k-1 \rightarrow k}}$ y $\Delta \hat{E_{k-1 \rightarrow k}}$ son los cambios ajustados en los valores de norte y este para el tramo de travesía desde el punto $k-1$ al punto $k$. El Ejemplo 4.1 se proporciona para ilustrar los pasos computacionales para el método de ajuste de tránsito.

### 4.2.2 Método de la Regla de la Brújula (Bowditch)

El método de la brújula (también conocido como método Bowditch) fue inventado entre 1773–1838 por el navegante americano conocido como Nathaniel Bowditch. El método fue diseñado originalmente para ajustar de manera estadística los errores aleatorios de cierre en travesías cerradas medidas con brújula y equipo de cinta. El método asume que las mediciones de dirección y distancia de un tramo de travesía no están correlacionadas y que las desviaciones estándar ($s_d$) de las mediciones de distancia y las de las mediciones de rumbo correspondientes ($s_\alpha$) son consistentes o equivalentes, es decir, $s_\alpha = d(s_d)$, donde $d$ es la distancia asociada. La otra suposición estadística de este método es que el error más probable en una medición de distancia será proporcional a la raíz cuadrada de esa medición de distancia (Sprinsky 1987; Crandall 1901). Dado que el equipo moderno de topografía proporciona precisiones de medición flexibles y diferentes tipos, las suposiciones de ajuste de la regla de la brújula se han vuelto innecesarias.

Sin embargo, el método de la regla de la brújula determina ajustes o correcciones a los cambios de coordenadas como sigue. Para un error total de cierre dado de una travesía en la dirección norte como $\epsilon_N$, la corrección ($c_{\Delta N_i}$) a aplicar al cambio en el valor de norte ($\Delta N_i$) para el tramo de travesía $i$ se puede dar como

$$
c_{\Delta N_i} = - \epsilon_N \left( \frac{S_i}{D} \right) \quad \text{(4.14)}
$$

donde $S_i$ es la longitud del tramo de travesía dado $i$ y $D$ es la longitud total de la travesía. De manera similar, la corrección ($c_{\Delta E_i}$) a aplicar al cambio en el valor de este ($\Delta E_i$) para el tramo de travesía $i$ se puede dar como

$$
c_{\Delta E_i} = - \epsilon_E \left( \frac{S_i}{D} \right) \quad \text{(4.15)}
$$

donde $\epsilon_E$ es el error total de cierre de la travesía en la dirección este. Nuevamente, los cambios de coordenadas ajustados para un tramo de travesía dado $i$ se dan por las Ecuaciones (4.10) y (4.11); y la coordenada ajustada norte ($\hat{N_k}$) y la coordenada ajustada este ($\hat{E_k}$) se calculan a partir de las Ecuaciones (4.12) y (4.13), respectivamente. ### 4.2.3 Método de la Regla de Crandall

El método de ajuste de la regla de Crandall fue desarrollado por Crandall (1901) para reflejar la instrumentación de ese tiempo aplicando el principio de mínima varianza. El método simula el método de mínimos cuadrados para ajustar errores de cierre en redes de travesía cerradas y asume que las mediciones de ángulo (o azimut) en las redes de travesía son más precisas que las mediciones de distancia. En este caso, el método atribuye los errores de cierre a las mediciones de distancia, distribuye las correcciones de error de cierre solo a las distancias y deja las mediciones de azimut (ángulo) sin ajustar. Esta suposición de mediciones de ángulo (azimut) sin errores, sin embargo, es incorrecta para equipos modernos de topografía, como la estación total. Las correcciones $c_{\Delta N_i}$ y $c_{\Delta E_i}$ que se deben aplicar a los cambios de coordenadas usando este método se pueden expresar como

$$
c_{\Delta N_i} = A (\Delta N_i)^2 + B (\Delta N_i \times \Delta E_i) \quad \text{(4.16)}
$$

$$
c_{\Delta E_i} = A (\Delta N_i \times \Delta E_i) + B (\Delta E_i)^2 \quad \text{(4.17)}
$$

$$
B = \Large{ \frac{ \left[ \epsilon_N \left( \sum_{i=1}^{n} \Delta N_i \times \Delta E_i \right) - \epsilon_E \left( \sum_{i=1}^{n} (\Delta N_i)^2 \right) \right] }{\left[ \left( \sum_{i=1}^{n} (\Delta E_i)^2 \right) \left( \sum_{i=1}^{n} (\Delta N_i)^2 \right) - \left( \sum_{i=1}^{n} (\Delta N_i \times \Delta E_i) \right)^2 \right]} } \quad \text{(4.18)}
$$

$$
A = \Large{\frac{- \left[ B \left( \sum_{i=1}^{n} \Delta N_i \times \Delta E_i \right) + \epsilon_N \right]}{\left( \sum_{i=1}^{n} (\Delta N_i)^2 \right) } }
\quad \text{(4.19)}
$$

Nuevamente, los cambios de coordenadas ajustados para un tramo de travesía dado $i$ se dan por las Ecuaciones (4.10) y (4.11); y la coordenada ajustada norte ($\hat{N_k}$) y la coordenada ajustada este ($\hat{E_k}$) se calculan a partir de las Ecuaciones (4.12) y (4.13), respectivamente. El Ejemplo 4.3 se proporciona para ilustrar los pasos computacionales para el método de ajuste de la regla de Crandall.

### 4.3 El Método de Mínimos Cuadrados

Hoy en día, los métodos tradicionales de ajuste de mediciones son inadecuados para levantamientos de precisión que involucran grandes mediciones redundantes (para verificar errores groseros en las mediciones), diferentes técnicas de medición (como en levantamientos de triangulación y trilateración), y precisiones de medición variables. Por ejemplo, los métodos tradicionales muestran claramente sus debilidades cuando se ajustan redes de travesía con enlaces cruzados y en el ajuste de redes de sistemas de navegación global por satélite (GNSS) posprocesados donde se involucran muchas observaciones. El método de ajuste que puede acomodar todas estas propiedades y evaluar mediciones redundantes al mismo tiempo es el método de mínimos cuadrados, que generalmente se refiere como estimación de mínimos cuadrados o ajuste de mínimos cuadrados. El método de mínimos cuadrados proporciona un tratamiento organizado y matemáticamente lógico de los errores aleatorios de observación y las inconsistencias en los modelos matemáticos. Utiliza un enfoque de distribución de errores en la aplicación de correcciones a las observaciones. Las características principales del ajuste de mínimos cuadrados son ajustar las observaciones para eliminar posibles cierres incorrectos y mantener los cambios en las observaciones lo más pequeños posible (de acuerdo con lo que se conoce como el criterio de mínimos cuadrados).

Los conceptos del método de mínimos cuadrados han evolucionado a lo largo de varios años. A continuación se presenta un resumen de algunos de los aspectos importantes de la historia del método (Merriman 1877; Cooper 1987). De hecho, los antecedentes del método comenzaron con los diversos estudios e investigaciones llevadas a cabo por los astrónomos y científicos, tales como los siguientes:

- En 1788, Laplace (científico francés) utilizó el llamado método de promedios para combinar diferentes datos recopilados bajo las mismas condiciones.
- Boscovich en 1757 y Laplace en 1799 sugirieron el uso del método de desviación absoluta mínima para combinar diferentes datos recopilados bajo diferentes condiciones.

- En 1795, 1806 y 1809, **Legendre** (matemático francés) y **Gauss** (científico alemán) concibieron el concepto de optimización de la suma de cuadrados de las correcciones a las mediciones. El nombre "método de mínimos cuadrados" fue publicado por primera vez por Legendre (1806), pero la descripción detallada del método fue proporcionada por Gauss (1809), quien afirmó haber derivado independientemente el método de mínimos cuadrados desde 1795. Sin embargo, se entiende que Legendre y Gauss abordaron el método de mínimos cuadrados de manera diferente:
  - **Legendre** utilizó modelos matemáticos para eliminar cierres incorrectos y proporcionó consistencia numérica de observaciones conectadas por modelos. Demostró el procedimiento ajustando ecuaciones lineales a algunas observaciones. En este caso, trató el método de mínimos cuadrados como un procedimiento de estimación numérica pura sin considerar la distribución de probabilidad.
  - **Gauss** asumió la normalidad de las mediciones para producir resultados con la mayor probabilidad de ser verdaderos. Descubrió que los errores observacionales siguen la ley de la función de distribución de probabilidad normal, que utilizó para proporcionar propiedades numéricas y estadísticas de los parámetros desconocidos. En su publicación de 1809, Gauss utilizó el método de mínimos cuadrados para calcular la órbita de cuerpos celestiales y conectó el método con los principios de probabilidad y distribución normal.
  - **Gauss** (1809) y **Laplace** (1811) especificaron un modelo matemático para la densidad de probabilidad de las observaciones. También demostraron que los errores observacionales siguen la ley de la distribución de probabilidad normal, con la media de las mediciones repetidas como el valor más probable del parámetro desconocido. Este método es comúnmente conocido como el **método de estimador de máxima verosimilitud**.

- En general, alrededor de 1800, **Legendre**, **Gauss** y **Laplace** fueron capaces de ajustar modelos funcionales a los datos mediante diferentes tipos de métodos de mínimos cuadrados.

El enfoque de solución real de las inconsistencias causadas por mediciones redundantes se entiende que fue proporcionado por **Legendre** y **Gauss**, y son considerados como los fundadores o padres del método de mínimos cuadrados. Sin embargo, el método de mínimos cuadrados se volvió atractivo para los geodestas algún tiempo después. Los geodestas emplearon el método para determinar el elipsoide que mejor se ajusta basado en mediciones de arco meridiano inconsistentes. Algunos de los geodestas que hicieron sus contribuciones de una forma u otra al método de mínimos cuadrados son los siguientes:

- **Adrian** (topógrafo americano) (1808) formuló de manera independiente el principio del análisis de mínimos cuadrados, pero no se refirió a él con ese nombre. También sugirió el uso de la distribución normal para describir errores en mediciones desde tan temprano como 1808, incluso antes de Gauss.
- **Bessel** (astrónomo y geodesta alemán) en 1837 resolvió el problema del ajuste de mínimos cuadrados de modelos paramétricos determinando los elementos de la figura de la Tierra (como el achatamiento, los radios polares y ecuatoriales, etc.) basándose en las coordenadas geodésicas disponibles. El método de ajuste de modelos paramétricos se ha convertido en el procedimiento de ajuste más útil hoy en día.
- **Helmert** (geodesta alemán) en 1907 determinó los coeficientes de la transformación lineal mediante el método paramétrico de mínimos cuadrados conocido como la **transformación de Helmert**.
- **Bjerhammar** (geodesta sueco) (1951) contribuyó a la simplificación del método de mínimos cuadrados con el uso de notaciones matriciales.

### 4.3.1 Criterio de Mínimos Cuadrados

Para ilustrar algunas de las características del método de mínimos cuadrados, considere un problema de resección en la Figura 4.2 en el cual las coordenadas $(x, y)$ del punto P deben ser determinadas. En este problema, los ángulos horizontales ($\alpha, \beta, \gamma$) y las distancias horizontales ($d_1, d_2, d_3$) con sus desviaciones estándar asociadas son medidos desde el punto P hasta tres puntos de control A, B y C cuyas coordenadas son bien conocidas.

Las seis mediciones en la Figura 4.2 deben ser ajustadas para hacer que se ajusten juntas en la determinación de las dos coordenadas desconocidas $(x, y)$ del punto P. En este caso, la suma de los ángulos en el punto P alrededor del horizonte debe ser igual a 360°, y las correcciones a las mediciones de distancia deben ser lo más pequeñas posible; todas estas condiciones deben ser satisfechas simultáneamente. Las correcciones (o residuos) a aplicar a las mediciones dependen de qué tan bien se ajusten las mediciones. Un enfoque para resolver las coordenadas del punto P es adivinar las coordenadas del punto P usando las coordenadas de los puntos A, B y C como guía, luego calcular independientemente las mediciones y usarlas para determinar los residuos. El método de mínimos cuadrados corrige iterativamente las coordenadas adivinadas iniciales de modo que la suma ponderada de las correcciones a las mediciones sea un valor mínimo. Esto significa que el método de mínimos cuadrados está diseñado para mantener los cambios (o ajustes) a las mediciones originales lo más pequeños posible. Esto conduce a lo que se conoce como el criterio de mínimos cuadrados o la condición de mínimos cuadrados. El criterio general de mínimos cuadrados establece que la suma de los cuadrados ponderados de las correcciones a las mediciones es un mínimo. Esto puede expresarse matemáticamente como

$$
v^T P v = \text{mínimo} \quad \text{(4.20)}
$$

donde $v$ es el vector de residuos de observación (correcciones a aplicar a las mediciones después del ajuste o las diferencias entre las mediciones reales y los valores predichos por el modelo) y $P$ es la matriz de pesos de las observaciones derivada de las desviaciones estándar de las observaciones. Si la matriz de pesos es diagonal (con elementos fuera de la diagonal como ceros), el criterio de mínimos cuadrados en la Ecuación (4.20) puede reescribirse como

$$
v_1^2 p_{11} + v_2^2 p_{22} + v_3^2 p_{33} + \cdots + v_n^2 p_{nm} = \text{mínimo} \quad \text{(4.21)}
$$

donde $P = \text{diag}(p_{11}, p_{22}, p_{33}, \ldots, p_{nm})$ es la matriz de pesos de las mediciones y $v = (v_1, v_2, v_3, \ldots, v_n)^T$ es el vector de residuos, que debe determinarse. En otras palabras, el método de mínimos cuadrados proporciona una forma de encontrar el valor más probable de una cantidad medida a través del uso del criterio de mínimos cuadrados. Para demostrar la aplicación del criterio de mínimos cuadrados, considere que una cantidad observable $\ell$ se ha medido repetidamente $n$ veces con el vector de mediciones como $\ell = (\ell_1, \ell_2, \ell_3, \ldots, \ell_n)^T$; la matriz de pesos de las mediciones como $P = \text{diag}(p_{11}, p_{22}, p_{33}, \ldots, p_{nm})$, asumiendo que diferentes medios fueron usados para medir el observable de modo que cada medición tiene diferente peso; la media de las observaciones como $\bar{\ell}$ (desconocido que debe determinarse); y el vector de residuos como

$$
v = (v_1, v_2, v_3, \ldots, v_n)^T \quad \text{(también desconocido que debe determinarse)}
$$

Los únicos valores conocidos para el observable son las mediciones y sus pesos. El criterio de mínimos cuadrados puede usarse para determinar los desconocidos como sigue. El criterio de mínimos cuadrados se formula como en la Ecuación (4.21) de la siguiente manera:

$$
v_1^2 p_{11} + v_2^2 p_{22} + v_3^2 p_{33} + \cdots + v_n^2 p_{nm} = \text{mínimo} \quad \text{(4.22)}
$$

El mínimo de la Ecuación (4.22) se encuentra estableciendo las derivadas parciales de la ecuación (con respecto a cada residuo de observación) a cero como sigue:

$$
\frac{\partial \left( v_1^2 p_{11} \right)}{\partial v_1} + \frac{\partial \left( v_2^2 p_{22} \right)}{\partial v_2} + \frac{\partial \left( v_3^2 p_{33} \right)}{\partial v_3} + \cdots + \frac{\partial \left( v_n^2 p_{nm} \right)}{\partial v_n} = 0 \quad \text{(4.23)}
$$

o

$$
2v_1p_{11} + 2v_2p_{22} + 2v_3p_{33} + \cdots + 2v_np_{nm} = 0 \quad \text{(4.24)}
$$

Recordando que un residuo ($v_i$) es la diferencia entre la medición ($\ell_i$) y su valor medio ($\bar{\ell}$), expresado como $v_i = (\bar{\ell} - \ell_i)$, la Ecuación (4.24) puede reescribirse como

$$
(\bar{\ell} - \ell_1)p_{11} + (\bar{\ell} - \ell_2)p_{22} + (\bar{\ell} - \ell_3)p_{33} + \cdots + (\bar{\ell} - \ell_n)p_{nm} = 0 \quad \text{(4.25)}
$$

La Ecuación (4.25) se puede resolver para dar

$$
\bar{\ell} = \frac{\ell_1p_{11} + \ell_2p_{22} + \ell_3p_{33} + \cdots + \ell_np_{nm}}{p_{11} + p_{22} + p_{33} + \cdots + p_{nm}} \quad \text{(4.26)}
$$

La Ecuación (4.26), que es simplemente la media ponderada de las mediciones, es la solución de mínimos cuadrados para las mediciones $\ell$ que tienen una matriz de pesos $P$. Se puede declarar que la estimación de mínimos cuadrados es simplemente un promedio ponderado de las observaciones para el parámetro desconocido. En este caso, todas las observaciones de alguna manera contribuyen al "mejor" parámetro estimado. La media ponderada es el valor más probable en este problema. El método ordinario de mínimos cuadrados se refiere como el mejor estimador lineal insesgado (BLUE) de parámetros en relación con el modelo de Gauss-Markov. Un modelo de Gauss-Markov es un modelo de regresión lineal en el cual los errores no correlacionados tienen expectativa cero e igual varianza. El término "mejor" significa que se logra el menor posible error cuadrático medio del parámetro estimado. También se sabe que cuando se minimiza la suma ponderada de los residuos cuadrados, como en la Ecuación (4.21), el parámetro estimado es BLUE si cada peso es igual al recíproco de la varianza de la medición. Si los errores están correlacionados, el estimador resultante es también BLUE si la matriz de pesos es igual a la inversa de la matriz de varianza-covarianza de las observaciones, como en el caso de la Ecuación (4.20). Las relaciones entre las desviaciones estándar de las mediciones, la matriz de varianza-covarianza, la matriz de cofactores y la matriz de pesos se discutieron en el Capítulo 1. De ese capítulo, se puede ver que la matriz de pesos $P$ de las observaciones se puede dar como

$$
P = \sigma_0^2 C_\ell^{-1} \quad \text{(4.27)}
$$

donde $C_\ell = \sigma_0^2 Q$ es la matriz de varianza-covarianza del vector de observación $\ell$ (con las desviaciones estándar cuadradas o varianzas de las observaciones como los elementos diagonales), $\sigma_0^2$ es el factor de varianza a priori de unidad de peso (que siempre se toma como la varianza de la población general con un valor de 1), y $Q$ es la matriz de cofactores del vector de observación.

### 4.4 Tipos de Modelos de Ajuste de Mínimos Cuadrados

Los conceptos de observables de levantamiento, ajustes, modelos funcionales y modelos estocásticos se discutieron en el Capítulo 1. Los modelos funcionales relacionan los observables del levantamiento (como distancias, direcciones, azimuts, ángulos y diferencias de elevación) con los parámetros desconocidos como las coordenadas de los puntos de la red de trabajo. Antes de que se pueda realizar cualquier ajuste, se debe formular un modelo funcional que relacione los observables con los parámetros, o se debe identificar y formular un modelo funcional que relacione los observables entre sí.

La necesidad de ajuste de modelos se basa en la redundancia (o sobredeterminación) de los modelos; no habrá necesidad de ajuste si los modelos no están sobredeterminados. Dependiendo de cómo estos observables se relacionen con los parámetros desconocidos o entre sí, los modelos matemáticos formados se pueden referir como paramétricos, condicionales o generales. Los modelos paramétricos tienen una observación ($\ell_i$) por ecuación, que es una función de algunos o todos los parámetros desconocidos ($x$). Dado que el número de ecuaciones es el mismo que el número de observaciones, el número de grados de libertad (redundancia) es el número de ecuaciones (u observaciones) menos el número de parámetros. Los modelos condicionales relacionan geométricamente solo las observaciones ($\ell$) entre sí sin involucrar ningún parámetro en las ecuaciones. Dado que el número de ecuaciones independientes involucradas puede ser menor que el número de observaciones, el número de grados de libertad para el ajuste de modelo condicional es el mismo que el número de ecuaciones de modelo independiente formuladas. Los modelos generales de alguna manera tienen las observaciones ($\ell$) entrelazadas con los parámetros ($x$) en las mismas ecuaciones; el número de ecuaciones involucradas menos el número de parámetros es el número de grados de libertad. Debe mencionarse, sin embargo, que el número de ecuaciones involucradas en el ajuste de modelo general no es necesariamente el mismo que el número de observaciones, como en el caso del ajuste de modelo paramétrico; generalmente es menor.

El caso especial de modelos generales en los que cada parámetro desconocido es una función de observaciones se denomina modelo especial en este libro. Este tipo de modelo se formuló para problemas de propagación de errores en el Capítulo 2. Las relaciones entre los diferentes modelos se muestran en la Figura 4.3. Ejemplos de cada modelo se dan en los capítulos correspondientes que tratan con métodos de ajuste de mínimos cuadrados paramétricos, condicionales y generales. La convención que se seguirá en los siguientes métodos de ajuste de mínimos cuadrados es que los símbolos que representan observaciones no ajustadas y los parámetros no ajustados no tendrán sombreros, como $\ell$ para observaciones (cantidades medidas) y $x$ como parámetros no ajustados (cantidades que no se miden pero cuyos valores se estimarán); las observaciones ajustadas se representarán con sombreros por $\hat{\ell}$, y los parámetros ajustados se representarán con sombreros por $\hat{x}$. Debe enfatizarse que siempre que se formule cada tipo de los diferentes modelos (paramétrico, condicional, general), el orden de disposición de las observaciones y los parámetros en cada lado del signo de igualdad en las ecuaciones es muy importante. La idea detrás del uso de cantidades ajustadas (aquellas representadas con "sombrero" en ellas) en la formulación de ecuaciones del modelo es indicar que esas cantidades van a ser ajustadas y las ecuaciones así formadas son completas y consistentes, es decir, no producirán ningún cierre incorrecto.

Dado que el álgebra de matrices es conveniente para representar ecuaciones de modelo en la estimación de mínimos cuadrados, se utilizará a lo largo de este libro. A modo de revisión, se presentan matrices introductorias en el Capítulo 1, y algunos lemas importantes de matrices se proporcionan en el Apéndice E. Debe mencionarse aquí que, siempre que se estén formulando ecuaciones de modelo matemático, no se deben usar inmediatamente los valores numéricos de las observaciones, sino los símbolos apropiados que representan esas mediciones, como $\hat{\ell_1}, \hat{\ell_2}, \Delta h_1, \Delta h_2$. También debe recordarse que en la etapa de formulación de ecuaciones, uno no está resolviendo las ecuaciones, sino solo presentándolas simbólicamente. También debe señalarse que, en esta etapa, lo importante es cómo identificar los parámetros desconocidos y las observaciones. Los parámetros desconocidos ($x$) son cantidades que uno quiere determinar, y las observaciones ($\ell$) son las mediciones o cantidades proporcionadas con sus desviaciones estándar o matriz de covarianza que se utilizarán para determinar los parámetros desconocidos.

### 4.5 Pasos de Ajuste de Mínimos Cuadrados

Después de que se hayan recolectado las mediciones del levantamiento, las mediciones deben corregirse por algunas condiciones físicas, tales como reducciones de marca a marca, correcciones de refracción a ángulos cenitales, correcciones debido a desviaciones de la vertical para convertir el azimut astronómico en geodésico, corrección de escala, etc. Después de que estas reducciones y correcciones se hayan aplicado a las mediciones, se obtienen **observaciones corregidas**. Las observaciones corregidas no deben confundirse con **observaciones ajustadas**. Las observaciones ajustadas son observaciones corregidas cuyos errores aleatorios han sido minimizados a través del proceso de ajuste que involucra observaciones redundantes; las observaciones ajustadas producen los mejores valores de las cantidades desconocidas. Las observaciones reducidas y corregidas se evalúan utilizando la herramienta de mínimos cuadrados, como se ilustra en la Figura 4.4.

Básicamente, el método de mínimos cuadrados requiere los siguientes pasos como se indica en la Figura 4.4:

- Formular algunos modelos matemáticos adecuados (paramétrico, condicional o general) y modelo estocástico (ver Capítulos 5, 12 y 13).
- Incorporar los modelos en el criterio de mínimos cuadrados (formando lo que se conoce como función de variación) (ver Secciones 5.5, 12.3 y 13.4).
- Encontrar las derivadas parciales del **criterio de mínimos cuadrados** o la función de variación (formulada a partir del criterio de mínimos cuadrados), estableciendo las derivadas parciales a cero y resolviéndolas. El "mínimo" del criterio de mínimos cuadrados o de la función de variación se encuentra estableciendo las derivadas parciales del criterio de mínimos cuadrados o de la función de variación a cero (ver Secciones 5.5–5.7, 12.3 y 13.5).
- Detectar posibles errores que puedan degradar la solución utilizando el enfoque de pruebas estadísticas (ver Capítulo 11).
- Calcular las soluciones finales y sus propiedades estadísticas (ver Secciones 5.7–5.8, 12.4 y 13.6–13.8).
- Evaluar los resultados del ajuste para determinar si se cumplen los estándares deseados.

El punto de partida del ajuste de mínimos cuadrados es la formulación de modelos matemáticos funcionales que relacionan las observaciones con las cantidades desconocidas (los parámetros), como en los casos de ajustes de modelos paramétricos y generales, o que relacionan las observaciones entre sí, como en el caso del ajuste de modelo condicional. Los modelos funcionales en geomática pueden ser no lineales. Sin embargo, el método de mínimos cuadrados generalmente se realiza con ecuaciones lineales. Esto requiere que se adopten medios para linearizar modelos no lineales. El método comúnmente utilizado para linearizar las ecuaciones en geomática es usar los términos de primer y cero orden de la expansión en serie de Taylor, como se discutió en el Capítulo 2. Dado que solo se utilizan los términos lineales (términos de primer y cero orden) de las series de Taylor para aproximar los modelos funcionales a ajustar, se requiere que los valores aproximados de las cantidades desconocidas se determinen lo suficientemente cerca de los resultados finales del ajuste, es decir, las cantidades ajustadas. Si este no es el caso, la expansión en serie de Taylor no representará suficientemente el modelo funcional original y los valores corregidos de los parámetros del ajuste deben usarse nuevamente (ahora como un nuevo conjunto de valores aproximados o iniciales) en el nuevo procedimiento de ajuste de mínimos cuadrados. Este procedimiento puede repetirse tantas veces como se considere necesario hasta que la solución final sea insignificativamente diferente de la anterior. Cuanto más cerca estén los valores iniciales de la solución final, menor será el número de iteraciones necesarias. Esto significa que la elección de los valores iniciales de las cantidades desconocidas a determinar en el ajuste de mínimos cuadrados es importante; sin embargo, no hay un medio especial para elegirlos. Por lo general, se puede usar algo de experiencia o algunos medios computacionales simples y rápidos para elegirlos.

### 4.6 Definición y Ajustes del Datum de la Red

Un **datum de red** es una referencia o una red base a la cual se adjunta otra red de mediciones a través del proceso de ajuste. En un ajuste donde se deben determinar las coordenadas de los puntos, siempre hay necesidad de un sistema de coordenadas, conocido también como datum. En otras palabras, se puede afirmar que al ajustar una red de levantamiento en la cual se desean coordenadas de puntos de la red, primero se debe adjuntar un sistema de coordenadas (un datum) a la red de levantamiento. Adjuntar un datum a una red de levantamiento es lo mismo que definir los **parámetros del datum** para la red. El ajuste de mínimos cuadrados de la red de levantamiento solo será posible si se definen los parámetros del datum. Los elementos del datum que deben conocerse antes de ser utilizados se conocen como **parámetros del datum**. Los parámetros típicos de un datum son su **ubicación espacial**, **orientación** y **escala**.

En geomática, generalmente se está preocupado con posicionar o determinar posiciones (o coordenadas) de puntos en o cerca de la superficie de la Tierra. Las posiciones o coordenadas de los puntos, a diferencia de las cantidades medibles como las distancias y los ángulos, deben especificarse con respecto a algunos parámetros del datum. El número de parámetros del datum necesarios en un ajuste depende del tipo de posicionamiento o coordenadas que se deben determinar. Un datum común para el posicionamiento espacial tridimensional (3-D) es generalmente un sistema de coordenadas ortogonales tridimensionales (X, Y, Z) con mano derecha que tiene una ubicación o origen bien definido ($x = 0, y = 0, z = 0$); la orientación de sus ejes de coordenadas X, Y y Z es conocida con respecto a la Tierra; y se conoce la unidad de longitud (o escala) a utilizar con el sistema de coordenadas. Estos siete parámetros (tres traducciones o coordenadas $x, y, z$ del origen, tres elementos de orientación de los ejes de coordenadas y una unidad de longitud o una escala para el sistema de coordenadas) se conocen como **parámetros del datum** para problemas de posicionamiento en 3-D. El datum para una red 3-D se define prácticamente especificando valores para los siete parámetros del datum de la siguiente manera:

- Especificar tres coordenadas ($x, y, z$) de al menos uno de los puntos de la red para definir la ubicación de la red en el espacio.
- Especificar tres acimutes espaciales de los ejes de coordenadas X, Y y Z de la red; o usar ángulos horizontales y cenitales, si se miden, para proporcionar dos parámetros de rotación con un acimut completando el parámetro de rotación restante.
- Medir la distancia espacial de al menos una de las líneas en la red para proporcionar la escala requerida para la red.
- Alternativamente, fijar tres coordenadas ($x, y, z$) de un punto con diferencias de altura medidas y acimutes medidos es suficiente; o fijar las coordenadas ($x, y, z$) de dos puntos y la altura de otro punto puede ser suficiente en el posicionamiento en 3-D.

Si se deben determinar coordenadas bidimensionales (2-D) ($x, y$) en un ajuste, el número de parámetros del datum será de cuatro: dos traducciones o coordenadas $x, y$ del origen, un elemento de orientación de los dos ejes de coordenadas, y una unidad de longitud o escala para el sistema de coordenadas. El datum para una red 2-D se define prácticamente especificando valores para los cuatro parámetros del datum de la siguiente manera:

- Especificar dos coordenadas ($x, y$) de al menos uno de los puntos de la red para definir la ubicación o el origen de la red en un plano de proyección de mapa.
- Especificar los acimutes de los ejes de coordenadas X y Y de la red; el acimut o el rumbo de una línea en la red proporcionará el parámetro de orientación necesario; los ángulos horizontales, sin embargo, no proporcionan ningún valor para el parámetro de orientación de una red 2-D.
- Medir la distancia horizontal de al menos una de las líneas en la red para proporcionar la escala necesaria para la red.
- Alternativamente, fijar las coordenadas ($x, y$) de dos puntos puede ser suficiente para el ajuste 2-D.

Si se deben determinar coordenadas unidimensionales (1-D) (elevaciones o alturas de puntos) en un ajuste, el número de parámetros del datum requeridos será de uno: la elevación o altura de un punto. El datum para una red 1-D (como una red de nivelación diferencial) se define si se especifica la elevación de un punto de referencia (con las mediciones de diferencia de altura proporcionando la escala necesaria). Note que los problemas 1-D no requieren un parámetro de orientación. Por lo general, la diferencia de altura, si se mide entre dos puntos de una red, proporcionará dos parámetros de rotación y un parámetro de escala si la red es 3-D y un parámetro de escala si la red es 2-D.

En resumen, cualquier ajuste de una red de levantamiento que solo satisfaga el número mínimo de parámetros del datum mencionados anteriormente se conoce como ajuste de restricción mínima.

### 4.6.1 Defecto del Datum y Defecto de Configuración

Los parámetros del datum que no se tienen en cuenta en el ajuste de mediciones de la red constituyen lo que se suele llamar **defectos del datum** o **defectos en el datum**. Un defecto del datum es la cantidad por la cual los parámetros del datum son desconocidos o faltan. Está estrechamente relacionado con los parámetros del datum que no se definen a través de observaciones. Por ejemplo, en un levantamiento de travesía en 2-D, en el cual generalmente se miden distancias y ángulos, si no hay un rumbo de una línea conocido, habrá 3 defectos del datum como resultado de no tener un rumbo y no tener coordenadas ($x, y$) fijas o restringidas. Si la travesía se realiza ahora con un rumbo de una línea conocido, habrá 2 defectos del datum por falta de coordenadas ($x, y$) fijas o restringidas. Una de las consecuencias de los defectos del datum es que hacen imposible el ajuste de mínimos cuadrados de la red de levantamiento donde las coordenadas son los parámetros desconocidos.

Los **defectos de configuración** surgen cuando el número o la elección de las observaciones no son suficientes para determinar de manera única las coordenadas de los puntos de la red. Por ejemplo, las redes de triangulación tienen defectos del datum así como defectos de configuración, debido a la falta de información sobre la escala; el defecto de configuración en este caso se puede eliminar midiendo una distancia; si los parámetros del datum se definen para la red, entonces estará libre de defectos del datum. A veces, se puede definir un datum para una red, pero la red aún puede tener un defecto de configuración, por ejemplo, si se realiza una medición lateral a un punto con solo una medición de distancia sin una medición de ángulo para conectar la medición lateral a la red principal, será imposible determinar las coordenadas de ese punto utilizando solo la medición de distancia; esta red se dice que tiene un defecto de configuración. Los puntos que están desconectados de la red principal crearán defectos de configuración para la red.

### 4.7 Restricciones en el Ajuste

**Ecuaciones de restricción** son las ecuaciones necesarias para completar una definición de datum. Esto es necesario ya que el datum para las coordenadas desconocidas a determinar está incompletamente definido por las mediciones. El tipo y el número de ecuaciones de restricción necesarias dependen del número de **parámetros del datum** que no están definidos por las mediciones de la red que se están ajustando, es decir, el número de defectos del datum. Estos defectos deben ser resueltos formulando ecuaciones de restricción apropiadas. El número de ecuaciones de restricción necesarias en un ajuste depende del tipo de posicionamiento o coordenadas que se deben determinar y también del número de parámetros del datum resueltos por las mediciones de levantamiento. Generalmente, algunos de los parámetros del datum resueltos por las mediciones de levantamiento son los siguientes:

- **Parámetro de escala:** Esto se resuelve mediante mediciones de distancia horizontal, mediciones de distancia de marca a marca de levantamientos GPS, mediciones de distancia de nivelación de precisión o mediciones de diferencia de altura.
- **Parámetros de posición:** Estos se resuelven proporcionando coordenadas apropiadas (1-D, 2-D o 3-D) de un punto dependiendo de si se están determinando coordenadas 1-D, 2-D o 3-D en el ajuste.
- **Parámetros de orientación:** Estos se resuelven midiendo o fijando un acimut (o rumbo) de una línea para el posicionamiento en 2-D ($x, y$). Para determinaciones 3-D ($x, y, z$), los parámetros de orientación se definirán midiendo un acimut y ángulos horizontales y verticales (cenitales); un acimut, ángulos horizontales y diferencias de altura; acimutes y diferencias de altura; o un acimut, distancias horizontales y diferencias de altura.

Uno puede fijar una coordenada incluyendo una ecuación apropiada que especifique el valor a asignar a la coordenada, tal como:

$$
x + v_x = \hat{x} \quad \text{(4.28)}
$$

$$
y + v_y = \hat{y} \quad \text{(4.29)}
$$

donde las coordenadas medidas ($x, y$) con sus correcciones ($v_x, v_y$) se expresan como funciones de las coordenadas ajustadas ($\hat{x}, \hat{y}$) ellas mismas. Cuando las coordenadas se expresan matemáticamente como en las Ecuaciones (4.28) y (4.29), las coordenadas se observan directamente, formando las **ecuaciones de restricción**. En este caso, siempre que se formen las ecuaciones paramétricas, es necesario agregar un conjunto de ecuaciones de restricción que completarán la definición del datum de coordenadas. Estas ecuaciones de restricción fuerzan a esos parámetros involucrados a cumplir las reglas definidas por las ecuaciones. La otra forma de aplicar estas ecuaciones de restricción en el ajuste es omitirlas al formar la primera matriz de diseño; en este caso, el número de parámetros a determinar en el ajuste se reduce. Este enfoque se adopta a lo largo de este libro.

### 4.7.1 Ajustes de Restricción Mínima

Cuando el número y los tipos de ecuaciones de restricción son suficientes para definir un datum dado, el ajuste resultante de tal sistema se llamará **ajuste de restricción mínima**. Este tipo de ajuste proporcionará coordenadas ajustadas que dependen únicamente de las observaciones y de la elección arbitraria de los parámetros del datum. El ajuste de mínimos cuadrados será imposible sin al menos imponer ecuaciones de restricción mínimas. En la práctica, se dice que una red de levantamiento tiene restricciones mínimas si cumple con cualquiera de los siguientes:

1. **En una red de nivelación 1-D** (donde solo se miden diferencias de elevación), la elevación de un punto en la red puede fijarse (no está sujeta a cambios) mientras que todos los puntos restantes de la red son libres de cambiar mediante el proceso de ajuste.

2. **En una red de triangulación 2-D** (donde solo se miden ángulos), las coordenadas de dos puntos en la red pueden fijarse (no están sujetas a cambios) mientras que todos los puntos restantes de la red son libres de cambiar (no fijados). En este caso, las coordenadas de uno de los dos puntos se utilizan para definir la ubicación de la red; el rumbo y la distancia calculada entre los dos puntos proporcionan la orientación (rumbo) y la escala necesaria para completar la definición del datum.

3. **En una red de trilateración 2-D** (donde solo se miden distancias), las coordenadas de un punto en la red pueden fijarse (para definir la ubicación de la red); una coordenada ($x$ o $y$) de otro punto de la red puede fijarse para completar el datum si las coordenadas aproximadas de los otros puntos de la red son "bien conocidas". En este caso, las mediciones de distancia proporcionan la escala y las coordenadas fijadas, y las coordenadas aproximadas de los puntos de la red definirán la orientación del acimut de la red.

4. **En una red combinada de triangulación y trilateración 2-D** (también conocida como red de triangulateración 2-D) en la que se mide un acimut de una línea, las coordenadas de un punto en la red pueden fijarse (para definir la ubicación de la red); el acimut y las mediciones de ángulo definirán la orientación de la red, mientras que las mediciones de distancia proporcionarán la escala necesaria para que la red complete la definición del datum.

5. **En una red de travesía 2-D** en la que se miden ángulos y distancias en estaciones de travesía, el datum se define de manera similar a los casos de la red de trilateración 2-D y la red combinada de triangulación y trilateración 2-D.

6. **En una red de travesía 3-D** en la que se miden ángulos horizontales, ángulos cenitales, distancias de pendiente y diferencias de elevación, el datum se define fijando las coordenadas 3-D de uno de los puntos de la red y el acimut de una línea.

Fijar cualquier otra coordenada en cualquiera de las redes anteriores sobrecargará la red, pero agregar más mediciones como ángulos, distancias e incluso acimutes no sobrecargará la red. Ajustar una red con las restricciones mínimas mencionadas anteriormente se conoce como ajuste de restricción mínima. Por ejemplo, la solución solo GPS (en la que las coordenadas ($x, y, z$) de un punto de referencia se conocen con los vectores de línea base ($dx, dy, dz$) medidos) es una solución de restricción mínima; las coordenadas del punto de referencia proporcionan el parámetro de ubicación, y el vector de línea base proporciona los parámetros de orientación y escala. También debe recordarse que si no se establece un datum para una red, será imposible ajustar la red mediante el método de mínimos cuadrados sin importar el número de observaciones redundantes involucradas.

### 4.7.2 Ajustes Sobrerrestringidos y Ajustes Restringidos por Peso

En un ajuste de restricción mínima, las coordenadas publicadas (si no se fijan en el ajuste) aún pueden moverse en una dirección (lo cual es normalmente inaceptable). Podría ser preferible asegurarse de que la red de control existente se mantenga fija restringiendo (o manteniendo fijas) sus coordenadas. Esto es común en la práctica, asumiendo que la calidad de la red de control es muy superior a la calidad de las nuevas observaciones que se agregan. Cuando una red está restringida a dos o más estaciones de control, se dice que está **sobrerrestringida**. Una red sobrerrestringida se ve obligada a acomodar (las mediciones), el control previamente establecido (errores en el levantamiento anterior) y el nuevo levantamiento. En esta red, cualquier error en los parámetros fijados deformará las observaciones ajustadas y los parámetros ajustados. En otras palabras, cuando se utilizan más ecuaciones de restricción en un ajuste de las que se necesitan, se dice que el ajuste está restringido o lo que comúnmente se conoce como **ajuste sobrerrestringido**. En este caso, las cantidades ajustadas resultantes están restringidas para estar en el sistema de coordenadas adecuado y también para ajustarse a las restricciones adicionales. Las elipses de error derivadas para los nuevos puntos, sin embargo, aparecerán demasiado optimistas en el sentido de que no se consideran los errores en los puntos de control. Cada vez que un nuevo levantamiento se ajusta a un sistema de coordenadas existente utilizando puntos de control existentes, se realiza un ajuste sobrerrestringido (o simplemente restringido).

Los ajustes restringidos tienen algunas ventajas sobre el ajuste de restricción mínima, que incluyen las siguientes:

- Colocar nuevos levantamientos en algún sistema de coordenadas más grande proporcionando el datum (origen y orientación) para el nuevo levantamiento.
- Proporcionar los medios para detectar errores groseros en el nuevo levantamiento; esto hace que las observaciones ajustadas resultantes del ajuste restringido sean más precisas que los valores observados y que las coordenadas ajustadas sean más precisas que las del ajuste de restricción mínima.
- Controlar o amortiguar el efecto acumulativo de los errores observacionales (accidentales) en las coordenadas ajustadas.

Algunas de las desventajas de los ajustes restringidos incluyen las siguientes:

- Los puntos de control fijados (con sus coordenadas asumidas como correctas y sin errores) para el ajuste pueden ser inadecuados en el sentido de que pueden llevar a cierres incorrectos que pueden ser mucho más grandes de lo esperado. Esto es generalmente el caso ya que los cierres incorrectos en tales ajustes a menudo se asocian con errores en los nuevos levantamientos (que pueden estar mal) en lugar de en los puntos de control fijados. La consecuencia de esta asociación es que la diferencia entre los valores observados y ajustados de las observaciones puede ser mucho mayor de lo que se puede explicar por error observacional (accidental).

- Cada ajuste restringido en el cual los puntos de control se mantienen fijos (teniendo coordenadas con varianzas cero) es subóptimo. Esto significa que los ajustes de mínimos cuadrados utilizando una matriz de pesos que no es inversamente proporcional a la matriz de covarianza de las observaciones no tienen la propiedad de varianza mínima (son subóptimos). Note que las coordenadas de los puntos de control fijados son pseudo-observaciones, pero sus matrices de covarianza no se utilizan en su ponderación ya que tienen valores cero como sus elementos. Por ejemplo, un levantamiento GPS de alta precisión puede distorsionarse o degradarse al ajustarse a puntos de control de menor calidad. En este caso, la matriz de covarianza de las observaciones ajustadas puede no ser menor que la de las observaciones reales como debería ser. Al fijar los puntos de control, los valores ajustados de las observaciones probablemente se volverán peores que las observaciones reales. Por ejemplo, se deben considerar los efectos de errores en los puntos de control fijados al ajustar vectores GPS que son precisos a 1 ppm en puntos de control fijados basados en la red original NAD83 que tiene una precisión de aproximadamente 1:300,000. Esto significa que las covarianzas de los puntos de control fijados deben considerarse en el ajuste para tener en cuenta los efectos de los errores de los puntos de control fijados.

Los ajustes restringidos suelen distorsionar las observaciones ya que producen observaciones ajustadas que son más inciertas o probablemente tengan mayores errores que las observaciones reales. Por esta razón, cuando los topógrafos procesan, por ejemplo, sus vectores de línea base GPS, prefieren el ajuste de restricción mínima al ajuste restringido, ya que los ajustes de restricción mínima no distorsionan las mediciones después del ajuste. El enfoque más correcto para ajustar las mediciones de levantamientos es realizar un ajuste restringido por peso, que considera tanto los errores en los nuevos levantamientos como en los puntos de control. Este enfoque, sin embargo, tiene el problema de incluir las coordenadas de los puntos de control para el ajuste, cambiando así sus valores después del ajuste. Dado que los topógrafos no quieren que las coordenadas de los puntos de control sigan cambiando en cada ajuste, la opción más aceptable para los topógrafos sigue siendo mantener los puntos de control fijos (definiendo el datum) y ajustar los nuevos levantamientos en relación con ellos.

El tipo de ajuste que se debe realizar en las estaciones de levantamiento de un nuevo levantamiento depende de los tipos de restricciones impuestas a las coordenadas de las estaciones de levantamiento. Las coordenadas de las estaciones de levantamiento pueden definirse como fijas (o absolutamente restringidas), restringidas por peso o libres. Las coordenadas de estaciones fijas (o absolutamente restringidas) tendrán coordenadas que permanecerán iguales después del ajuste ya que sus coordenadas no están abiertas a ningún ajuste. Se dice que estas estaciones tienen coordenadas con pesos infinitos o una matriz de cofactores con valores cero como sus elementos. Cuando las estaciones de levantamiento se fijan para definir un datum, esas estaciones fijas se consideran como una base de referencia de varianza cero. En este caso, la matriz de covarianza de las coordenadas estimadas dependerá de la elección de la base de referencia de varianza cero (datum). Sin embargo, es posible utilizar transformaciones de similitud para transformar una matriz de cofactores de un datum o base de referencia de varianza cero a otro en una base de referencia de varianza cero diferente. Los pasos de transformación se discuten más adelante en el Capítulo 14. Si las coordenadas iniciales de una estación de levantamiento (antes del ajuste) están asociadas con algunas desviaciones estándar, la estación se considera una estación restringida por peso. Esto significa que las coordenadas de las estaciones restringidas por peso no se conocen lo suficientemente bien como para mantenerse fijas y, por lo tanto, están abiertas a ajustes en la medida limitada por las desviaciones estándar asociadas de las coordenadas de la estación; estas coordenadas de la estación pueden usarse como pseudo-observaciones con sus desviaciones estándar como factor de ponderación. Las desviaciones estándar de los puntos ponderados mejorarán después del ajuste y también pueden mejorar otros parámetros a través de la correlación. El ajuste de estaciones ponderadas también se conoce como mínimos cuadrados bayesianos.

Por lo general, una estimación de mínimos cuadrados basada en coordenadas iniciales que están asociadas con una matriz de cofactores (**restricciones ponderadas**) dará resultados de ajuste que son casi idénticos a los obtenidos por el ajuste de **restricción mínima**; los valores ajustados basados en este método se encuentran en algún punto entre los valores derivados utilizando el método sobrerrestringido y aquellos basados en el método de restricción mínima. Esto puede interpretarse como que los datos previamente estimados en un ajuste no distorsionan significativamente las nuevas mediciones. En este caso, la geometría (posición relativa de los puntos) está definida por las mediciones, y no se permite que las restricciones de peso afecten la geometría. Cuando las coordenadas de las estaciones de levantamiento son desconocidas y están abiertas a un ajuste ilimitado o no restringido, las estaciones se denominan **estaciones libres**.

### 4.8 Comparación de Diferentes Métodos de Ajuste

En tiempos recientes, los profesionales de la geomática se han interesado en cómo los métodos tradicionales de ajuste de travesías se comparan con el método de mínimos cuadrados. Los métodos tradicionales de ajuste bien conocidos en topografía son los métodos de tránsito, Bowditch y Crandall. Otro algoritmo tradicional de ajuste de travesías conocido como el método de Tuttle fue discutido y comparado con los procedimientos de Bowditch y Crandall en Sprinsky (1987). Aunque los métodos tradicionales de ajuste son inadecuados para ajustar redes de levantamiento de alta precisión que involucran redes de trilateración y triangulación, siguen siendo muy atractivos para los topógrafos de tierras que se especializan en el establecimiento de límites de propiedad. Esta sección utiliza dos ejemplos (Ejemplos 4.4 y 4.5) para explicar los supuestos y las relaciones entre los métodos tradicionales de ajuste y el método de ajuste de mínimos cuadrados. Se espera que los profesionales de la geomática aprecien las limitaciones de los diferentes métodos de ajuste y cómo los métodos están relacionados a través de los ejemplos dados. Además, se anima a los profesionales de la geomática a verificar cualquier resultado de ajuste obtenido a través de software comercial realizando cálculos rutinarios de travesías manualmente en lugar de simplemente aceptar valores proporcionados por algunos paquetes de software. Los resultados de ajuste basados en los métodos de tránsito, Crandall, Bowditch y mínimos cuadrados se presentan, y algunos aspectos de los métodos tradicionales se comparan con el método de ajuste de mínimos cuadrados.

Para cuantificar el efecto de cualquier ajuste basado en cualquier método, se analizan las observaciones ajustadas (distancias y ángulos) en lugar de las coordenadas planas calculadas a partir de esas observaciones. Las correcciones a las observaciones deben basarse en la calidad de las observaciones; grandes correcciones a observaciones de alta calidad indican un posible error sistemático en las observaciones. Los pasos para cuantificar el efecto de cualquier ajuste pueden darse de la siguiente manera:

- Calcular las coordenadas a partir de las mediciones de campo.
- Calcular ángulos y distancias mediante cálculos inversos utilizando las coordenadas calculadas.
- Comparar las cantidades calculadas inversamente con las mediciones reales para ver cuánto se desvían de las mediciones reales.

### 4.8.1 Discusiones Generales

A partir de la Tabla 4.9, se puede observar que las correcciones previas al ajuste son las mismas que los residuos del método de Crandall. Esto es consistente con la idea de que el método de Crandall no ajusta ángulos (rumbo); los residuos se deben en realidad a las correcciones previas al ajuste aplicadas a los ángulos. Esto es diferente en el caso del ajuste de red de itinerarios alternos, como se muestra en la Tabla 4.19. Se puede ver en la Tabla 4.19 que los residuos del método de Crandall son diferentes de las correcciones previas al ajuste de los ángulos, lo que parece sugerir que el método de Crandall ajustó los rumbos; esto se debe al punto común R. Sin embargo, si el punto R se trata como dos puntos diferentes en el ajuste, los residuos serán los mismos que las correcciones previas al ajuste. Se puede ver que aquellos puntos conectados al punto R, como Q, S, U y W, tienen residuos diferentes, pero los puntos no conectados T y P tienen residuos que son iguales a sus correcciones previas al ajuste. En general, a partir de los dos ejemplos dados, se puede ver que el ajuste máximo de distancia esperado en un promedio de 200 m está dentro de ±10 mm en el método tradicional (en el método de Crandall) y está dentro de ±4 mm usando el método de mínimos cuadrados. Los ajustes angulares están dentro de ±10'' en el método tradicional (en el método de la brújula) y dentro de ±7'' usando el método de mínimos cuadrados. Las coordenadas ajustadas de los tres métodos tradicionales varían de las de los mínimos cuadrados dentro de ±5 mm. Aunque los resultados de los tres métodos tradicionales adicionales son comparables con los del método de mínimos cuadrados, hay una serie de ventajas significativas del método de mínimos cuadrados sobre los métodos tradicionales, las cuales incluyen lo siguiente:

1. Tiene una ventaja sobre los otros métodos, ya que se basa en la teoría de la probabilidad matemática, lo que lo hace más riguroso. También proporciona las correcciones más pequeñas a las mediciones después de ajustar las observaciones juntas sin cierre cero.
2. Permite que todas las observaciones se ponderen adecuadamente usando sus precisiones esperadas; los otros métodos no permiten esto.
3. El mismo procedimiento se sigue al resolver muchos diferentes tipos de problemas geomáticos.
4. Permite que se realicen análisis estadísticos completos después de los ajustes, lo cual incluye determinar las precisiones esperadas de las cantidades ajustadas. Dado que los estándares actuales de precisión de las encuestas se basan en los conceptos de estimación de mínimos cuadrados, los topógrafos solo pueden probar adecuadamente sus observaciones en cumplimiento con los estándares de precisión ajustando sus datos usando el método de estimación de mínimos cuadrados.
5. Los sistemas de información geográfica (GIS) dependen en gran medida de los procedimientos rigurosos de ajuste y análisis estadístico de los métodos de estimación de mínimos cuadrados.
6. Se puede usar para realizar la planificación previa a la encuesta, como en el análisis previo y el diseño de la red.

Las desventajas del método de mínimos cuadrados ordinarios son generalmente con respecto a otros métodos de ajuste, como la colocalización de mínimos cuadrados. En este caso, el método de mínimos cuadrados ordinarios se puede considerar como teniendo propiedades de suavizado excesivo cuando se aplica a problemas de predicción o interpolación. El método de mínimos cuadrados ordinarios (también conocido como norma L₂) también puede ser débil comparado con otro método de estimación como el método de norma L₁ (también conocido como alternativa robusta a mínimos cuadrados, que minimiza la suma de valores absolutos de residuos o residuos ponderados) (Huber 1964; Barrodale y Roberts 1974; Harvey 1993; Bektas y Sisman 2010) cuando se trata de observaciones que pueden ser afectadas por valores atípicos. En este caso, no se puede considerar ningún método de ajuste particular como el mejor para todas las aplicaciones. Los siguientes comentarios generales se pueden hacer con respecto a los errores de cierre del levantamiento de itinerario tradicional:

1. La relación del error de cierre con la longitud del itinerario ha sido útil para indicar la existencia de algunos errores graves y ha sido un indicador de la precisión del itinerario. Sin embargo, no hay garantía de que todas las mediciones estén libres de errores graves o que la geometría del itinerario sea tal que todas las estaciones estén dentro de una precisión relativa deseada entre sí.
2. La relación del error de cierre con la longitud del itinerario (por ejemplo, 1:5000) se ha utilizado tradicionalmente para determinar la precisión del trabajo de levantamiento legal. Esta relación tiene limitaciones para encuestas muy grandes y encuestas que involucran líneas muy cortas. En levantamientos de itinerarios muy largos, el error permitido puede ser engañoso y, en algunos casos, puede ocultar un error grave. Para una longitud de itinerario de 100 m, por ejemplo, el cierre lineal esperado será de 0,02 m, lo que puede ser difícil de cumplir según la especificación de cierre de 1/5000.
3. En los sistemas de información geográfica y de información de tierras donde se integran varios conjuntos de datos, es necesario conocer la precisión posicional. Por lo tanto, ya no es suficiente evaluar la precisión de un levantamiento legal en términos de la relación del error de cierre con la longitud del itinerario. Además, no hay una correlación simple entre las precisiones relativas del índice de cierre y las precisiones posicionales radiales del 95%.
